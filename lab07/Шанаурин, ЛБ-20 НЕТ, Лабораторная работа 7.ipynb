```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display, Markdown

# –ß—Ç–µ–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö –∏–∑ —Ñ–∞–π–ª–∞ CSV
data = pd.read_csv('–í–∞—Ä–∏–∞–Ω—Ç 26.3.csv')

display(Markdown(r"### –ò–º–ø–æ—Ä—Ç –¥–∞–Ω–Ω—ã—Ö –∏ –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ —Ç–∞–±–ª–∏—Ü—ã —Å–æ–ø—Ä—è–∂–µ–Ω–Ω–æ—Å—Ç–∏"))

# –í—ã–≤–æ–¥ –ø–µ—Ä–≤—ã—Ö —Å—Ç—Ä–æ–∫ –¥–∞–Ω–Ω—ã—Ö –¥–ª—è –æ–∑–Ω–∞–∫–æ–º–ª–µ–Ω–∏—è
display(data.head())

# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ —Ç–∞–±–ª–∏—Ü—ã —Å–æ–ø—Ä—è–∂–µ–Ω–Ω–æ—Å—Ç–∏
contingency_table = pd.crosstab(data['–í–æ–ø—Ä–æ—Å 1'], data['–í–æ–ø—Ä–æ—Å 2'])
print("\n–¢–∞–±–ª–∏—Ü–∞ —Å–æ–ø—Ä—è–∂–µ–Ω–Ω–æ—Å—Ç–∏:")
display(contingency_table)

# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —Ç–∞–±–ª–∏—Ü—ã —Å–æ–ø—Ä—è–∂–µ–Ω–Ω–æ—Å—Ç–∏
plt.figure(figsize=(8, 6))
sns.heatmap(contingency_table, annot=True, cmap="YlGnBu", fmt='d')
plt.title('–¢–∞–±–ª–∏—Ü–∞ —Å–æ–ø—Ä—è–∂–µ–Ω–Ω–æ—Å—Ç–∏ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ X –∏ Y')
plt.xlabel('–í–æ–ø—Ä–æ—Å 2 (–ø—Ä–∏–∑–Ω–∞–∫ Y)')
plt.ylabel('–í–æ–ø—Ä–æ—Å 1 (–ø—Ä–∏–∑–Ω–∞–∫ X)')
plt.tight_layout()

# –ü—Ä–æ–≤–µ—Ä–∫–∞ –æ–∂–∏–¥–∞–µ–º—ã—Ö —á–∞—Å—Ç–æ—Ç
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)

display(Markdown(r"#### –û–∂–∏–¥–∞–µ–º—ã–µ —á–∞—Å—Ç–æ—Ç—ã:"))
print(pd.DataFrame(expected, index=contingency_table.index, columns=contingency_table.columns))

# –ü—Ä–æ–≤–µ—Ä–∫–∞ —É—Å–ª–æ–≤–∏–π –ø—Ä–∏–º–µ–Ω–∏–º–æ—Å—Ç–∏ –∫—Ä–∏—Ç–µ—Ä–∏—è —Ö–∏-–∫–≤–∞–¥—Ä–∞—Ç
expected_less_than_5 = (expected < 5).sum()
total_cells = expected.size
percent_less_than_5 = (expected_less_than_5 / total_cells) * 100
min_expected = expected.min()

print(f"\n–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —è—á–µ–µ–∫ —Å –æ–∂–∏–¥–∞–µ–º–æ–π —á–∞—Å—Ç–æ—Ç–æ–π < 5: {expected_less_than_5} –∏–∑ {total_cells} ({percent_less_than_5:.2f}%)")
print(f"–ú–∏–Ω–∏–º–∞–ª—å–Ω–∞—è –æ–∂–∏–¥–∞–µ–º–∞—è —á–∞—Å—Ç–æ—Ç–∞: {min_expected:.2f}")

# –†–∞—Å—á–µ—Ç –∫—Ä–∏—Ç–µ—Ä–∏—è —Ö–∏-–∫–≤–∞–¥—Ä–∞—Ç
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)

print(f"\n–ó–Ω–∞—á–µ–Ω–∏–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏ —Ö–∏-–∫–≤–∞–¥—Ä–∞—Ç: {chi2:.4f}")
print(f"–°—Ç–µ–ø–µ–Ω–∏ —Å–≤–æ–±–æ–¥—ã: {dof}")
print(f"p-–∑–Ω–∞—á–µ–Ω–∏–µ: {p:.4f}")

# –ö—Ä–∏—Ç–∏—á–µ—Å–∫–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —Ö–∏-–∫–≤–∞–¥—Ä–∞—Ç –¥–ª—è —É—Ä–æ–≤–Ω—è –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ 0.05
alpha = 0.05
critical_value = stats.chi2.ppf(1 - alpha, dof)
print(f"–ö—Ä–∏—Ç–∏—á–µ—Å–∫–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —Ö–∏-–∫–≤–∞–¥—Ä–∞—Ç (Œ± = {alpha}): {critical_value:.4f}")

# –í—ã–≤–æ–¥ –æ –Ω–∞–ª–∏—á–∏–∏ –≤–∑–∞–∏–º–æ—Å–≤—è–∑–∏
if p < alpha:
    print("\n–ù—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ –æ–± –æ—Ç—Å—É—Ç—Å—Ç–≤–∏–∏ —Å–≤—è–∑–∏ –æ—Ç–≤–µ—Ä–≥–∞–µ—Ç—Å—è.")
    print("–°—É—â–µ—Å—Ç–≤—É–µ—Ç —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º–∞—è —Å–≤—è–∑—å –º–µ–∂–¥—É –ø—Ä–∏–∑–Ω–∞–∫–∞–º–∏ X –∏ Y.")
else:
    print("\n–ù—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞ –æ–± –æ—Ç—Å—É—Ç—Å—Ç–≤–∏–∏ —Å–≤—è–∑–∏ –Ω–µ –æ—Ç–≤–µ—Ä–≥–∞–µ—Ç—Å—è.")
    print("–ù–µ—Ç —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º–æ–π —Å–≤—è–∑–∏ –º–µ–∂–¥—É –ø—Ä–∏–∑–Ω–∞–∫–∞–º–∏ X –∏ Y.")

# –†–∞—Å—á–µ—Ç –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–∞ –ö—Ä–∞–º–µ—Ä–∞
n = contingency_table.sum().sum()  # –æ–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–∞–±–ª—é–¥–µ–Ω–∏–π
r = len(contingency_table)         # –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç—Ä–æ–∫
k = len(contingency_table.columns) # –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç–æ–ª–±—Ü–æ–≤
min_dim = min(r - 1, k - 1)        # –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ –∏–∑–º–µ—Ä–µ–Ω–∏–µ

cramer_v = np.sqrt(chi2 / (n * min_dim))
print(f"\n–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –ö—Ä–∞–º–µ—Ä–∞ V: {cramer_v:.4f}")

# –ò–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è —Å–∏–ª—ã —Å–≤—è–∑–∏
if cramer_v < 0.1:
    strength = "–Ω–µ—Å—É—â–µ—Å—Ç–≤–µ–Ω–Ω–∞—è"
elif cramer_v < 0.3:
    strength = "—Å–ª–∞–±–∞—è"
elif cramer_v < 0.5:
    strength = "—Å—Ä–µ–¥–Ω—è—è"
else:
    strength = "—Å–∏–ª—å–Ω–∞—è"

print(f"–°–∏–ª–∞ —Å–≤—è–∑–∏: {strength}")

display(Markdown(r"""
#### –ò–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ –∏ –≤—ã–≤–æ–¥—ã

–ù–∞ –æ—Å–Ω–æ–≤–∞–Ω–∏–∏ –ø—Ä–æ–≤–µ–¥–µ–Ω–Ω–æ–≥–æ –∞–Ω–∞–ª–∏–∑–∞ –º–æ–∂–Ω–æ —Å–¥–µ–ª–∞—Ç—å —Å–ª–µ–¥—É—é—â–∏–µ –≤—ã–≤–æ–¥—ã:

- –ü—Ä–∏–º–µ–Ω–∏–º–æ—Å—Ç—å –∫—Ä–∏—Ç–µ—Ä–∏—è —Ö–∏-–∫–≤–∞–¥—Ä–∞—Ç: –ö—Ä–∏—Ç–µ—Ä–∏–π —Ö–∏-–∫–≤–∞–¥—Ä–∞—Ç –ø—Ä–∏–º–µ–Ω–∏–º –∫ –¥–∞–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ, —Ç–∞–∫ –∫–∞–∫ –≤—Å–µ –æ–∂–∏–¥–∞–µ–º—ã–µ —á–∞—Å—Ç–æ—Ç—ã –≤ —Ç–∞–±–ª–∏—Ü–µ —Å–æ–ø—Ä—è–∂–µ–Ω–Ω–æ—Å—Ç–∏ –ø—Ä–µ–≤—ã—à–∞—é—Ç 5, —á—Ç–æ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É–µ—Ç —Ç—Ä–µ–±–æ–≤–∞–Ω–∏—è–º –¥–ª—è –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ–≥–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è –¥–∞–Ω–Ω–æ–≥–æ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–≥–æ –º–µ—Ç–æ–¥–∞.
- –ù–∞–ª–∏—á–∏–µ –≤–∑–∞–∏–º–æ—Å–≤—è–∑–∏: –†–∞—Å—á–µ—Ç –∫—Ä–∏—Ç–µ—Ä–∏—è —Ö–∏-–∫–≤–∞–¥—Ä–∞—Ç –ø–æ–∫–∞–∑–∞–ª —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º—É—é –≤–∑–∞–∏–º–æ—Å–≤—è–∑—å –º–µ–∂–¥—É –ø—Ä–∏–∑–Ω–∞–∫–∞–º–∏ X –∏ Y (–≤–æ–ø—Ä–æ—Å—ã 1 –∏ 2). –ó–Ω–∞—á–µ–Ω–∏–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏ —Ö–∏-–∫–≤–∞–¥—Ä–∞—Ç —Å–æ—Å—Ç–∞–≤–∏–ª–æ 31.7563, —á—Ç–æ –ø—Ä–µ–≤—ã—à–∞–µ—Ç –∫—Ä–∏—Ç–∏—á–µ—Å–∫–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ 5.9915 –¥–ª—è —É—Ä–æ–≤–Ω—è –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ 0.05. –ü–æ–ª—É—á–µ–Ω–Ω–æ–µ p-–∑–Ω–∞—á–µ–Ω–∏–µ —Ä–∞–≤–Ω–æ 0.0000, —á—Ç–æ –∑–Ω–∞—á–∏—Ç–µ–ª—å–Ω–æ –º–µ–Ω—å—à–µ —É—Ä–æ–≤–Ω—è –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ 0.05. –≠—Ç–æ –ø–æ–∑–≤–æ–ª—è–µ—Ç –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É –æ–± –æ—Ç—Å—É—Ç—Å—Ç–≤–∏–∏ —Å–≤—è–∑–∏ –º–µ–∂–¥—É –ø—Ä–∏–∑–Ω–∞–∫–∞–º–∏.
- –°–∏–ª–∞ —Å–≤—è–∑–∏: –ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –ö—Ä–∞–º–µ—Ä–∞ V, –∏—Å–ø–æ–ª—å–∑—É–µ–º—ã–π –¥–ª—è –æ—Ü–µ–Ω–∫–∏ —Å–∏–ª—ã —Å–≤—è–∑–∏ –º–µ–∂–¥—É –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–º–∏ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–º–∏, —Å–æ—Å—Ç–∞–≤–∏–ª 0.5499. –°–æ–≥–ª–∞—Å–Ω–æ –æ–±—â–µ–ø—Ä–∏–Ω—è—Ç–æ–π –∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏–∏, —ç—Ç–æ –∑–Ω–∞—á–µ–Ω–∏–µ —É–∫–∞–∑—ã–≤–∞–µ—Ç –Ω–∞ —Å—Ä–µ–¥–Ω—é—é —Å–∏–ª—É —Å–≤—è–∑–∏ –º–µ–∂–¥—É –∏–∑—É—á–∞–µ–º—ã–º–∏ –ø—Ä–∏–∑–Ω–∞–∫–∞–º–∏.
- –ò–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è —Ç–∞–±–ª–∏—Ü—ã —Å–æ–ø—Ä—è–∂–µ–Ω–Ω–æ—Å—Ç–∏: –ê–Ω–∞–ª–∏–∑–∏—Ä—É—è —Ç–∞–±–ª–∏—Ü—É —Å–æ–ø—Ä—è–∂–µ–Ω–Ω–æ—Å—Ç–∏, –º–æ–∂–Ω–æ –∑–∞–º–µ—Ç–∏—Ç—å, —á—Ç–æ —É –∏—Å–ø—ã—Ç—É–µ–º—ã—Ö —Å –æ—Ç–≤–µ—Ç–æ–º A –Ω–∞ –ø–µ—Ä–≤—ã–π –≤–æ–ø—Ä–æ—Å –∑–Ω–∞—á–∏—Ç–µ–ª—å–Ω–æ —á–∞—â–µ –≤—Å—Ç—Ä–µ—á–∞–µ—Ç—Å—è –æ—Ç–≤–µ—Ç a –Ω–∞ –≤—Ç–æ—Ä–æ–π –≤–æ–ø—Ä–æ—Å (40 –Ω–∞–±–ª—é–¥–µ–Ω–∏—è –ø—Ä–æ—Ç–∏–≤ –æ–∂–∏–¥–∞–µ–º—ã—Ö 25.667), –≤ —Ç–æ –≤—Ä–µ–º—è –∫–∞–∫ —É –∏—Å–ø—ã—Ç—É–µ–º—ã—Ö —Å –æ—Ç–≤–µ—Ç–æ–º B –Ω–∞ –ø–µ—Ä–≤—ã–π –≤–æ–ø—Ä–æ—Å —á–∞—â–µ –≤—Å—Ç—Ä–µ—á–∞—é—Ç—Å—è –æ—Ç–≤–µ—Ç—ã b –∏ c –Ω–∞ –≤—Ç–æ—Ä–æ–π –≤–æ–ø—Ä–æ—Å (17 –∏ 24 –Ω–∞–±–ª—é–¥–µ–Ω–∏–π —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ –ø—Ä–æ—Ç–∏–≤ –æ–∂–∏–¥–∞–µ–º—ã—Ö 10.47619 –∏ 16.190476).

–¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, –º–µ–∂–¥—É –ø—Ä–∏–∑–Ω–∞–∫–∞–º–∏ X –∏ Y —Å—É—â–µ—Å—Ç–≤—É–µ—Ç —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏ –∑–Ω–∞—á–∏–º–∞—è –≤–∑–∞–∏–º–æ—Å–≤—è–∑—å —Å—Ä–µ–¥–Ω–µ–π —Å–∏–ª—ã, —á—Ç–æ –≥–æ–≤–æ—Ä–∏—Ç –æ —Ç–æ–º, —á—Ç–æ –æ—Ç–≤–µ—Ç—ã –∏—Å–ø—ã—Ç—É–µ–º—ã—Ö –Ω–∞ –ø–µ—Ä–≤—ã–π –≤–æ–ø—Ä–æ—Å —Ç–µ—Å—Ç–∞ –≤ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω–æ–π —Å—Ç–µ–ø–µ–Ω–∏ —Å–≤—è–∑–∞–Ω—ã —Å –∏—Ö –æ—Ç–≤–µ—Ç–∞–º–∏ –Ω–∞ –≤—Ç–æ—Ä–æ–π –≤–æ–ø—Ä–æ—Å.
"""))
```
