```python
features_of_interest = ['BEDS', 'BATH', 'PROPERTYSQFT', 'PRICE']
df_analysis = df[features_of_interest].copy()

# Проверим на нормальность распределения каждого признака
from scipy.stats import shapiro
from scipy.stats import pearsonr, spearmanr, kendalltau

print("\nТест Шапиро-Уилка на нормальность распределения:")
for column in df_analysis.columns:
    # Берём выборку из 5000 элементов (или меньше, если данных меньше)
    # из-за ограничений теста Шапиро-Уилка
    sample = df_analysis[column].sample(min(5000, len(df_analysis)))
    stat, p = shapiro(sample)
    print(f"{column}: статистика={stat:.4f}, p-value={p:.4f}")
    if p < 0.05:
        print(f"  Распределение {column} не является нормальным (p < 0.05)")
    else:
        print(f"  Распределение {column} может быть нормальным (p >= 0.05)")

# Визуализация распределений с помощью гистограмм
plt.figure(figsize=(14, 10))
for i, column in enumerate(df_analysis.columns):
    plt.subplot(2, 2, i+1)
    sns.histplot(df_analysis[column], kde=True)
    plt.title(f'Распределение {column}')
plt.tight_layout()
plt.show()
# plt.savefig('distributions.png')
# plt.close()

# Матрица корреляций Пирсона (параметрическая - предполагает нормальное распределение)
pearson_corr = df_analysis.corr(method='pearson')
print("\nМатрица корреляций Пирсона:")
print(pearson_corr)

# Матрица корреляций Спирмена (непараметрическая - основана на рангах)
spearman_corr = df_analysis.corr(method='spearman')
print("\nМатрица корреляций Спирмена:")
print(spearman_corr)

# Расчет корреляции Кендалла τ (тау) для каждой пары переменных
kendall_corr_matrix = df_analysis.corr(method='kendall')
print("\nМатрица корреляций Кендалла τ:")
print(kendall_corr_matrix)

# Визуализация матриц корреляции
plt.figure(figsize=(20, 6))

plt.subplot(1, 3, 1)
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Корреляция Пирсона')

plt.subplot(1, 3, 2)
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Корреляция Спирмена')

plt.subplot(1, 3, 3)
sns.heatmap(kendall_corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Корреляция Кендалла τ')

plt.tight_layout()
plt.show()
# plt.savefig('correlation_matrices.png')
# plt.close()

# Детальный анализ корреляций с их p-значениями
def calculate_correlation_with_p_values(df):
    """Рассчитывает корреляцию Пирсона, Спирмена и Кендалла с p-значениями."""
    results = []
    for i, col1 in enumerate(df.columns):
        for j, col2 in enumerate(df.columns):
            if i < j:  # Только для пар признаков, чтобы избежать дубликатов
                # Корреляция Пирсона
                pearson_corr, pearson_p = pearsonr(df[col1], df[col2])
                # Корреляция Спирмена
                spearman_corr, spearman_p = spearmanr(df[col1], df[col2])
                # Корреляция Кендалла
                kendall_corr, kendall_p = kendalltau(df[col1], df[col2])
                
                results.append({
                    'Признак 1': col1,
                    'Признак 2': col2,
                    'Корреляция Пирсона': pearson_corr,
                    'P-значение (Пирсон)': pearson_p,
                    'Корреляция Спирмена': spearman_corr,
                    'P-значение (Спирмен)': spearman_p,
                    'Корреляция Кендалла': kendall_corr,
                    'P-значение (Кендалл)': kendall_p
                })
    return pd.DataFrame(results)

correlation_details = calculate_correlation_with_p_values(df_analysis)
print("\nДетальный анализ корреляций с p-значениями:")
print(correlation_details)

# Создание пар графиков рассеяния
plt.figure(figsize=(12, 10))
sns.pairplot(df_analysis, diag_kind='kde', plot_kws={'alpha': 0.6})
plt.suptitle('Парные графики рассеяния между BEDS, BATH, PROPERTYSQFT и PRICE', y=1.02)
plt.show()
# plt.savefig('pairplot.png')
# plt.close()

# Альтернативный вариант с использованием scatter_matrix
pd.plotting.scatter_matrix(df_analysis, figsize=(12, 10), alpha=0.8, diagonal='kde')
plt.suptitle('Матрица графиков рассеяния между BEDS, BATH, PROPERTYSQFT и PRICE', y=1.02)
plt.tight_layout()
plt.show()
# plt.savefig('scatter_matrix.png')
# plt.close()

# Визуализация зависимостей между признаками и целевой переменной PRICE
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

axes[0].scatter(df_analysis['BEDS'], df_analysis['PRICE'], alpha=0.5)
axes[0].set_title('Зависимость PRICE от BEDS')
axes[0].set_xlabel('BEDS')
axes[0].set_ylabel('PRICE')

axes[1].scatter(df_analysis['BATH'], df_analysis['PRICE'], alpha=0.5)
axes[1].set_title('Зависимость PRICE от BATH')
axes[1].set_xlabel('BATH')
axes[1].set_ylabel('PRICE')

axes[2].scatter(df_analysis['PROPERTYSQFT'], df_analysis['PRICE'], alpha=0.5)
axes[2].set_title('Зависимость PRICE от PROPERTYSQFT')
axes[2].set_xlabel('PROPERTYSQFT')
axes[2].set_ylabel('PRICE')

plt.tight_layout()
plt.show()
# plt.savefig('price_dependencies.png')
# plt.close()

# Дополнительная визуализация: регрессионные графики с seaborn
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

sns.regplot(x='BEDS', y='PRICE', data=df_analysis, ax=axes[0], scatter_kws={'alpha':0.3})
axes[0].set_title('Регрессия PRICE ~ BEDS')

sns.regplot(x='BATH', y='PRICE', data=df_analysis, ax=axes[1], scatter_kws={'alpha':0.3})
axes[1].set_title('Регрессия PRICE ~ BATH')

sns.regplot(x='PROPERTYSQFT', y='PRICE', data=df_analysis, ax=axes[2], scatter_kws={'alpha':0.3})
axes[2].set_title('Регрессия PRICE ~ PROPERTYSQFT')

plt.tight_layout()
plt.show()
# plt.savefig('regression_plots.png')
# plt.close()

# Анализ сильных и слабых связей
print("\nАнализ силы связей между признаками:")
for i, row in correlation_details.iterrows():
    feature1 = row['Признак 1']
    feature2 = row['Признак 2']
    
    # Определяем, какой коэффициент корреляции больше подходит
    is_normal1 = shapiro(df_analysis[feature1].sample(min(5000, len(df_analysis))))[1] >= 0.05
    is_normal2 = shapiro(df_analysis[feature2].sample(min(5000, len(df_analysis))))[1] >= 0.05
    
    if is_normal1 and is_normal2:
        preferred_corr = 'Пирсона'
        corr_value = row['Корреляция Пирсона']
        p_value = row['P-значение (Пирсон)']
    else:
        preferred_corr = 'Спирмена'
        corr_value = row['Корреляция Спирмена']
        p_value = row['P-значение (Спирмен)']
    
    print(f"\n{feature1} и {feature2}:")
    print(f"  Рекомендуемый коэффициент корреляции: {preferred_corr}")
    print(f"  Значение корреляции: {corr_value:.4f}, p-value: {p_value:.4e}")
    
    # Интерпретация силы связи
    abs_corr = abs(corr_value)
    if abs_corr < 0.3:
        strength = "слабая"
    elif abs_corr < 0.7:
        strength = "средняя"
    else:
        strength = "сильная"
    
    direction = "положительная" if corr_value > 0 else "отрицательная"
    significance = "статистически значимая" if p_value < 0.05 else "статистически незначимая"
    
    print(f"  Интерпретация: {strength} {direction} связь, {significance}")
```

### 1. Обоснование выбора метода корреляционного анализа
При выборе метода корреляционного анализа необходимо учитывать распределение данных и характер взаимосвязей между признаками.

Согласно результатам теста Шапиро-Уилка, все исследуемые признаки (BEDS, BATH, PROPERTYSQFT и PRICE) не имеют нормального распределения (p < 0.05 для всех признаков). В таких случаях применение корреляции Пирсона может привести к искажению результатов, поскольку этот метод предполагает нормальность распределения данных.

#### Обоснование выбора метода для каждой пары признаков:

##### BEDS и BATH:
- Оба признака дискретные (целочисленные)
- Распределения не являются нормальными
- Рекомендуемый метод: ранговая корреляция Спирмена
##### BEDS и PROPERTYSQFT:
- BEDS - дискретный признак, PROPERTYSQFT - непрерывный
- Распределения не являются нормальными
- Возможно наличие нелинейной зависимости
- Рекомендуемый метод: ранговая корреляция Спирмена
##### BATH и PROPERTYSQFT:
- BATH - дискретный признак, PROPERTYSQFT - непрерывный
- Распределения не являются нормальными
- Возможно наличие нелинейной зависимости
- Рекомендуемый метод: ранговая корреляция Спирмена
##### BEDS и PRICE:
- Распределения не являются нормальными
- Возможно наличие выбросов в цене
- Рекомендуемый метод: ранговая корреляция Спирмена
##### BATH и PRICE:
- Распределения не являются нормальными
- Возможно наличие выбросов в цене
- Рекомендуемый метод: ранговая корреляция Спирмена
##### PROPERTYSQFT и PRICE:
- Распределения не являются нормальными

```python
plt.rcParams['figure.figsize'] = (14, 10)

# Функция для построения графика распределения категориального признака
def plot_categorical_distribution(data, column, title, ax, threshold_percent=1):
    # Подсчет частот категорий
    value_counts = data[column].value_counts()
    
    # Вычисление порога для малочисленных категорий (в процентах от общего числа)
    threshold = len(data) * threshold_percent / 100
    
    # Выделение малочисленных категорий
    rare_categories = value_counts[value_counts < threshold].index.tolist()
    
    # Вывод информации о малочисленных категориях
    if rare_categories:
        print(f"\nМалочисленные категории в {column} (менее {threshold_percent}% наблюдений):")
        for cat in rare_categories:
            print(f"- {cat}: {value_counts[cat]} ({value_counts[cat]/len(data)*100:.2f}%)")
    
    # Подготовка данных для графика
    total_records = len(data)
    categories = value_counts.index
    counts = value_counts.values
    percentages = (counts / total_records) * 100
    
    # Построение графика
    sns.barplot(x=categories, y=counts, ax=ax, alpha=0.8)
    
    # Добавление процентных значений над столбцами
    for i, (count, percentage) in enumerate(zip(counts, percentages)):
        if percentage < 5:  # Для малых значений размещаем текст над столбцом
            ax.text(i, count + 5, f'{count}\n({percentage:.1f}%)', 
                    ha='center', va='bottom', fontsize=8, rotation=0)
        else:
            ax.text(i, count/2, f'{count}\n({percentage:.1f}%)', 
                    ha='center', va='center', fontsize=8, color='white')
    
    # Настройка графика
    ax.set_title(title, fontsize=14, pad=20)
    ax.set_xlabel('')
    ax.set_ylabel('Количество наблюдений', fontsize=12)
    
    # Поворот меток оси X для лучшей читаемости
    plt.setp(ax.get_xticklabels(), rotation=45, ha='right', fontsize=10)
    
    # Выделение малочисленных категорий другим цветом
    if rare_categories:
        bars = ax.patches
        for i, cat in enumerate(categories):
            if cat in rare_categories:
                bars[i].set_facecolor('red')
                bars[i].set_alpha(1.0)
    
    return rare_categories

# Создаем фигуру с подграфиками
fig, axes = plt.subplots(2, 2, figsize=(20, 16))
axes = axes.flatten()

# Анализ распределения BROKERTITLE
rare_broker = plot_categorical_distribution(
    df, 'BROKERTITLE', 'Распределение по брокерским компаниям (BROKERTITLE)', axes[0], threshold_percent=1)

# Анализ распределения TYPE
rare_type = plot_categorical_distribution(
    df, 'TYPE', 'Распределение по типам недвижимости (TYPE)', axes[1], threshold_percent=1)

# Анализ распределения LOCALITY
rare_locality = plot_categorical_distribution(
    df, 'LOCALITY', 'Распределение по местоположению (LOCALITY)', axes[2], threshold_percent=1)

# Анализ распределения SUBLOCALITY
rare_sublocality = plot_categorical_distribution(
    df, 'SUBLOCALITY', 'Распределение по районам (SUBLOCALITY)', axes[3], threshold_percent=1)

plt.tight_layout()
plt.savefig('categorical_distributions.png', dpi=300, bbox_inches='tight')
plt.show()

# Создаем отдельный отчет о малочисленных категориях
print("\n\n===== ОТЧЕТ О МАЛОЧИСЛЕННЫХ КАТЕГОРИЯХ =====")
categories = {'BROKERTITLE': rare_broker, 'TYPE': rare_type, 
              'LOCALITY': rare_locality, 'SUBLOCALITY': rare_sublocality}

for col, rare_cats in categories.items():
    if rare_cats:
        print(f"\n{col}:")
        counts = df[col].value_counts()
        for cat in rare_cats:
            count = counts[cat]
            percent = count / len(df) * 100
            print(f"  - {cat}: {count} наблюдений ({percent:.2f}%)")
    else:
        print(f"\n{col}: малочисленных категорий не обнаружено")

# Рекомендации по обработке малочисленных категорий
print("\n\n===== РЕКОМЕНДАЦИИ =====")
print("Для работы с малочисленными категориями предлагается несколько стратегий:")
print("1. Объединение малочисленных категорий в одну группу 'Other'")
print("2. Удаление наблюдений с малочисленными категориями (если доля незначительна)")
print("3. Использование методов регуляризации при построении моделей")
print("4. Применение техник one-hot encoding со снижением размерности для уменьшения влияния редких категорий")

# Дополнительный анализ: количество уникальных значений и процент заполнения
print("\n\n===== ДОПОЛНИТЕЛЬНАЯ СТАТИСТИКА =====")
for col in ['BROKERTITLE', 'TYPE', 'LOCALITY', 'SUBLOCALITY']:
    unique_count = df[col].nunique()
    null_count = df[col].isnull().sum()
    fill_rate = (len(df) - null_count) / len(df) * 100
    print(f"{col}:")
    print(f"  - Уникальных значений: {unique_count}")
    print(f"  - Заполненность: {fill_rate:.2f}%")
    print(f"  - Пропущенных значений: {null_count} ({null_count/len(df)*100:.2f}%)")

# Пример объединения малочисленных категорий
print("\n\n===== ПРИМЕР ОБРАБОТКИ МАЛОЧИСЛЕННЫХ КАТЕГОРИЙ =====")
for col, rare_cats in categories.items():
    if rare_cats:
        # Создаем копию колонки с объединенными категориями
        new_col_name = f"{col}_GROUPED"
        df[new_col_name] = df[col].copy()
        
        # Заменяем малочисленные категории на 'Other'
        df.loc[df[col].isin(rare_cats), new_col_name] = 'Other'
        
        # Выводим информацию о новом распределении
        new_counts = df[new_col_name].value_counts()
        print(f"\n{col} -> {new_col_name}:")
        print(f"  - Исходное количество категорий: {df[col].nunique()}")
        print(f"  - Новое количество категорий: {df[new_col_name].nunique()}")
        if 'Other' in new_counts:
            print(f"  - Категория 'Other': {new_counts['Other']} наблюдений ({new_counts['Other']/len(df)*100:.2f}%)")
```

При построении моделей и проведении анализа необходимо учитывать эту особенность. Проведенное укрупнение категорий (с 262 до 14 для BROKERTITLE, с 9 до 8 для TYPE и т.д.) значительно упрощает анализ без существенной потери информативности данных.

```python
from scipy.stats import chi2_contingency

# Шаг 1: Найдем топ-5 агентств по количеству продаж
def analyze_broker_locality_relationship(df):
    print("Анализ связи между агентствами (BROKERTITLE) и районами (LOCALITY)")
    
    # Находим топ-5 агентств по количеству продаж
    top5_brokers = df['BROKERTITLE'].value_counts().nlargest(5)
    print("\nТоп-5 агентств по количеству продаж:")
    for broker, count in top5_brokers.items():
        print(f"{broker}: {count} объектов")
    
    # Фильтруем данные, оставляя только топ-5 агентств
    top5_broker_names = top5_brokers.index
    df_top5 = df[df['BROKERTITLE'].isin(top5_broker_names)]
    print(f"\nОтфильтровано {len(df_top5)} записей для топ-5 агентств")
    
    # Шаг 2: Строим таблицу сопряженности
    contingency_table = pd.crosstab(df_top5['BROKERTITLE'], df_top5['LOCALITY'])
    print("\nТаблица сопряженности (первые 5 колонок):")
    display(contingency_table.iloc[:, :5])
    
    # Шаг 3: Оцениваем применимость критерия хи-квадрат
    # Проверяем, сколько клеток имеют ожидаемые частоты менее 5
    expected = chi2_contingency(contingency_table)[3]
    low_exp_count = np.sum(expected < 5)
    low_exp_percent = low_exp_count / expected.size * 100
    print(f"\nПроверка применимости критерия хи-квадрат:")
    print(f"Количество клеток с ожидаемыми частотами < 5: {low_exp_count} из {expected.size} ({low_exp_percent:.1f}%)")
    
    # Если слишком много клеток с малыми частотами, объединяем малочисленные LOCALITY
    if low_exp_percent > 20:
        print("\nСлишком много клеток с малыми ожидаемыми частотами. Объединяем малочисленные районы.")
        
        # Определяем малочисленные районы
        locality_counts = df_top5['LOCALITY'].value_counts()
        major_localities = locality_counts[locality_counts > locality_counts.quantile(0.75)].index
        
        # Объединяем малочисленные районы в категорию "Others"
        df_top5_grouped = df_top5.copy()
        df_top5_grouped['LOCALITY_GROUPED'] = df_top5['LOCALITY'].apply(
            lambda x: x if x in major_localities else 'Others'
        )
        
        # Строим новую таблицу сопряженности
        new_contingency_table = pd.crosstab(df_top5_grouped['BROKERTITLE'], df_top5_grouped['LOCALITY_GROUPED'])
        print("\nТаблица сопряженности после группировки:")
        display(new_contingency_table)
        
        # Проверяем применимость хи-квадрат снова
        new_expected = chi2_contingency(new_contingency_table)[3]
        new_low_exp_count = np.sum(new_expected < 5)
        new_low_exp_percent = new_low_exp_count / new_expected.size * 100
        print(f"\nПосле группировки - клетки с ожидаемыми частотами < 5: {new_low_exp_count} из {new_expected.size} ({new_low_exp_percent:.1f}%)")
        
        # Используем сгруппированную таблицу для дальнейшего анализа
        contingency_table = new_contingency_table
    
    # Шаг 4: Применяем критерий хи-квадрат
    chi2, p, dof, expected = chi2_contingency(contingency_table)
    
    print("\nРезультаты критерия хи-квадрат:")
    print(f"Хи-квадрат: {chi2:.2f}")
    print(f"p-значение: {p:.6f}")
    print(f"Степени свободы: {dof}")
    
    # Интерпретируем результаты
    alpha = 0.05
    if p < alpha:
        print(f"\nТак как p-значение ({p:.6f}) < {alpha}, отвергаем нулевую гипотезу.")
        print("Существует статистически значимая связь между агентством (BROKERTITLE) и районом (LOCALITY).")
        
        # Шаг 5: Вычисляем коэффициент Крамера, если связь существует
        n = contingency_table.sum().sum()
        phi2 = chi2 / n
        r, k = contingency_table.shape
        phi2_corrected = max(0, phi2 - ((k-1)*(r-1))/(n-1))
        cramers_v = np.sqrt(phi2_corrected / min(k-1, r-1))
        
        print(f"\nКоэффициент Крамера V: {cramers_v:.4f}")
        
        # Интерпретация силы связи по коэффициенту Крамера
        if cramers_v < 0.1:
            strength = "несущественная"
        elif cramers_v < 0.2:
            strength = "слабая"
        elif cramers_v < 0.4:
            strength = "средняя"
        elif cramers_v < 0.6:
            strength = "относительно сильная"
        elif cramers_v < 0.8:
            strength = "сильная"
        else:
            strength = "очень сильная"
            
        print(f"Сила связи между признаками: {strength} (V = {cramers_v:.4f})")
        
        # Визуализация таблицы сопряженности
        plt.figure(figsize=(12, 8))
        sns.heatmap(contingency_table, annot=True, fmt='d', cmap='YlGnBu')
        plt.title('Распределение объектов по агентствам и районам')
        plt.ylabel('Агентство (BROKERTITLE)')
        plt.xlabel('Район (LOCALITY)')
        plt.tight_layout()
        plt.show()
    else:
        print(f"\nТак как p-значение ({p:.6f}) >= {alpha}, не можем отвергнуть нулевую гипотезу.")
        print("Нет статистически значимой связи между агентством (BROKERTITLE) и районом (LOCALITY).")

# Вызов функции анализа
analyze_broker_locality_relationship(df)
```

```python
from matplotlib.ticker import FuncFormatter

# Функция для форматирования цен в миллионах долларов
def millions_formatter(x, pos):
    return f'${x/1000000:.1f}M'

# Функция для создания диаграмм "ящик с усами"
def create_boxplot(df, feature, target='PRICE', top_n=10, figsize=(14, 8)):
    # Определяем топ-N категорий по частоте
    value_counts = df[feature].value_counts()
    top_categories = value_counts.head(top_n).index.tolist()
    
    # Создаем новый признак с объединением редких категорий в "Others"
    df_copy = df.copy()
    df_copy[f'{feature}_grouped'] = df_copy[feature].apply(
        lambda x: x if x in top_categories else 'Others')
    
    # Сортируем категории по медиане цены
    category_medians = df_copy.groupby(f'{feature}_grouped')[target].median().sort_values()
    category_order = category_medians.index.tolist()
    
    # Создаем диаграмму
    plt.figure(figsize=figsize)
    ax = sns.boxplot(data=df_copy, x=f'{feature}_grouped', y=target, order=category_order)
    
    # Настройка форматирования осей
    ax.yaxis.set_major_formatter(FuncFormatter(millions_formatter))
    plt.xticks(rotation=45, ha='right')
    plt.title(f'Распределение цен в зависимости от {feature}', fontsize=16)
    plt.ylabel('Цена', fontsize=14)
    plt.xlabel(feature, fontsize=14)
    plt.tight_layout()
    
    # Печатаем статистику для категории "Others"
    if 'Others' in df_copy[f'{feature}_grouped'].unique():
        others_stats = df_copy[df_copy[f'{feature}_grouped'] == 'Others'][target].describe()
        print(f"\nСтатистика для категории 'Others' в признаке {feature}:")
        print(others_stats)
        
        # Процент наблюдений в категории "Others"
        others_pct = df_copy[f'{feature}_grouped'].value_counts(normalize=True)['Others'] * 100
        print(f"Процент наблюдений в категории 'Others': {others_pct:.2f}%")
    
    # Сравниваем медианные цены по категориям
    medians = df_copy.groupby(f'{feature}_grouped')[target].median()
    print(f"\nМедианные цены по категориям {feature}:")
    for category, median in medians.items():
        print(f"{category}: ${median:,.0f}")
    
    return ax, df_copy[f'{feature}_grouped']

# Анализ качественных признаков
categorical_features = ['BROKERTITLE', 'TYPE', 'LOCALITY', 'SUBLOCALITY']

# Создание диаграмм "ящик с усами" для каждого качественного признака
grouped_features = {}
for feature in categorical_features:
    print(f"\n{'='*50}")
    print(f"Анализ признака: {feature}")
    print(f"{'='*50}")
    ax, grouped_feature = create_boxplot(df, feature)
    grouped_features[feature] = grouped_feature
    plt.show()

# Дополнительно: анализ возможности объединения близких категорий
# Для примера объединим близкие категории по LOCALITY
def merge_similar_categories(df, feature, grouped_feature, target='PRICE', 
                            similarity_threshold=0.1, min_freq_pct=0.05):
    """
    Объединяет категории с близкими медианными ценами
    
    Args:
        df: Исходный датафрейм
        feature: Исходный признак
        grouped_feature: Сгруппированный признак (с категорией Others)
        target: Целевой признак (цена)
        similarity_threshold: Порог сходства медианных цен (в долях)
        min_freq_pct: Минимальная частота категории для сохранения
        
    Returns:
        Series с новым объединенным признаком
    """
    # Получаем медианы цен по категориям
    category_medians = df.groupby(grouped_feature)[target].median().sort_values()
    
    # Создаем новый признак с объединенными категориями
    result = grouped_feature.copy()
    
    # Частоты категорий
    frequency = grouped_feature.value_counts(normalize=True)
    
    # Создаем словарь для объединения
    merge_dict = {}
    
    # Предыдущая категория и её медиана
    prev_category = None
    prev_median = None
    new_group = None
    
    # Для каждой категории, отсортированной по медиане
    for category, median in category_medians.items():
        # Пропускаем Others
        if category == 'Others':
            merge_dict[category] = category
            continue
            
        # Если частота категории достаточно высока, сохраняем её отдельно
        if frequency[category] >= min_freq_pct:
            # Если это первая категория или медиана существенно отличается от предыдущей
            if prev_median is None or abs(median - prev_median) / prev_median > similarity_threshold:
                new_group = category
            # Если медиана близка к предыдущей, объединяем категории
            else:
                new_group = f"{prev_category}+{category}"
                merge_dict[prev_category] = new_group
            
            merge_dict[category] = new_group
            prev_category = category
            prev_median = median
        else:
            # Малочисленные категории объединяем в "Small Categories"
            merge_dict[category] = "Small Categories"
    
    # Применяем словарь для преобразования
    for old, new in merge_dict.items():
        result[result == old] = new
        
    return result

# Пример объединения близких категорий для LOCALITY
print("\nОбъединение близких категорий для LOCALITY:")
locality_merged = merge_similar_categories(df, 'LOCALITY', grouped_features['LOCALITY'])

# Визуализация результата объединения
plt.figure(figsize=(14, 8))
ax = sns.boxplot(data=df, x=locality_merged, y='PRICE')
ax.yaxis.set_major_formatter(FuncFormatter(millions_formatter))
plt.xticks(rotation=45, ha='right')
plt.title(f'Распределение цен в зависимости от объединенных категорий LOCALITY', fontsize=16)
plt.ylabel('Цена', fontsize=14)
plt.xlabel('LOCALITY (объединенные категории)', fontsize=14)
plt.tight_layout()
plt.show()
```
