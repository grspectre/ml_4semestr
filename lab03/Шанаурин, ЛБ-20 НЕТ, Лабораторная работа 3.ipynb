### Пояснение:

- операция «удалить» означает, что все соответствующие строки должны быть удалены их фрейма, а не соответствующие остаться;
- операция «сохранить» означает, что во фрейме должны быть удалены строки не соответствующие условию, а остальные – остаться;
- общее (сложное) условие формируется из двух простых, при этом варианты взаимного выполнения обоих условий следующие:
- «одновременно»: оба простых условия должны быть истинными (True);
- «только первое»: первое простое условие должны быть True, а второе - False;
- «только второе»: первое простое условие должны быть False, а второе - True;
- «хотя бы одно»: или первое или второе простое условие должны быть True;
- «ни одно»: оба простых условия должны быть False.

## Задача 4.
Используя исходный фрейм с пассажирами «Титаника» из второй задачи вычислить величину, указанную в вариантах для каждого студента.
### Условия
Вид расчёта: Сумма  
Для каких данных: все числовые  
При каком условии: мужчин

````python
import pandas as pd
from IPython.display import display, HTML

display(HTML('<h2>Задача 2</h2>'))

pd.set_option('display.max_colwidth', None)
pd.set_option('display.float_format', '{:.2f}'.format)

titanic_df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')

# Выведем первые строки исходного фрейма
print("Исходный фрейм данных:")
display(titanic_df.head(10))

# Применяем условия
# Первое условие: пассажир младше 15 лет
condition1 = titanic_df['Age'] < 15
# Второе условие: пассажир мужского пола
condition2 = titanic_df['Sex'] == 'male'

# "Только второе" означает: condition1 == False и condition2 == True
filtered_df = titanic_df[~condition1 & condition2]

# Выведем первые строки обработанного фрейма
print("\nОбработанный фрейм (сохранены только пассажиры мужского пола 15 лет и старше):")
display(filtered_df.head(10))

# Проверка: выводим количество строк до и после обработки
print(f"\nКоличество строк в исходном фрейме: {titanic_df.shape[0]}")
print(f"Количество строк в обработанном фрейме: {filtered_df.shape[0]}")

# Исходный фрейм
print("\nСрез исходного фрейма (Sex, Age):")
display(titanic_df[['Sex', 'Age']].head(15))

# Обработанный фрейм
print("\nСрез обработанного фрейма (Sex, Age):")
display(filtered_df[['Sex', 'Age']].head(15))


display(HTML('<h2>Задача 3</h2>'))

# Выведем первые 4 и последние 4 строки фрейма до сортировки
print("До сортировки:")
print("Первые 4 строки:")
display(filtered_df.head(4))
print("\nПоследние 4 строки:")
display(filtered_df.tail(4))

# Сортируем фрейм по возрасту (по убыванию)
sorted_df = filtered_df.sort_values(by='Age', ascending=False)

# Выведем первые 4 и последние 4 строки фрейма после сортировки
print("\nПосле сортировки:")
print("Первые 4 строки:")
display(sorted_df.head(4))
print("\nПоследние 4 строки:")
display(sorted_df.tail(4))

display(HTML('''<p style="font-size: 20px;">
После сортировки фрейма по колонке 'Age' в порядке убывания, в начале фрейма находятся самые старшие пассажиры-мужчины, а в конце - самые младшие (но не младше 15 лет, так как мы их отфильтровали в задаче 2). Если в данных есть пропущенные значения возраста (NaN), они будут размещены в конце отсортированного фрейма.
</p>'''))


display(HTML('<h2>Задача 4</h2>'))

# Фильтруем исходный фрейм, оставляя только мужчин
men_df = titanic_df[titanic_df['Sex'] == 'male']

# Выбираем только числовые колонки
numeric_columns = men_df.select_dtypes(include=['number']).columns
numeric_data = men_df[numeric_columns]

# Вычисляем сумму всех числовых данных
sums = numeric_data.sum()

print("Сумма числовых данных для мужчин:")
summary = 0
for column, sum_value in sums.items():
    summary += sum_value
    print(f"{column}: {sum_value:.2f}")

print(f"\nОбщая сумма всех числовых значений для мужчин: {summary}")
````

## Задача 5*
Создайте два DataFrame и объедините их, указав метод объединения (outer) в методе join().

````python
import pandas as pd
import numpy as np
from IPython.display import display

# Создаем первый DataFrame
df1 = pd.DataFrame({
    'А': ['А0', 'А1', 'А2', 'А3'],
    'Б': ['Б0', 'Б1', 'Б2', 'Б3'],
    'В': ['В0', 'В1', 'В2', 'В3']
}, index=['K0', 'K1', 'K2', 'K3'])

# Создаем второй DataFrame
df2 = pd.DataFrame({
    'Г': ['Г0', 'Г1', 'Г2', 'Г3'],
    'Д': ['Д0', 'Д1', 'Д2', 'Д3'],
    'Е': ['Е0', 'Е1', 'Е2', 'Е3']
}, index=['K0', 'K2', 'K4', 'K5'])

# Выводим первый DataFrame
print("Первый DataFrame (df1):")
display(df1)
print("\n")

# Выводим второй DataFrame
print("Второй DataFrame (df2):")
display(df2)
print("\n")

# Объединяем DataFrame с помощью метода join() с параметром how='outer'
result = df1.join(df2, how='outer')

# Выводим результат
print("Результат объединения df1.join(df2, how='outer'):")
display(result)
````
