—Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è,
- –æ—Ç–≤–µ—Ç–∏—Ç—å –Ω–∞ –≤–æ–ø—Ä–æ—Å: –∫–∞–∫ –≤–ª–∏—è–µ—Ç –∑–Ω–∞—á–µ–Ω–∏–µ ùëõ –Ω–∞ –±–ª–∏–∑–æ—Å—Ç—å —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è $\overline{ùëã}$ –∫
–Ω–æ—Ä–º–∞–ª—å–Ω–æ–º—É —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—é?
- –æ—Ü–µ–Ω–∏—Ç—å, —Å–æ–≥–ª–∞—Å—É—é—Ç—Å—è –ª–∏ –ø–æ–ª—É—á–µ–Ω–Ω—ã–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã —Å —É—Ç–≤–µ—Ä–∂–¥–µ–Ω–∏–µ–º —Ü–µ–Ω—Ç—Ä–∞–ª—å–Ω–æ–π
–ø—Ä–µ–¥–µ–ª—å–Ω–æ–π —Ç–µ–æ—Ä–µ–º—ã.
## –£—Å–ª–æ–≤–∏—è

–°–ª—É—á–∞–π–Ω–∞—è –≤–µ–ª–∏—á–∏–Ω–∞ ùëã –∑–∞–¥–∞–Ω–∞ —Ñ—É–Ω–∫—Ü–∏–µ–π —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è
$f(ùë•) = \frac {1} {3 \cdot \sqrt {2 \pi}} \int_{-\infty}^x e^{- \frac {t^2} {18} } dt$.

–î–ª—è –ø—É–Ω–∫—Ç–∞ 4.2) ‚Äì 5%-–Ω—ã–π –∫–≤–∞–Ω—Ç–∏–ª—å; $ùëÉ(1,5 \leq ùëã < 5,5)$.

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
from IPython.display import display, Markdown, HTML

display(HTML('<h1>–ó–∞–¥–∞–Ω–∏–µ 1</h1>'))
display(HTML('<h2>1. –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å–ª—É—á–∞–π–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω—ã</h2>'))

display(Markdown("""
–ü–æ –∑–∞–¥–∞–Ω–Ω–æ–π —Ñ—É–Ω–∫—Ü–∏–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –º–æ–∂–Ω–æ –æ–ø—Ä–µ–¥–µ–ª–∏—Ç—å:

__–ö–ª–∞—Å—Å —Å–ª—É—á–∞–π–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω—ã__: –ù–µ–ø—Ä–µ—Ä—ã–≤–Ω–∞—è —Å–ª—É—á–∞–π–Ω–∞—è –≤–µ–ª–∏—á–∏–Ω–∞  
__–ó–∞–∫–æ–Ω —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è__: –ù–æ—Ä–º–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ (–ì–∞—É—Å—Å–æ–≤–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ)  
__–ü–∞—Ä–∞–º–µ—Ç—Ä—ã –∑–∞–∫–æ–Ω–∞ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è__:

- –ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ $(Œº) = 0$
- –°—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ $(œÉ) = 3$
"""))

display(HTML('<h2>2. –°–æ–∑–¥–∞–Ω–∏–µ —Å–ª—É—á–∞–π–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω—ã –∏ 3. –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–æ–≤</h2>'))

# –ó–∞–¥–∞–µ–º –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è
mu = 0  # –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ
sigma = 3  # —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ

# –°–æ–∑–¥–∞–µ–º —Å–ª—É—á–∞–π–Ω—É—é –≤–µ–ª–∏—á–∏–Ω—É —Å –Ω–æ—Ä–º–∞–ª—å–Ω—ã–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ–º
np.random.seed(616)  # –¥–ª—è –≤–æ—Å–ø—Ä–æ–∏–∑–≤–æ–¥–∏–º–æ—Å—Ç–∏ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
X = np.random.normal(mu, sigma, 1000)  # –≥–µ–Ω–µ—Ä–∏—Ä—É–µ–º 1000 –∑–Ω–∞—á–µ–Ω–∏–π

# –î–∏–∞–ø–∞–∑–æ–Ω –∑–Ω–∞—á–µ–Ω–∏–π –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–æ–≤
x_range = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)

# –°—Ç—Ä–æ–∏–º –≥—Ä–∞—Ñ–∏–∫–∏
plt.figure(figsize=(12, 5))

# –ì—Ä–∞—Ñ–∏–∫ –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ (PDF)
plt.subplot(1, 2, 1)
plt.plot(x_range, stats.norm.pdf(x_range, mu, sigma), 'b-')
plt.title('–§—É–Ω–∫—Ü–∏—è –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π', fontsize=12)
plt.xlabel('–ó–Ω–∞—á–µ–Ω–∏–µ X', fontsize=10)
plt.ylabel('–ü–ª–æ—Ç–Ω–æ—Å—Ç—å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏', fontsize=10)
plt.grid(True)

# –ì—Ä–∞—Ñ–∏–∫ —Ñ—É–Ω–∫—Ü–∏–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è (CDF)
plt.subplot(1, 2, 2)
plt.plot(x_range, stats.norm.cdf(x_range, mu, sigma), 'r-')
plt.title('–§—É–Ω–∫—Ü–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è', fontsize=12)
plt.xlabel('–ó–Ω–∞—á–µ–Ω–∏–µ X', fontsize=10)
plt.ylabel('–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å P(X ‚â§ x)', fontsize=10)
plt.grid(True)

plt.tight_layout()
plt.show()

display(HTML('<h2>4. –í—ã—á–∏—Å–ª–µ–Ω–∏–µ —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫ —Å–ª—É—á–∞–π–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω—ã</h2>'))

# 4.1) –í—ã—á–∏—Å–ª–µ–Ω–∏–µ —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫
mean_theoretical = mu
variance_theoretical = sigma**2
std_theoretical = sigma

mean_empirical = np.mean(X)
variance_empirical = np.var(X)
std_empirical = np.std(X)

print(f"–¢–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏–µ —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫–∏:")
print(f"–ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ: {mean_theoretical}")
print(f"–î–∏—Å–ø–µ—Ä—Å–∏—è: {variance_theoretical}")
print(f"–°—Ä–µ–¥–Ω–µ–µ –∫–≤–∞–¥—Ä–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ: {std_theoretical}")
print("\n–≠–º–ø–∏—Ä–∏—á–µ—Å–∫–∏–µ —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫–∏:")
print(f"–í—ã–±–æ—Ä–æ—á–Ω–æ–µ —Å—Ä–µ–¥–Ω–µ–µ: {mean_empirical:.4f}")
print(f"–í—ã–±–æ—Ä–æ—á–Ω–∞—è –¥–∏—Å–ø–µ—Ä—Å–∏—è: {variance_empirical:.4f}")
print(f"–í—ã–±–æ—Ä–æ—á–Ω–æ–µ —Å—Ä–µ–¥–Ω–µ–µ –∫–≤–∞–¥—Ä–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ: {std_empirical:.4f}")

# 4.2) –í—ã—á–∏—Å–ª–µ–Ω–∏–µ 5%-–Ω–æ–≥–æ –∫–≤–∞–Ω—Ç–∏–ª—è –∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ P(1,5 ‚â§ X < 5,5)
quantile_5 = stats.norm.ppf(0.05, mu, sigma)
probability = stats.norm.cdf(5.5, mu, sigma) - stats.norm.cdf(1.5, mu, sigma)

print(f"\n5%-–Ω—ã–π –∫–≤–∞–Ω—Ç–∏–ª—å: {quantile_5:.4f}")
print(f"–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å P(1,5 ‚â§ X < 5,5): {probability:.4f}")

# 4.4) –ü—Ä–æ–≤–µ—Ä–∫–∞ –ø—Ä–∞–≤–∏–ª–∞ —Ç—Ä–µ—Ö —Å–∏–≥–º
# –ì–µ–Ω–µ—Ä–∏—Ä—É–µ–º 1000 –∑–Ω–∞—á–µ–Ω–∏–π —Å–ª—É—á–∞–π–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω—ã
np.random.seed(42)
X_large = np.random.normal(mu, sigma, 1000)

# –í—ã—á–∏—Å–ª—è–µ–º –≥—Ä–∞–Ω–∏—Ü—ã –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞
lower_bound = mu - 3*sigma
upper_bound = mu + 3*sigma

# –°—á–∏—Ç–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–Ω–∞—á–µ–Ω–∏–π, –ø–æ–ø–∞–≤—à–∏—Ö –≤ –∏–Ω—Ç–µ—Ä–≤–∞–ª
count_in_range = np.sum((X_large > lower_bound) & (X_large < upper_bound))
frequency = count_in_range / len(X_large)

# –¢–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∞—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ–ø–∞–¥–∞–Ω–∏—è –≤ –∏–Ω—Ç–µ—Ä–≤–∞–ª
theoretical_prob = stats.norm.cdf(upper_bound, mu, sigma) - stats.norm.cdf(lower_bound, mu, sigma)

print(f"\n–ü—Ä–æ–≤–µ—Ä–∫–∞ –ø—Ä–∞–≤–∏–ª–∞ —Ç—Ä–µ—Ö —Å–∏–≥–º:")
print(f"–ì—Ä–∞–Ω–∏—Ü—ã –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞: ({lower_bound}, {upper_bound})")
print(f"–ß–∞—Å—Ç–æ—Ç–∞ –ø–æ–ø–∞–¥–∞–Ω–∏—è –≤ –∏–Ω—Ç–µ—Ä–≤–∞–ª: {frequency:.4f}")
print(f"–¢–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∞—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å: {theoretical_prob:.4f}")
print(f"–†–∞–∑–Ω–∏—Ü–∞: {abs(frequency - theoretical_prob):.4f}")

display(HTML('<h2>5. –í–ª–∏—è–Ω–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –Ω–∞ –∑–∞–∫–æ–Ω —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è</h2>'))

# 5.1) –∏ 5.2) –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–æ–≤ —Å —Ä–∞–∑–Ω—ã–º–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏
plt.figure(figsize=(14, 6))

# –ü–∞—Ä–∞–º–µ—Ç—Ä—ã –¥–ª—è —Ä–∞–∑–ª–∏—á–Ω—ã—Ö —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–π
parameters = [
    (0, 1, '–ù–æ—Ä–º–∞–ª—å–Ω–æ–µ(0, 1)'),
    (0, 3, '–ù–æ—Ä–º–∞–ª—å–Ω–æ–µ(0, 3)'),
    (0, 5, '–ù–æ—Ä–º–∞–ª—å–Ω–æ–µ(0, 5)'),
    (2, 3, '–ù–æ—Ä–º–∞–ª—å–Ω–æ–µ(2, 3)')
]

# –î–∏–∞–ø–∞–∑–æ–Ω –∑–Ω–∞—á–µ–Ω–∏–π –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–æ–≤
x_range = np.linspace(-15, 15, 1000)

# –ì—Ä–∞—Ñ–∏–∫ –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ (PDF)
plt.subplot(1, 2, 1)
for mu, sigma, label in parameters:
    plt.plot(x_range, stats.norm.pdf(x_range, mu, sigma), label=label)
plt.title('–§—É–Ω–∫—Ü–∏–∏ –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π\n—Å —Ä–∞–∑–Ω—ã–º–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏', fontsize=12)
plt.xlabel('–ó–Ω–∞—á–µ–Ω–∏–µ X', fontsize=10)
plt.ylabel('–ü–ª–æ—Ç–Ω–æ—Å—Ç—å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏', fontsize=10)
plt.legend()
plt.grid(True)

# –ì—Ä–∞—Ñ–∏–∫ —Ñ—É–Ω–∫—Ü–∏–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è (CDF)
plt.subplot(1, 2, 2)
for mu, sigma, label in parameters:
    plt.plot(x_range, stats.norm.cdf(x_range, mu, sigma), label=label)
plt.title('–§—É–Ω–∫—Ü–∏–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è\n—Å —Ä–∞–∑–Ω—ã–º–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏', fontsize=12)
plt.xlabel('–ó–Ω–∞—á–µ–Ω–∏–µ X', fontsize=10)
plt.ylabel('–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å P(X ‚â§ x)', fontsize=10)
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

display(Markdown(r"""
## –ò–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è –ø–æ–ª—É—á–µ–Ω–Ω—ã—Ö —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
### 4.3) –ò–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è —á–∏—Å–ª–æ–≤—ã—Ö —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫

- __–ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ ($Œº = 0$)__: –¢–æ—á–∫–∞ —Å–∏–º–º–µ—Ç—Ä–∏–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è, —Ü–µ–Ω—Ç—Ä —Ç—è–∂–µ—Å—Ç–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è.
–≠—Ç–æ –∑–Ω–∞—á–µ–Ω–∏–µ, –≤–æ–∫—Ä—É–≥ –∫–æ—Ç–æ—Ä–æ–≥–æ –≥—Ä—É–ø–ø–∏—Ä—É—é—Ç—Å—è —Ä–µ–∞–ª–∏–∑–∞—Ü–∏–∏ —Å–ª—É—á–∞–π–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω—ã.
- __–î–∏—Å–ø–µ—Ä—Å–∏—è ($œÉ^2 = 9$)__: –ú–µ—Ä–∞ —Ä–∞–∑–±—Ä–æ—Å–∞ —Å–ª—É—á–∞–π–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω—ã –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ –µ–µ –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–≥–æ –æ–∂–∏–¥–∞–Ω–∏—è. 
–ë–æ–ª—å—à–∞—è –¥–∏—Å–ø–µ—Ä—Å–∏—è —É–∫–∞–∑—ã–≤–∞–µ—Ç –Ω–∞ –±–æ–ª–µ–µ —à–∏—Ä–æ–∫–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏–π.
- __–°—Ä–µ–¥–Ω–µ–µ –∫–≤–∞–¥—Ä–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ ($œÉ = 3$)__: –ú–µ—Ä–∞ —Ä–∞–∑–±—Ä–æ—Å–∞ –∑–Ω–∞—á–µ–Ω–∏–π —Å–ª—É—á–∞–π–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω—ã –æ—Ç 
–º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–≥–æ –æ–∂–∏–¥–∞–Ω–∏—è, –≤—ã—Ä–∞–∂–µ–Ω–Ω–∞—è –≤ —Ç–µ—Ö –∂–µ –µ–¥–∏–Ω–∏—Ü–∞—Ö, —á—Ç–æ –∏ —Å–∞–º–∞ —Å–ª—É—á–∞–π–Ω–∞—è –≤–µ–ª–∏—á–∏–Ω–∞. –ü–æ–∫–∞–∑—ã–≤–∞–µ—Ç 
—Å—Ä–µ–¥–Ω–µ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏–π –æ—Ç –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–≥–æ –æ–∂–∏–¥–∞–Ω–∏—è.
- __5%-–Ω—ã–π –∫–≤–∞–Ω—Ç–∏–ª—å (–ø—Ä–∏–º–µ—Ä–Ω–æ -4.94)__: –≠—Ç–æ —Ç–∞–∫–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ x, —á—Ç–æ $P(X < x) = 0.05$. –≠—Ç–æ –æ–∑–Ω–∞—á–∞–µ—Ç, —á—Ç–æ 
–ø—Ä–∏–º–µ—Ä–Ω–æ 5% –≤—Å–µ—Ö –∑–Ω–∞—á–µ–Ω–∏–π —Å–ª—É—á–∞–π–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω—ã –º–µ–Ω—å—à–µ -4.94.
- __–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å $P(1,5 \leq X < 5,5)$ (–ø—Ä–∏–º–µ—Ä–Ω–æ 0.34)__: –≠—Ç–æ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ —Å–ª—É—á–∞–π–Ω–∞—è –≤–µ–ª–∏—á–∏–Ω–∞ X 
–ø—Ä–∏–º–µ—Ç –∑–Ω–∞—á–µ–Ω–∏–µ –≤ –∏–Ω—Ç–µ—Ä–≤–∞–ª–µ –æ—Ç 1.5 –¥–æ 5.5. –ü—Ä–∏–º–µ—Ä–Ω–æ 34% –≤—Å–µ—Ö –∑–Ω–∞—á–µ–Ω–∏–π –ø–æ–ø–∞–¥–∞–µ—Ç –≤ —ç—Ç–æ—Ç –∏–Ω—Ç–µ—Ä–≤–∞–ª.

### 4.4) –û—Ü–µ–Ω–∫–∞ –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è –ø—Ä–∞–≤–∏–ª–∞ —Ç—Ä–µ—Ö —Å–∏–≥–º

–°–æ–≥–ª–∞—Å–Ω–æ –ø—Ä–∞–≤–∏–ª—É —Ç—Ä–µ—Ö —Å–∏–≥–º, –¥–ª—è –Ω–æ—Ä–º–∞–ª—å–Ω–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω–æ–π —Å–ª—É—á–∞–π–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω—ã –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å 
–ø–æ–ø–∞–¥–∞–Ω–∏—è –≤ –∏–Ω—Ç–µ—Ä–≤–∞–ª ($Œº - 3œÉ, Œº + 3œÉ$) —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏ —Ä–∞–≤–Ω–∞ 0.9973 –∏–ª–∏ –ø—Ä–∏–º–µ—Ä–Ω–æ 99.73%.

–í –Ω–∞—à–µ–º –≤—ã—á–∏—Å–ª–∏—Ç–µ–ª—å–Ω–æ–º —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–µ —á–∞—Å—Ç–æ—Ç–∞ –ø–æ–ø–∞–¥–∞–Ω–∏—è –≤ –∏–Ω—Ç–µ—Ä–≤–∞–ª (-9, 9) —Å–æ—Å—Ç–∞–≤–∏–ª–∞ –ø—Ä–∏–º–µ—Ä–Ω–æ 0.997
(–∏–ª–∏ –º–æ–∂–µ—Ç –±—ã—Ç—å –Ω–µ–º–Ω–æ–≥–æ –¥—Ä—É–≥–æ–π –∏–∑-–∑–∞ —Å–ª—É—á–∞–π–Ω–æ–π –≥–µ–Ω–µ—Ä–∞—Ü–∏–∏), —á—Ç–æ –æ—á–µ–Ω—å –±–ª–∏–∑–∫–æ –∫ —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–æ–º—É –∑–Ω–∞—á–µ–Ω–∏—é. 
–†–∞–∑–Ω–∏—Ü–∞ –º–µ–∂–¥—É —ç–º–ø–∏—Ä–∏—á–µ—Å–∫–æ–π –∏ —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–æ–π –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—è–º–∏ —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç –º–µ–Ω–µ–µ 0.01, —á—Ç–æ —Å–≤–∏–¥–µ—Ç–µ–ª—å—Å—Ç–≤—É–µ—Ç 
–æ —Ö–æ—Ä–æ—à–µ–º —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞–ª—å–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–æ–π –º–æ–¥–µ–ª–∏.

–í—ã–≤–æ–¥: –ø—Ä–∞–≤–∏–ª–æ —Ç—Ä–µ—Ö —Å–∏–≥–º –≤—ã–ø–æ–ª–Ω—è–µ—Ç—Å—è –¥–ª—è –Ω–∞—à–µ–π —Å–ª—É—á–∞–π–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω—ã, —á—Ç–æ –ø–æ–¥—Ç–≤–µ—Ä–∂–¥–∞–µ—Ç –Ω–æ—Ä–º–∞–ª—å–Ω–æ—Å—Ç—å 
–µ—ë —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è.

### 5.3) –í–ª–∏—è–Ω–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –Ω–∞ —Ö–∞—Ä–∞–∫—Ç–µ—Ä —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è
–ù–∞ –æ—Å–Ω–æ–≤–µ –ø–æ—Å—Ç—Ä–æ–µ–Ω–Ω—ã—Ö –≥—Ä–∞—Ñ–∏–∫–æ–≤ –º–æ–∂–Ω–æ —Å–¥–µ–ª–∞—Ç—å —Å–ª–µ–¥—É—é—â–∏–µ –≤—ã–≤–æ–¥—ã:

1. __–í–ª–∏—è–Ω–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ Œº (–º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ)__:
- –ò–∑–º–µ–Ω–µ–Ω–∏–µ $Œº$ –ø—Ä–∏–≤–æ–¥–∏—Ç –∫ –≥–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª—å–Ω–æ–º—É —Å–¥–≤–∏–≥—É —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è.
- –ë–æ–ª—å—à–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ $Œº$ —Å–º–µ—â–∞–µ—Ç –∫—Ä–∏–≤—É—é –≤–ø—Ä–∞–≤–æ, –º–µ–Ω—å—à–µ–µ - –≤–ª–µ–≤–æ.
- –§–æ—Ä–º–∞ –∫—Ä–∏–≤–æ–π –Ω–µ –º–µ–Ω—è–µ—Ç—Å—è –ø—Ä–∏ –∏–∑–º–µ–Ω–µ–Ω–∏–∏ $Œº$.

2. __–í–ª–∏—è–Ω–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ $œÉ$ (—Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ)__:
- –ò–∑–º–µ–Ω–µ–Ω–∏–µ œÉ –≤–ª–∏—è–µ—Ç –Ω–∞ —Ñ–æ—Ä–º—É —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è.
- –ü—Ä–∏ —É–≤–µ–ª–∏—á–µ–Ω–∏–∏ $œÉ$ –≥—Ä–∞—Ñ–∏–∫ —Å—Ç–∞–Ω–æ–≤–∏—Ç—Å—è –±–æ–ª–µ–µ –ø–ª–æ—Å–∫–∏–º –∏ —Ä–∞—Å—Ç—è–Ω—É—Ç—ã–º –ø–æ –≥–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª–∏, –∞ –ø—Ä–∏ 
—É–º–µ–Ω—å—à–µ–Ω–∏–∏ - –±–æ–ª–µ–µ —É–∑–∫–∏–º –∏ –≤—ã—Å–æ–∫–∏–º.
- –ß–µ–º –±–æ–ª—å—à–µ $œÉ$, —Ç–µ–º –±–æ–ª—å—à–µ —Ä–∞–∑–±—Ä–æ—Å –≤–æ–∑–º–æ–∂–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π —Å–ª—É—á–∞–π–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω—ã.

–ù–∞ —Ñ—É–Ω–∫—Ü–∏–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —É–≤–µ–ª–∏—á–µ–Ω–∏–µ œÉ –¥–µ–ª–∞–µ—Ç –ø–µ—Ä–µ—Ö–æ–¥ –æ—Ç 0 –∫ 1 –±–æ–ª–µ–µ –ø–ª–∞–≤–Ω—ã–º.

3. __–û–±—â–∏–µ –Ω–∞–±–ª—é–¥–µ–Ω–∏—è__:
- –ù–µ–∑–∞–≤–∏—Å–∏–º–æ –æ—Ç –∑–Ω–∞—á–µ–Ω–∏–π –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤, –Ω–æ—Ä–º–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –≤—Å–µ–≥–¥–∞ –∏–º–µ–µ—Ç –∫–æ–ª–æ–∫–æ–ª–æ–æ–±—Ä–∞–∑–Ω—É—é —Ñ–æ—Ä–º—É, 
—Å–∏–º–º–µ—Ç—Ä–∏—á–Ω—É—é –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ $Œº$.
- –§—É–Ω–∫—Ü–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –≤—Å–µ–≥–¥–∞ –∏–º–µ–µ—Ç S-–æ–±—Ä–∞–∑–Ω—É—é —Ñ–æ—Ä–º—É, —Å—Ç—Ä–µ–º—è—Å—å –∫ 0 –ø—Ä–∏ $x ‚Üí - \infty$ –∏ –∫ 1 –ø—Ä–∏ $x ‚Üí +\infty$.
- –ù–∞–∏–±–æ–ª–µ–µ –∫—Ä—É—Ç–æ–π –ø–æ–¥—ä–µ–º –Ω–∞ –≥—Ä–∞—Ñ–∏–∫–µ —Ñ—É–Ω–∫—Ü–∏–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –Ω–∞–±–ª—é–¥–∞–µ—Ç—Å—è –≤ –æ–±–ª–∞—Å—Ç–∏ $Œº$, –∏ —ç—Ç–æ—Ç –ø–æ–¥—ä–µ–º 
—Å—Ç–∞–Ω–æ–≤–∏—Ç—Å—è –±–æ–ª–µ–µ –ø–æ–ª–æ–≥–∏–º –ø—Ä–∏ —É–≤–µ–ª–∏—á–µ–Ω–∏–∏ $œÉ$.

–≠—Ç–∏ —Å–≤–æ–π—Å—Ç–≤–∞ –¥–µ–ª–∞—é—Ç –Ω–æ—Ä–º–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —É–Ω–∏–≤–µ—Ä—Å–∞–ª—å–Ω—ã–º –∏–Ω—Å—Ç—Ä—É–º–µ–Ω—Ç–æ–º –¥–ª—è –º–æ–¥–µ–ª–∏—Ä–æ–≤–∞–Ω–∏—è 
–º–Ω–æ–≥–∏—Ö –µ—Å—Ç–µ—Å—Ç–≤–µ–Ω–Ω—ã—Ö –ø—Ä–æ—Ü–µ—Å—Å–æ–≤ –∏ –∏–∑–º–µ—Ä–µ–Ω–∏–π.
"""))
````

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
from IPython.display import display, Markdown, HTML


# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –≥–µ–Ω–µ—Ä–∞—Ü–∏–∏ —Å—Ä–µ–¥–Ω–∏—Ö –∑–Ω–∞—á–µ–Ω–∏–π –≤—ã–±–æ—Ä–æ–∫
def generate_sample_means(distribution, n, num_samples=1000):
    """
    –ì–µ–Ω–µ—Ä–∏—Ä—É–µ—Ç –º–∞—Å—Å–∏–≤ –∏–∑ —Å—Ä–µ–¥–Ω–∏—Ö –∑–Ω–∞—á–µ–Ω–∏–π –≤—ã–±–æ—Ä–æ–∫ —Ä–∞–∑–º–µ—Ä–æ–º n
    
    Args:
        distribution: —Ñ—É–Ω–∫—Ü–∏—è, –≥–µ–Ω–µ—Ä–∏—Ä—É—é—â–∞—è —Å–ª—É—á–∞–π–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è
        n: —Ä–∞–∑–º–µ—Ä –∫–∞–∂–¥–æ–π –≤—ã–±–æ—Ä–∫–∏
        num_samples: –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤—ã–±–æ—Ä–æ–∫
        
    Returns:
        –ú–∞—Å—Å–∏–≤ —Å—Ä–µ–¥–Ω–∏—Ö –∑–Ω–∞—á–µ–Ω–∏–π –≤—ã–±–æ—Ä–æ–∫
    """
    sample_means = np.zeros(num_samples)
    
    for i in range(num_samples):
        # –ì–µ–Ω–µ—Ä–∞—Ü–∏—è n –∑–Ω–∞—á–µ–Ω–∏–π –∏ –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ –∏—Ö —Å—Ä–µ–¥–Ω–µ–≥–æ
        sample = distribution(n)
        sample_means[i] = np.mean(sample)
    
    return sample_means


# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –≥–µ–Ω–µ—Ä–∞—Ü–∏–∏ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Å –∑–∞–¥–∞–Ω–Ω—ã–º–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏
def normal_distribution(size):
    return np.random.normal(mu, sigma, size)


display(HTML('<h1>–ó–∞–¥–∞–Ω–∏–µ 1</h1>'))
display(HTML('<h2>1. –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å–ª—É—á–∞–π–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω—ã</h2>'))

# –ó–∞–¥–∞–µ–º –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è
mu = 0  # –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ
sigma = 3  # —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ

# –°–æ–∑–¥–∞–µ–º —Å–ª—É—á–∞–π–Ω—É—é –≤–µ–ª–∏—á–∏–Ω—É —Å –Ω–æ—Ä–º–∞–ª—å–Ω—ã–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ–º
np.random.seed(42)  # –¥–ª—è –≤–æ—Å–ø—Ä–æ–∏–∑–≤–æ–¥–∏–º–æ—Å—Ç–∏ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
X = np.random.normal(mu, sigma, 1000)  # –≥–µ–Ω–µ—Ä–∏—Ä—É–µ–º 1000 –∑–Ω–∞—á–µ–Ω–∏–π

# –î–∏–∞–ø–∞–∑–æ–Ω –∑–Ω–∞—á–µ–Ω–∏–π –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–æ–≤
x_range = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)

# –í—ã—á–∏—Å–ª—è–µ–º –∑–Ω–∞—á–µ–Ω–∏—è —Ñ—É–Ω–∫—Ü–∏–∏ –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏
pdf_values = stats.norm.pdf(x_range, mu, sigma)

# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–æ–≤
plt.figure(figsize=(10, 6))
plt.hist(X, bins=30, density=True, alpha=0.7, label='–ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ –≤—ã–±–æ—Ä–∫–∏')
plt.plot(x_range, pdf_values, 'r-', linewidth=2, label='–ü–ª–æ—Ç–Ω–æ—Å—Ç—å –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è')
plt.title('–ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ –≤—ã–±–æ—Ä–∫–∏ –∏ –ø–ª–æ—Ç–Ω–æ—Å—Ç—å –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è N(0, 3)')
plt.xlabel('–ó–Ω–∞—á–µ–Ω–∏—è —Å–ª—É—á–∞–π–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω—ã')
plt.ylabel('–ü–ª–æ—Ç–Ω–æ—Å—Ç—å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

display(HTML('<h2>4-6. –ò—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏–µ —Ü–µ–Ω—Ç—Ä–∞–ª—å–Ω–æ–π –ø—Ä–µ–¥–µ–ª—å–Ω–æ–π —Ç–µ–æ—Ä–µ–º—ã</h2>'))

# –ó–Ω–∞—á–µ–Ω–∏—è n –¥–ª—è –∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏—è
n_values = [2, 5, 10, 30, 100]

# –°–æ–∑–¥–∞–µ–º —Ñ–∏–≥—É—Ä—É –¥–ª—è –≥—Ä–∞—Ñ–∏–∫–æ–≤
plt.figure(figsize=(15, 12))

# –î–ª—è –∫–∞–∂–¥–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è n
for i, n in enumerate(n_values):
    # –ì–µ–Ω–µ—Ä–∏—Ä—É–µ–º —Å—Ä–µ–¥–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è
    sample_means = generate_sample_means(normal_distribution, n)
    
    # –†–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ–º –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Å–æ–≥–ª–∞—Å–Ω–æ –¶–ü–¢
    mean_theoretical = mu
    std_theoretical = sigma / np.sqrt(n)
    
    # –†–∞—Å—Å—á–∏—Ç—ã–≤–∞–µ–º —Ñ–∞–∫—Ç–∏—á–µ—Å–∫–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã
    mean_actual = np.mean(sample_means)
    std_actual = np.std(sample_means)
    
    # –°–æ–∑–¥–∞–µ–º –ø–æ–¥–≥—Ä–∞—Ñ–∏–∫
    plt.subplot(len(n_values), 1, i+1)
    
    # –°—Ç—Ä–æ–∏–º –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—É
    plt.hist(sample_means, bins=30, density=True, alpha=0.7, 
             label=f'–ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ —Å—Ä–µ–¥–Ω–∏—Ö (n={n})')
    
    # –°—Ç—Ä–æ–∏–º —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫—É—é –ø–ª–æ—Ç–Ω–æ—Å—Ç—å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏
    x = np.linspace(mu - 4*std_theoretical, mu + 4*std_theoretical, 1000)
    plt.plot(x, stats.norm.pdf(x, mean_theoretical, std_theoretical), 'r-', linewidth=2,
             label=f'–¢–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∞—è –ø–ª–æ—Ç–Ω–æ—Å—Ç—å N({mean_theoretical:.2f}, {std_theoretical:.2f})')
    
    plt.title(f'n = {n}: –¢–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ Œº = {mean_theoretical:.4f}, œÉ = {std_theoretical:.4f} | '
              f'–§–∞–∫—Ç–∏—á–µ—Å–∫–∏–µ: Œº = {mean_actual:.4f}, œÉ = {std_actual:.4f}')
    plt.xlabel('–ó–Ω–∞—á–µ–Ω–∏—è —Å–ª—É—á–∞–π–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω—ã XÃÑ')
    plt.ylabel('–ü–ª–æ—Ç–Ω–æ—Å—Ç—å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏')
    plt.legend()
    plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ 5%-–Ω–æ–≥–æ –∫–≤–∞–Ω—Ç–∏–ª—è –∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ P(1.5 ‚â§ X < 5.5)
quantile_5 = stats.norm.ppf(0.05, mu, sigma)
probability = stats.norm.cdf(5.5, mu, sigma) - stats.norm.cdf(1.5, mu, sigma)

print(f"5%-–Ω—ã–π –∫–≤–∞–Ω—Ç–∏–ª—å —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è X: {quantile_5:.4f}")
print(f"P(1.5 ‚â§ X < 5.5): {probability:.4f}")

display(HTML('<h2>7. –ê–Ω–∞–ª–∏–∑ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ –∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏—è</h2>'))

display(Markdown(r"""
### –í–∏–∑—É–∞–ª—å–Ω–∞—è –æ—Ü–µ–Ω–∫–∞ —Å—Ö–æ–∂–µ—Å—Ç–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è X –∏ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è

–ò—Å—Ö–æ–¥–Ω–∞—è —Å–ª—É—á–∞–π–Ω–∞—è –≤–µ–ª–∏—á–∏–Ω–∞ $X$ —É–∂–µ –∏–º–µ–µ—Ç –Ω–æ—Ä–º–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏
$Œº = 0$ –∏ $œÉ = 3$, –ø–æ—ç—Ç–æ–º—É –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ –≤—ã–±–æ—Ä–∫–∏ –∏ —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∞—è –∫—Ä–∏–≤–∞—è –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ 
–Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –æ—á–µ–Ω—å —Ö–æ—Ä–æ—à–æ —Å–æ–≥–ª–∞—Å—É—é—Ç—Å—è –¥—Ä—É–≥ —Å –¥—Ä—É–≥–æ–º. –≠—Ç–æ –µ—Å—Ç–µ—Å—Ç–≤–µ–Ω–Ω–æ, 
–ø–æ—Å–∫–æ–ª—å–∫—É –¥–∞–Ω–Ω—ã–µ –±—ã–ª–∏ —Å–≥–µ–Ω–µ—Ä–∏—Ä–æ–≤–∞–Ω—ã –Ω–µ–ø–æ—Å—Ä–µ–¥—Å—Ç–≤–µ–Ω–Ω–æ –∏–∑ —ç—Ç–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è.

–í–ª–∏—è–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è $n$ –Ω–∞ –±–ª–∏–∑–æ—Å—Ç—å —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è $\overline{X}$ –∫ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–º—É —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—é
–ü–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º –∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏—è –º–æ–∂–Ω–æ —Å–¥–µ–ª–∞—Ç—å —Å–ª–µ–¥—É—é—â–∏–µ –≤—ã–≤–æ–¥—ã:

1. __–ü—Ä–∏ $n = 2$__: –§–æ—Ä–º–∞ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Å—Ä–µ–¥–Ω–∏—Ö –∑–Ω–∞—á–µ–Ω–∏–π —É–∂–µ –ø—Ä–∏–±–ª–∏–∂–∞–µ—Ç—Å—è –∫ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–º—É 
—Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—é, –Ω–æ –µ—â—ë –∑–∞–º–µ—Ç–Ω—ã –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è. –°—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ —Å—Ä–µ–¥–Ω–∏—Ö –∑–Ω–∞—á–µ–Ω–∏–π 
–ø—Ä–∏–º–µ—Ä–Ω–æ —Ä–∞–≤–Ω–æ $\frac {œÉ} {\sqrt{2}} = \frac{3} {1.414} ‚âà 2.12$, —á—Ç–æ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É–µ—Ç —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–æ–º—É 
–∑–Ω–∞—á–µ–Ω–∏—é.
2. __–ü—Ä–∏ $n = 5$__: –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å—Ä–µ–¥–Ω–∏—Ö –∑–Ω–∞—á–µ–Ω–∏–π —Å—Ç–∞–Ω–æ–≤–∏—Ç—Å—è –±–æ–ª–µ–µ –ø–æ—Ö–æ–∂–∏–º –Ω–∞ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–µ, 
—Å –º–µ–Ω—å—à–∏–º —Ä–∞–∑–±—Ä–æ—Å–æ–º ($\frac {œÉ} {\sqrt{5}} = \frac{3} {2.236} ‚âà 1.34$).
3. __–ü—Ä–∏ $n = 10$__: –§–æ—Ä–º–∞ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –µ—â–µ –±–ª–∏–∂–µ –∫ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–º—É, —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ —É–º–µ–Ω—å—à–∞–µ—Ç—Å—è 
–¥–æ $\frac {œÉ} {\sqrt{10}} = \frac{3} {3.162} ‚âà 0.95$.
4. __–ü—Ä–∏ $n = 30$__: –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø—Ä–∞–∫—Ç–∏—á–µ—Å–∫–∏ –Ω–µ–æ—Ç–ª–∏—á–∏–º–æ –æ—Ç –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ, —Å –µ—â–µ –º–µ–Ω—å—à–∏–º —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω—ã–º
–æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ–º ($\frac {œÉ} {\sqrt{30}} = \frac{3} {5.477} ‚âà 0.55$).
5. __–ü—Ä–∏ $n = 100$__: –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å—Ä–µ–¥–Ω–∏—Ö –∑–Ω–∞—á–µ–Ω–∏–π –ø—Ä–∞–∫—Ç–∏—á–µ—Å–∫–∏ –∏–¥–µ–∞–ª—å–Ω–æ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É–µ—Ç –Ω–æ—Ä–º–∞–ª—å–Ω–æ–º—É 
—Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—é —Å –æ—á–µ–Ω—å –º–∞–ª—ã–º —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω—ã–º –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ–º ($\frac {œÉ} {\sqrt{100}} = \frac{3} {10} ‚âà 0.3$).

–¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, —Å —É–≤–µ–ª–∏—á–µ–Ω–∏–µ–º $n$:

- –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å—Ä–µ–¥–Ω–∏—Ö –∑–Ω–∞—á–µ–Ω–∏–π –≤—Å—ë –±–æ–ª—å—à–µ –ø—Ä–∏–±–ª–∏–∂–∞–µ—Ç—Å—è –∫ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–º—É, –¥–∞–∂–µ –µ—Å–ª–∏ –∏—Å—Ö–æ–¥–Ω–æ–µ 
—Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —É–∂–µ –±—ã–ª–æ –Ω–æ—Ä–º–∞–ª—å–Ω—ã–º.
- –†–∞–∑–±—Ä–æ—Å –∑–Ω–∞—á–µ–Ω–∏–π (—Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ) —É–º–µ–Ω—å—à–∞–µ—Ç—Å—è –ø—Ä–æ–ø–æ—Ä—Ü–∏–æ–Ω–∞–ª—å–Ω–æ $\sqrt{n}$, —á—Ç–æ –¥–µ–ª–∞–µ—Ç
—Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –±–æ–ª–µ–µ "—É–∑–∫–∏–º" –∏ —Å–∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∏—Ä–æ–≤–∞–Ω–Ω—ã–º –≤–æ–∫—Ä—É–≥ –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–≥–æ –æ–∂–∏–¥–∞–Ω–∏—è.
- –ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ —Å—Ä–µ–¥–Ω–∏—Ö –∑–Ω–∞—á–µ–Ω–∏–π —Å—Ç—Ä–µ–º–∏—Ç—Å—è –∫ –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–º—É –æ–∂–∏–¥–∞–Ω–∏—é –∏—Å—Ö–æ–¥–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è
(–≤ –¥–∞–Ω–Ω–æ–º —Å–ª—É—á–∞–µ $Œº = 0$).

### –°–æ–≥–ª–∞—Å–æ–≤–∞–Ω–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ —Å —É—Ç–≤–µ—Ä–∂–¥–µ–Ω–∏–µ–º —Ü–µ–Ω—Ç—Ä–∞–ª—å–Ω–æ–π –ø—Ä–µ–¥–µ–ª—å–Ω–æ–π —Ç–µ–æ—Ä–µ–º—ã

–ü–æ–ª—É—á–µ–Ω–Ω—ã–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã –ø–æ–ª–Ω–æ—Å—Ç—å—é —Å–æ–≥–ª–∞—Å—É—é—Ç—Å—è —Å —É—Ç–≤–µ—Ä–∂–¥–µ–Ω–∏–µ–º —Ü–µ–Ω—Ç—Ä–∞–ª—å–Ω–æ–π –ø—Ä–µ–¥–µ–ª—å–Ω–æ–π —Ç–µ–æ—Ä–µ–º—ã. –ú—ã –Ω–∞–±–ª—é–¥–∞–µ–º, —á—Ç–æ:

1. –ü—Ä–∏ —É–≤–µ–ª–∏—á–µ–Ω–∏–∏ —Ä–∞–∑–º–µ—Ä–∞ –≤—ã–±–æ—Ä–∫–∏ n, —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å—Ä–µ–¥–Ω–∏—Ö –∑–Ω–∞—á–µ–Ω–∏–π $\overline{X}$ —Å—Ç–∞–Ω–æ–≤–∏—Ç—Å—è –≤—Å—ë 
–±–ª–∏–∂–µ –∫ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–º—É —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—é.
2. –ü–∞—Ä–∞–º–µ—Ç—Ä—ã —ç—Ç–æ–≥–æ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—Ç —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏–º:
- –ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ $\overline{X}$ —Ä–∞–≤–Ω–æ –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–º—É –æ–∂–∏–¥–∞–Ω–∏—é –∏—Å—Ö–æ–¥–Ω–æ–π —Å–ª—É—á–∞–π–Ω–æ–π
–≤–µ–ª–∏—á–∏–Ω—ã X (–≤ –¥–∞–Ω–Ω–æ–º —Å–ª—É—á–∞–µ 0).
- –°—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ $\overline{X}$ —Ä–∞–≤–Ω–æ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–º—É –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—é X, –¥–µ–ª–µ–Ω–Ω–æ–º—É 
–Ω–∞ –∫–≤–∞–¥—Ä–∞—Ç–Ω—ã–π –∫–æ—Ä–µ–Ω—å –∏–∑ n ($\frac {œÉ} {\sqrt{n}}).
3. –î–∞–∂–µ –¥–ª—è –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ –Ω–µ–±–æ–ª—å—à–∏—Ö –∑–Ω–∞—á–µ–Ω–∏–π n (–Ω–∞–ø—Ä–∏–º–µ—Ä, n = 10) –º—ã —É–∂–µ –≤–∏–¥–∏–º —Ö–æ—Ä–æ—à–µ–µ –ø—Ä–∏–±–ª–∏–∂–µ–Ω–∏–µ 
–∫ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–º—É —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—é.

–í–∞–∂–Ω–æ –æ—Ç–º–µ—Ç–∏—Ç—å, —á—Ç–æ –≤ –Ω–∞—à–µ–º —Å–ª—É—á–∞–µ –∏—Å—Ö–æ–¥–Ω–∞—è —Å–ª—É—á–∞–π–Ω–∞—è –≤–µ–ª–∏—á–∏–Ω–∞ X —É–∂–µ –∏–º–µ–ª–∞ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ, –ø–æ—ç—Ç–æ–º—É –∏ —Å—Ä–µ–¥–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è —Ç–∞–∫–∂–µ –∏–º–µ—é—Ç –Ω–æ—Ä–º–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø—Ä–∏ –ª—é–±–æ–º n. –û–¥–Ω–∞–∫–æ, —Ü–µ–Ω—Ç—Ä–∞–ª—å–Ω–∞—è –ø—Ä–µ–¥–µ–ª—å–Ω–∞—è —Ç–µ–æ—Ä–µ–º–∞ —É—Ç–≤–µ—Ä–∂–¥–∞–µ—Ç, —á—Ç–æ —ç—Ç–æ—Ç —Ä–µ–∑—É–ª—å—Ç–∞—Ç –±—É–¥–µ—Ç –≤–µ—Ä–µ–Ω –∏ –¥–ª—è –±–æ–ª—å—à–∏–Ω—Å—Ç–≤–∞ –¥—Ä—É–≥–∏—Ö —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–π –ø—Ä–∏ –¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ –±–æ–ª—å—à–æ–º n, —á—Ç–æ –¥–µ–ª–∞–µ—Ç –µ—ë –æ–¥–Ω–∏–º –∏–∑ –Ω–∞–∏–±–æ–ª–µ–µ —Ñ—É–Ω–¥–∞–º–µ–Ω—Ç–∞–ª—å–Ω—ã—Ö —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ —Ç–µ–æ—Ä–∏–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π.
"""))
````
