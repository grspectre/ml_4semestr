### Создание простейшего графика.

####**Создание области построения**

Метод **figure()** модуля **pyplot**  создает объект **Figure** (экземпляр класса  **Figure**) - область построения, в которой будут размещаться все элементы графика.
Можно определять размеры области, цвет фона, границы и т. п. с помощью соответствующих параметров.

[Документация метода figure() с описанием всех доступных параметров](https://matplotlib.org/stable/api/figure_api.html#module-matplotlib.figure)

 [Документация класса Figure с описанием всех доступных параметров и методов](https://matplotlib.org/stable/api/figure_api.html)

\
####**Создание линейного графика**

Метод **plot()** модуля **pyplot**  добавляет линию в области построения (линия - экземпляр класса  **Line2D**).

Можно определять тип, цвет, толщину линии, маркеры и т. п. с помощью соответствующих параметров.

[Документация метода plot() с описанием всех доступных настроек](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html)

```python
x = np.linspace(0, 5, 11)                   # генерация массива аргументов
y = x**2                                    # генерация значений функции y(x)

plt.figure()                                # создание объекта - области построения (экземпляра класса Figure) с параметрами по умолчанию
plt.plot(x, y, 'r')                         # построение линии с координатами точек, заданными в массивах x и y, красным цветом (остальные параметры графика - по умолчанию)

z = x**3                                    # генерация значений функции z(x)
plt.plot(x, z, 'b')                         # добавление новой линии на график: координаты точек в массивах x и z, синим цветом

plt.xlabel('x')                             # определение подписи оси абсцисс
plt.ylabel('y, z')                          # определение подписи оси ординат
plt.title('Графики $y = x^2$ и $z = x^3$')  # определение заголовка
```

```python
plt.figure()                                            # создание объекта - области построения (экземпляра класса Figure) с параметрами по умолчанию
plt.plot(x, y, color='k', marker='o')                   # построение черной сплошной (по умолчанию) линии с круглыми маркерами
plt.plot(x, z, color='k', linestyle='--', marker='s')   # добавление черной пунктирной линии с квадратными маркерами
plt.xlabel('x')                                         # определение подписи оси абсцисс
plt.ylabel('y, z')                                      # определение подписи оси ординат
plt.title('Графики $y = x^2$ и $z = x^3$')              # определение заголовка
```

### Задание размеров графика и добавление легенды

```python
plt.figure(figsize = (7, 4))                   # создание области построения с заданными размерами (размеры в дюймах)
plt.plot(x, x**2, 'r', label = '$y = x^2$')    # параметр label определяет надписи для легенды
plt.plot(x, x**3, 'b', label = '$z = x^3$')
plt.xlabel('x')
plt.ylabel('y, z')
plt.title('Графики $y = x^2$ и $z = x^3$')
plt.legend(loc='upper left')                   # создание легенды (параметр loc определяет положение легенды)
```

[Документация метода legend() с описанием всех доступных настроек](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html)

```python
plt.figure(figsize = (7, 4))
plt.plot(x, x**2, 'r', label = '$y = x^2$')
plt.plot(x, x**3, 'b', label = '$z = x^3$')
plt.xlabel('x')
plt.ylabel('y, z')
plt.title('Графики $y = x^2$ и $z = x^3$')
plt.legend(loc="upper right", bbox_to_anchor=(1.3, 0.9))    # расположение легенды вне поля графика с указанием координат
```

### Создание нескольких графиков в раздельных системах координат

Метод **subplots()** модуля **pyplot**  создает область построения с набором подграфиков (возвращает экземпляр класса **Figure** и массив подграфиков - экземпляров класса **Axes**).

Можно определять количество строк и столбцов в области построения, в которых будут размещены подграфики, а также другие параметры.

[Документация метода subplots() с описанием всех доступных настроек](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html#matplotlib.pyplot.subplots)

Параметры каждого подграфика задаются путем обращения к соответствующему объекту Axes (элементу массива, созданного методом subplots() ).

[Документация класса Axes с описанием всех доступных настроек](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.html#matplotlib.axes.Axes)

```python
fig, axes = plt.subplots(nrows = 1, ncols = 3, figsize = (10, 4))   # Создание подграфиков, размещенных в 1 строке и 3 столбцах; параметр figsize определяет размеры ВСЕЙ области построения
i = 2
for ax in axes:                                                     # Цикл по всем объектам Axes (подграфикам). Для каждого подграфика
    ax.plot(x, x**i, 'r')                                               # определение линии, выводимой на этом подграфике
    ax.set_xlabel('x')                                                  # определение параметров подграфиков (подписей по осям и заголовков)
    ax.set_ylabel('y')
    ax.set_title('$y = x^%s$' % i)
    i += 1
fig.tight_layout()                                                  # расположение графиков без наложения
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize = (7, 8))   # Создание подграфиков, размещенных в 2 строках и 2 столбцах
i = 1
for axs in axes:                                                   # Цикл по строкам (первый индекс axes)
  for ax in axs:                                                   # цикл по столбцам (второй индекс axes)
    ax.plot(x, x**i, 'r')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('$y = x^%s$' % i)
    i += 1
fig.tight_layout()
```

К отдельным подграфикам можно обращаться по индексам массива axes

```python
fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize = (7, 8))   # Создание подграфиков, размещенных в 2 строках и 2 столбцах
i = 1
for axs in axes:                                                   # Цикл по строкам (первый индекс axes)
  for ax in axs:                                                   # цикл по столбцам (второй индекс axes)
    ax.plot(x, x**i, 'r')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('$y = x^%s$' % i)
    i += 1
fig.tight_layout()

axes[0][0].set_title('Линейная зависимость $y = x$')               # Изменение параметров отдельных подграфиков
axes[0][1].set_title('Квадратичная зависимость $y = x^2$')
```

### Совмещение графиков

Метод **add_axes()** класса **Figure**  добавляет подграфик (экземпляр **Axes**) в область построения.

Можно определять положение начала координат и размеры каждого подграфика, тип системы координат, а также другие параметры.
Полное описание метода - в документации класса **Figure** (см. выше).

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.9, 0.9])    # Определение положения начала координат для каждой пары осей (первые два элемента списка)
axes2 = fig.add_axes([0.17, 0.65, 0.4, 0.3])   # и размеров графиков в долях от размеров всей области построения (последние два два элемента списка)

axes1.plot(x, y, 'r')
axes1.set_xlabel('x')
axes1.set_ylabel('y')

axes2.plot(y, x, 'g')
axes2.set_xlabel('y')
axes2.set_ylabel('x')

axes1.set_title('Совмещенные графики $y = x^2$ и обратной ей функции')
```

### Scatter (облако точек)

Метод **scatter()** модуля **pyplot**  реализует построение "облака точек"  -  точек с указанными массивами координат.


[Документация метода scatter() с описанием всех доступных настроек](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html)

```python
plt.figure(figsize = (7, 4))
x = np.linspace(0, 5, 500)                             # создание массива аргументов
y = x + np.random.randn(x.size)                        # создание массива значений функции y=x с добавлением случайных значений (для получения разброса)

plt.scatter(x, y, color = 'g')                         # построение облака точек (зеленого цвета)
plt.plot(x, x, 'r')                                    # построение линии тренда (зависимости y=x без учета разброса)

plt.xlabel('x')
plt.ylabel('y')
plt.title('Облако точек (линейный тренд с рассеянием)')
```

### Гистограмма

```python
n = np.random.randn(100000)
fig, axes = plt.subplots(1, 2, figsize=(12,4))
axes[0].hist(n)
axes[0].set_xlim(min(n), max(n))
axes[0].set_title('Гистограмма c параметрами по умолчанию')
axes[1].hist(n, cumulative = True, bins = 50)
axes[1].set_xlim(min(n), max(n))
axes[1].set_title('Кумулятивная гистограмма с 50 интервалами')
```
