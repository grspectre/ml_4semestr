### Пояснение
В удобочитаемой форме это можно переписать следующим способом:
```python
arr = []
for i in np.linspace(-2, 12, 15):
    arr.append([])
    for j in range(1, 8):
        arr[len(arr)-1].append(3 * i + j)
arr = np.array(arr)
```

То есть Исходный массив создаётся с помощью __list comprehension__, где внешний цикл перебирает значения переменной $i$, полученные из функции `np.linspace(-2, 12, 15)`. Эта функция формирует __15__ равномерно распределённых чисел __от −2 до 12__. Внутренний цикл с переменной $j$ пробегает целые значения __от 1 до 7__. Для каждого значения $i$ вычисляется выражение $3 \cdot i + j$. Таким образом, каждая строка создаётся для конкретного $i$ и состоит из __7__ элементов, в которых значение $j$ последовательно изменяется __от 1 до 7__. Результирующий двумерный массив имеет размерность `15 × 7`.

````python
import numpy as np

# Для воспроизводимости результатов устанавливаем seed
np.random.seed(42)

print('Задача 3')

print("""Генерируем матрицы A и B заданных размерностей:
A имеет размерность 5 x 7, B – 7 x 5.
Здесь для простоты используем целочисленные псевдослучайные значения в диапазоне [1, 20).""")
A = np.random.randint(1, 20, size=(5, 7))
B = np.random.randint(1, 20, size=(7, 5))

print("Матрица A (5x7):\n", A)
print("\nМатрица B (7x5):\n", B)

print("""1. Находим произведение A * B и B * A
Умножение матриц производится с помощью оператора @""")

prod_A_B = A @ B  # Результат имеет размерность 5 x 5, так как (5 x 7) * (7 x 5) = (5 x 5)
prod_B_A = B @ A  # Результат имеет размерность 7 x 7, так как (7 x 5) * (5 x 7) = (7 x 7)

print("\nПроизведение A * B (размер 5x5):\n", prod_A_B)
print("\nПроизведение B * A (размер 7x7):\n", prod_B_A)

print("""Комментарий к полученным результатам:
Важно заметить, что умножение матриц не является коммутативной операцией. 
В данном случае A * B дает квадратную матрицу размером 5 x 5, а B * A – квадратную матрицу размером 7 x 7. 
Это различие обусловлено размерами исходных матриц.""")

print("""2. Вычисляем матрицу C по формуле: C = B * A + 3E, где E – единичная матрица.
Поскольку произведение B * A имеет размер 7 x 7, то и единичная матрица E должна быть размером 7 x 7.""")
E = np.eye(7)
C = prod_B_A + 3 * E

print("\nМатрица C = B * A + 3E:\n", C)

print("3. Вычисление определителя матрицы C")
det_C = np.linalg.det(C)
print("\nОпределитель матрицы C: {:.3f}".format(det_C))

print("""4. Проверяем, существует ли обратная матрица для C. 
Если определитель не равен нулю (с учётом численных погрешностей), матрица обратима.""")
epsilon = 1e-10  # допустимый порог для сравнений с нулём, т.к. сравниваем float

if abs(det_C) > epsilon:
    C_inv = np.linalg.inv(C)
    print("\nОбратная матрица для C:\n", C_inv)
else:
    print("\nМатрица C вырождена (определитель равен нулю), обратной матрицы не существует.")

print("\n\nЗадача 4")

min_rows_np = np.min(A, axis=1)

min_rows_loop = []
for i in range(A.shape[0]):
    row = A[i, :]
    min_rows_loop.append(np.min(row))  # можно также использовать встроенную функцию min(row)
min_rows_loop = np.array(min_rows_loop)

print("\nМинимальные значения по строкам:")
print("Встроенная функция numpy:", min_rows_np)
print("С использованием цикла:   ", min_rows_loop)

min_cols_np = np.min(A, axis=0)
min_cols_loop = []
for j in range(A.shape[1]):
    col = A[:, j]
    min_cols_loop.append(np.min(col))  # или min(col)
min_cols_loop = np.array(min_cols_loop)

print("\nМинимальные значения по столбцам:")
print("Встроенная функция numpy:", min_cols_np)
print("С использованием цикла:   ", min_cols_loop)

min_all_np = np.min(A)
min_all_loop = A[0, 0]
for i in range(A.shape[0]):
    for j in range(A.shape[1]):
        if A[i, j] < min_all_loop:
            min_all_loop = A[i, j]
print("\nМинимальное значение по всей матрице A:")
print("Встроенная функция numpy:", min_all_np)
print("С использованием цикла:   ", min_all_loop)

g = np.sqrt(A.astype(float))

print("\nРезультат применения функции g(z)=sqrt(z) к матрице A:\n", g)
````

### Задача 8.
Найти минимум указанной функции (по вариантам). Реализовать запуск алгоритма из нескольких начальных точек, проанализировать полученные результаты.

По результатам анализа сформировать небольшой отчет:
- повлиял ли выбор начальной точки на получаемые результаты; если да, то каким образом;
- какое из начальных приближений можно считать лучшим и почему;
- для наиболее удачного начального приближения
  - какое количество итераций понадобилось для завершения работы алгоритма?
  - какое значение аргумента найдено в качестве min функции?
  - каково значение функции в найденной точке min?
  - каково значение градиента (для функции одной переменной–производной) в найденной точке min?
- как можно в целом оценить успешность выполненного поиска.

Указание: для более глубокого понимания результатов экспериментов рекомендуется также найти точку минимума данной функции аналитическими методами (из курса математического анализа).

__Данные__:

$$
\frac {x + 1} {(x - 1)^2}
$$

````python
import numpy as np
from scipy.optimize import minimize

# Определим функцию с защитой от деления на ноль (при x близком к 1)
def f(x):
    # x приходит как массив размерности (1,), поэтому обращаемся к x[0]
    if np.abs(x[0] - 1) < 1e-8:
        return 1e10  # большое значение, если x почти равен 1
    return (x[0] + 1) / ((x[0] - 1) ** 2)

# Зададим аналитическую производную f'(x) = - (x+3)/(x-1)^3
def grad_f(x):
    if np.abs(x[0] - 1) < 1e-8:
        return np.array([1e10])
    return np.array([ - (x[0] + 3) / ((x[0] - 1) ** 3) ])

# Множество начальных приближений
initial_points = {
    "x0 = -4": [-4.0],
    "x0 = -3": [-3.0],
    "x0 = -2": [-2.0],
    "x0 = 0": [0.0],
    "x0 = 2": [2.0]
}

results = {}

print("Поиск минимума функции f(x)= (x+1)/(x-1)^2\n")
for label, x0 in initial_points.items():
    # Попробуем метод BFGS с использованием аналитического градиента
    res = minimize(f, x0, method="BFGS", jac=grad_f, options={"disp": False})
    results[label] = res
    print(f"{label}")
    # Для некоторых методов (например, BFGS) число итераций можно получить из res.nit
    if hasattr(res, "nit"):
        iters = res.nit
    else:
        iters = "не определено"
    print(f"  Итераций: {iters}")
    print(f"  Найденное значение аргумента: {res.x[0]:.5f}")
    print(f"  Значение функции: {res.fun:.5f}")
    # Вычислим значение градиента в найденной точке
    grad_val = grad_f(res.x)[0]
    print(f"  Градиент в найденной точке: {grad_val:.5e}\n")
````

## Анализ результатов эксперимента
При запуске алгоритма оптимизации из различных начальных точек получаем следующие наблюдения:

### Влияние выбора начальной точки:
При начальных точках, находящихся в левом отрезке (например, $x_0 = −4$, $x_0 = −3$, $x_0 = −2$), алгоритм стабильно сходится к точке 
$x ≈ −3$, что соответствует аналитически найденному минимума. При этом значение функции становится примерно $−0.125$.

Если запускать оптимизацию с начальной точкой, принадлежащей правой части (например, $x_0 = 2$ или $x_0 = 0$), алгоритм может столкнуться с проблемой перехода через точку $x = 1$ (близость которой вызывает резкий рост функции из-за деления на малую величину) или же он направляется к минимуму в области, где функция принимает положительные значения, стремясь к пределу 0 при $x → +∞$. Таким образом, выбор начальной точки значительно влияет на область сходимости алгоритма.

### Лучший выбор начального приближения:

Исходя из экспериментов, лучшим оказывается начальное приближение из левого отрезка (например, $x_0 = −3$ или $x_0 = −2$). Они приводят к минимальному значению функции, равному $−0.125$, что соответствует аналитически полученной точке минимума.

Такой выбор оправдан, поскольку аналитический анализ показывает, что критическая точка $x = −3$ даёт глобальный минимум, а оптимизация из этой области сходится к нужному решению быстро и без «опасных» областей (особенности в точке $x = 1$).

### Детали для наиболее удачного приближения (например, $x_0 = −3):

- Количество итераций: число итераций зависит от работы оптимизатора; в примере с методом BFGS это значение выводится в res.nit (обычно оно невелико, порядка нескольких итераций).
- Найденное значение аргумента минимума: $x ≈ −3.00000$.
- Значение функции в найденной точке: $f(x) ≈ −0.12500$.
- Значение градиента в найденной точке: $f′(x)$ практически равно нулю (например, порядка $1e−7$ или меньше), что подтверждает достижение стационарной точки.

### Общая оценка успешности поиска:

При разумном выборе начальных приближений (из области, избегающей особенностей) классический алгоритм оптимизации (например, метод BFGS) успешно сходится к глобальному минимуму. Анализ показывает, что применение аналитического градиента дополнительно помогает алгоритму быстро приближаться к решению. Однако важным является контроль за выбором области поиска – поскольку у функции есть особенность в точке $x = 1$, оптимизацию имеет смысл проводить с начальными точками, находящимися «безопасно» от разрыва.

````python
import numpy as np
import matplotlib.pyplot as plt

# Определяем функции
def f(x):
    return (x + 1) / (x - 1)**2

def f_prime(x):
    return -(x + 3) / (x - 1)**3

# Настраиваем диапазоны для трех подграфиков
# Первый диапазон: от -6 до -2 (около критической точки x = -3)
x1 = np.linspace(-6, -2, 400)
# Второй диапазон: от -1 до 0.9 (приближение к асимптоте с левой стороны)
x2 = np.linspace(-1, 0.9, 400)
# Третий диапазон: от 1.1 до 6 (правая ветвь после асимптоты)
x3 = np.linspace(1.1, 6, 400)

# Создаем фигуру с 2 рядами и 3 столбцами подграфиков
fig, axs = plt.subplots(nrows=2, ncols=3, figsize=(16, 8))
fig.suptitle("Графики функции f(x) и её производной f'(x)", fontsize=16)

# Список диапазонов и соответствующих значений для удобства обхода
ranges = [(x1, 'Диапазон: [-6, -2]'), (x2, 'Диапазон: [-1, 0.9]'), (x3, 'Диапазон: [1.1, 6]')]

# Первый ряд: графики исходной функции f(x)
for idx, (x_range, title_range) in enumerate(ranges):
    ax = axs[0, idx]
    ax.plot(x_range, f(x_range),
            color='darkcyan',    # Выбираем насыщенный цвет для выделения графика
            linestyle='--',      # Пунктирный стиль линии для наглядного представления графика
            linewidth=2)         # Толщина линии увеличена, чтобы график хорошо смотрелся на печати и экране
    ax.set_title(f"График f(x)\n{title_range}", fontsize=12)
    ax.set_xlabel("x", fontsize=10)
    ax.set_ylabel("f(x)", fontsize=10)
    ax.grid(True)

# Второй ряд: графики производной f'(x)
for idx, (x_range, title_range) in enumerate(ranges):
    ax = axs[1, idx]
    ax.plot(x_range, f_prime(x_range),
            color='magenta',    # Яркий цвет для отличия от исходного графика функции
            linestyle=':',      # Точечный стиль линии для визуального различия от пунктирного графика функции
            linewidth=2)        # Толщина линии, аналогично первому ряду
    ax.set_title(f"График f'(x)\n{title_range}", fontsize=12)
    ax.set_xlabel("x", fontsize=10)
    ax.set_ylabel("f'(x)", fontsize=10)
    ax.grid(True)

plt.tight_layout(rect=[0, 0, 1, 0.95])  # Настройка расположения графиков, оставляем место для общего заголовка
plt.show()
````

## Пояснения и комментарии
В данном коде:
- Функция np.linspace используется для равномерного разбиения выбранных диапазонов, что обеспечивает плавность линий графика.
Каждый подграфик создаётся с индивидуальными подписями: заголовок содержит информацию о том, какой именно диапазон значений представлен, а подписи осей позволяют понимать, какие величины отображаются.
- Визуальное оформление (цвет, тип и толщина линии) задается явно через соответствующие параметры функции plot. Это позволяет отделить внешний вид графиков от настроек по умолчанию и сделать их более наглядными.
- Сетка (ax.grid(True)) помогает лучше ориентироваться в координатной плоскости и отслеживать поведение функции и её производной.
- Расположение графиков в двух рядах и трех столбцах позволяет удобно сравнивать поведение исходной функции и её производной по выбранным диапазонам.
### Связь с решением задачи 8

Из результатов задачи 8 мы знаем, что функция имеет критическую точку при $x = −3$ и разрыв (вертикальную асимптоту) в точке $x = 1$. На графиках:

- В первом столбце (диапазон $[−6, −2]$) видно, что функция достигает минимума в окрестности $x = −3$. График производной в этом диапазоне пересекает ось абсцисс именно в этой точке, что соответствует условию экстремума.
- Во втором столбце наблюдается резкое изменение поведения функции при приближении к $x = 1$ (в данном случае до $0.9$ включительно), что отображается также и на графике производной.
- В третьем столбце (диапазон $[1.1, 6]$) графики демонстрируют ветвь функции и соответствующую производную после разрыва, что позволяет оценить асимптотическое поведение функции справа от вертикальной асимптоты.

### Выводы
Из полученных графиков можно сделать следующие выводы:

- Критическая точка $x = −3$ четко видна как минимум функции на первом подграфике, что подтверждает результаты аналитического исчисления.
- Поведение функции вблизи вертикальной асимптоты $x = 1$ ясно прослеживается: слева функция стремится к бесконечности, что свидетельствует о разрыве.
- График производной дополнительно демонстрирует, как меняется наклон функции на различных участках, помогая интерпретировать рост/убывание исходной функции.
- Выбранные цвета и стили линий делают различие между графиками функции и её производной, что упрощает визуальный анализ.

Данный подход демонстрирует, как с помощью библиотеки Matplotlib можно не только построить графики, но и обеспечить их грамотное структурирование и визуальное оформление для глубокого анализа полученных результатов.
