```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
from scipy.stats import uniform
from IPython.display import display, Markdown, HTML

display(HTML('<h2>2. –°–æ–∑–¥–∞–Ω–∏–µ —Å–ª—É—á–∞–π–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω—ã —Å —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω—ã–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ–º</h2>'))
# –ü–∞—Ä–∞–º–µ—Ç—Ä—ã —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è
a = 2  # –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ
b = 10  # –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ

# –°–æ–∑–¥–∞–Ω–∏–µ —Å–ª—É—á–∞–π–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω—ã —Å —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω—ã–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ–º
X_uniform = uniform(loc=a, scale=b-a)

# –ì–µ–Ω–µ—Ä–∞—Ü–∏—è –≤—ã–±–æ—Ä–∫–∏ –¥–ª—è –¥–µ–º–æ–Ω—Å—Ç—Ä–∞—Ü–∏–∏
sample_uniform = X_uniform.rvs(size=1000)

print(f"–ü–µ—Ä–≤—ã–µ 10 —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –≤—ã–±–æ—Ä–∫–∏ –∏–∑ —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –Ω–∞ [2, 10]:")
print(sample_uniform[:10])
display(HTML('<h2>3. –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–æ–≤</h2>'))
display(HTML('<h3>3.1 –ì—Ä–∞—Ñ–∏–∫ —Ñ—É–Ω–∫—Ü–∏–∏ –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π</h3>'))

# –ó–∞–¥–∞–µ–º —Ç–æ—á–∫–∏ –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–æ–≤
x = np.linspace(0, 12, 1000)

# –ì—Ä–∞—Ñ–∏–∫ —Ñ—É–Ω–∫—Ü–∏–∏ –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π
plt.figure(figsize=(10, 6))
plt.plot(x, X_uniform.pdf(x), 'r-', lw=2, label=f'f(x), a={a}, b={b}')
plt.grid(True, alpha=0.3)
plt.title('–§—É–Ω–∫—Ü–∏—è –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.axvline(x=a, color='g', linestyle='--', label=f'x = {a}')
plt.axvline(x=b, color='b', linestyle='--', label=f'x = {b}')
plt.fill_between(x, X_uniform.pdf(x), where=((x >= a) & (x <= b)), color='red', alpha=0.3)
plt.legend()
plt.show()

display(HTML('<h3>3.2 –ì—Ä–∞—Ñ–∏–∫ —Ñ—É–Ω–∫—Ü–∏–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è</h3>'))

# –ì—Ä–∞—Ñ–∏–∫ —Ñ—É–Ω–∫—Ü–∏–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è
plt.figure(figsize=(10, 6))
plt.plot(x, X_uniform.cdf(x), 'b-', lw=2, label=f'F(x), a={a}, b={b}')
plt.grid(True, alpha=0.3)
plt.title('–§—É–Ω–∫—Ü–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.axhline(y=1, color='k', linestyle='-', alpha=0.3)
plt.axvline(x=a, color='g', linestyle='--', label=f'x = {a}')
plt.axvline(x=b, color='b', linestyle='--', label=f'x = {b}')
plt.legend()
plt.show()

display(HTML('<h2>4. –í—ã—á–∏—Å–ª–µ–Ω–∏–µ —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫</h2>'))
display(HTML('<h3>4.1 –ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ, –¥–∏—Å–ø–µ—Ä—Å–∏—è –∏ —Å—Ä–µ–¥–Ω–µ–µ –∫–≤–∞–¥—Ä–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ</h3>'))

# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ —á–∏—Å–ª–æ–≤—ã—Ö —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫
mean = X_uniform.mean()
var = X_uniform.var()
std = X_uniform.std()

print(f"–ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ: {mean:.4f}")
print(f"–î–∏—Å–ø–µ—Ä—Å–∏—è: {var:.4f}")
print(f"–°—Ä–µ–¥–Ω–µ–µ –∫–≤–∞–¥—Ä–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ: {std:.4f}")


display(HTML('<h3>4.2 –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –º–µ–¥–∏–∞–Ω—ã –∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ P(3 < X < 7)</h3>'))

# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –º–µ–¥–∏–∞–Ω—ã
median = X_uniform.median()
print(f"–ú–µ–¥–∏–∞–Ω–∞: {median:.4f}")

# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ P(3 < X < 7)
prob = X_uniform.cdf(7) - X_uniform.cdf(3)
print(f"P(3 < X < 7): {prob:.4f}")

display(HTML('<h3>4.3 –ò–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è –ø–æ–ª—É—á–µ–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π</h3>'))

display(Markdown(r"""
–ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –Ω–∞ –∏–Ω—Ç–µ—Ä–≤–∞–ª–µ $[a, b]$ —Ä–∞–≤–Ω–æ 
$(a+b)/2 = (2+10)/2=6$, —á—Ç–æ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É–µ—Ç —Å–µ—Ä–µ–¥–∏–Ω–µ –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞. –≠—Ç–æ –æ–∑–Ω–∞—á–∞–µ—Ç, 
—á—Ç–æ –ø—Ä–∏ –º–Ω–æ–≥–æ–∫—Ä–∞—Ç–Ω–æ–º –ø–æ–≤—Ç–æ—Ä–µ–Ω–∏–∏ —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞ —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —Å–ª—É—á–∞–π–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω—ã 
–±—É–¥–µ—Ç —Å—Ç—Ä–µ–º–∏—Ç—å—Å—è –∫ 6.

–î–∏—Å–ø–µ—Ä—Å–∏—è —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Ä–∞–≤–Ω–∞ 
${(b‚àía)}^2/12=(10‚àí2)^2/12=64/12‚âà5.33$. –≠—Ç–æ –º–µ—Ä–∞ —Ä–∞–∑–±—Ä–æ—Å–∞ –∑–Ω–∞—á–µ–Ω–∏–π 
—Å–ª—É—á–∞–π–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω—ã –≤–æ–∫—Ä—É–≥ –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–≥–æ –æ–∂–∏–¥–∞–Ω–∏—è.

–°—Ä–µ–¥–Ω–µ–µ –∫–≤–∞–¥—Ä–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ (–°–ö–û) —Ä–∞–≤–Ω–æ 
$\sqrt {{(b‚àía)}^2} / 12 = \sqrt {64/12} ‚âà 2.31$. –°–ö–û –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç, 
–Ω–∞—Å–∫–æ–ª—å–∫–æ –≤ —Å—Ä–µ–¥–Ω–µ–º –∑–Ω–∞—á–µ–Ω–∏—è —Å–ª—É—á–∞–π–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω—ã –æ—Ç–∫–ª–æ–Ω—è—é—Ç—Å—è 
–æ—Ç –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–≥–æ –æ–∂–∏–¥–∞–Ω–∏—è.

–ú–µ–¥–∏–∞–Ω–∞ —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Å–æ–≤–ø–∞–¥–∞–µ—Ç —Å –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–∏–º 
–æ–∂–∏–¥–∞–Ω–∏–µ–º –∏ —Ä–∞–≤–Ω–∞ 6. –≠—Ç–æ –æ–∑–Ω–∞—á–∞–µ—Ç, —á—Ç–æ –ø–æ–ª–æ–≤–∏–Ω–∞ –≤—Å–µ—Ö –≤–æ–∑–º–æ–∂–Ω—ã—Ö 
–∑–Ω–∞—á–µ–Ω–∏–π —Å–ª—É—á–∞–π–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω—ã –º–µ–Ω—å—à–µ 6, –∞ –ø–æ–ª–æ–≤–∏–Ω–∞ - –±–æ–ª—å—à–µ.

–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ä–∞–≤–Ω–∞ 

$\frac {7‚àí3} {b‚àía}=\frac{4}{8}=0.5$. –≠—Ç–æ –æ–∑–Ω–∞—á–∞–µ—Ç, —á—Ç–æ —Å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å—é
$0.5$ (–∏–ª–∏ 50%) —Å–ª—É—á–∞–π–Ω–∞—è –≤–µ–ª–∏—á–∏–Ω–∞ X –ø—Ä–∏–º–µ—Ç –∑–Ω–∞—á–µ–Ω–∏–µ –∏–∑ –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞
(3, 7).
"""))

display(HTML('<h2>5. –í–ª–∏—è–Ω–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –Ω–∞ —Ö–∞—Ä–∞–∫—Ç–µ—Ä –∑–∞–∫–æ–Ω–∞ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è</h2>'))
display(HTML('<h3>5.1 –í–ª–∏—è–Ω–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –Ω–∞ —Ñ—É–Ω–∫—Ü–∏—é –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏</h3>'))


# –°–æ–∑–¥–∞–µ–º –Ω–µ—Å–∫–æ–ª—å–∫–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–π —Å —Ä–∞–∑–Ω—ã–º–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏
X1 = uniform(loc=2, scale=8)  # a=2, b=10
X2 = uniform(loc=4, scale=8)  # a=4, b=12
X3 = uniform(loc=6, scale=8)  # a=6, b=14
X4 = uniform(loc=2, scale=4)  # a=2, b=6
X5 = uniform(loc=2, scale=12) # a=2, b=14

# –ì—Ä–∞—Ñ–∏–∫ –ø—Ä–∏ –∏–∑–º–µ–Ω–µ–Ω–∏–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ a (–ø—Ä–∏ —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω–æ–º b-a)
plt.figure(figsize=(12, 6))
x = np.linspace(0, 16, 1000)

plt.subplot(1, 2, 1)
plt.plot(x, X1.pdf(x), 'r-', lw=2, label='a=2, b=10')
plt.plot(x, X2.pdf(x), 'g-', lw=2, label='a=4, b=12')
plt.plot(x, X3.pdf(x), 'b-', lw=2, label='a=6, b=14')
plt.grid(True, alpha=0.3)
plt.title('–í–ª–∏—è–Ω–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ a –Ω–∞ —Ñ—É–Ω–∫—Ü–∏—é –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏\n(–ø—Ä–∏ —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω–æ–º b-a=8)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()

# –ì—Ä–∞—Ñ–∏–∫ –ø—Ä–∏ –∏–∑–º–µ–Ω–µ–Ω–∏–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ b-a (–ø—Ä–∏ —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω–æ–º a)
plt.subplot(1, 2, 2)
plt.plot(x, X4.pdf(x), 'r-', lw=2, label='a=2, b=6')
plt.plot(x, X1.pdf(x), 'g-', lw=2, label='a=2, b=10')
plt.plot(x, X5.pdf(x), 'b-', lw=2, label='a=2, b=14')
plt.grid(True, alpha=0.3)
plt.title('–í–ª–∏—è–Ω–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ b –Ω–∞ —Ñ—É–Ω–∫—Ü–∏—é –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏\n(–ø—Ä–∏ —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω–æ–º a=2)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()

plt.tight_layout()
plt.show()

display(HTML('<h3>5.2 –í–ª–∏—è–Ω–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –Ω–∞ —Ñ—É–Ω–∫—Ü–∏—é —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è</h3>'))

# –ì—Ä–∞—Ñ–∏–∫ –§–† –ø—Ä–∏ –∏–∑–º–µ–Ω–µ–Ω–∏–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ a (–ø—Ä–∏ —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω–æ–º b-a)
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(x, X1.cdf(x), 'r-', lw=2, label='a=2, b=10')
plt.plot(x, X2.cdf(x), 'g-', lw=2, label='a=4, b=12')
plt.plot(x, X3.cdf(x), 'b-', lw=2, label='a=6, b=14')
plt.grid(True, alpha=0.3)
plt.title('–í–ª–∏—è–Ω–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ a –Ω–∞ —Ñ—É–Ω–∫—Ü–∏—é —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è\n(–ø—Ä–∏ —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω–æ–º b-a=8)')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.legend()

# –ì—Ä–∞—Ñ–∏–∫ –§–† –ø—Ä–∏ –∏–∑–º–µ–Ω–µ–Ω–∏–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ b-a (–ø—Ä–∏ —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω–æ–º a)
plt.subplot(1, 2, 2)
plt.plot(x, X4.cdf(x), 'r-', lw=2, label='a=2, b=6')
plt.plot(x, X1.cdf(x), 'g-', lw=2, label='a=2, b=10')
plt.plot(x, X5.cdf(x), 'b-', lw=2, label='a=2, b=14')
plt.grid(True, alpha=0.3)
plt.title('–í–ª–∏—è–Ω–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ b –Ω–∞ —Ñ—É–Ω–∫—Ü–∏—é —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è\n(–ø—Ä–∏ —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω–æ–º a=2)')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.legend()

plt.tight_layout()
plt.show()

display(HTML('<h3>5.3 –í—ã–≤–æ–¥—ã –æ –≤–ª–∏—è–Ω–∏–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –Ω–∞ —Ö–∞—Ä–∞–∫—Ç–µ—Ä –∑–∞–∫–æ–Ω–∞ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è</h3>'))

display(Markdown(r"""
–ü—Ä–æ–∞–Ω–∞–ª–∏–∑–∏—Ä–æ–≤–∞–≤ –≥—Ä–∞—Ñ–∏–∫–∏, –º–æ–∂–Ω–æ —Å–¥–µ–ª–∞—Ç—å —Å–ª–µ–¥—É—é—â–∏–µ –≤—ã–≤–æ–¥—ã –æ –≤–ª–∏—è–Ω–∏–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –Ω–∞ —Ö–∞—Ä–∞–∫—Ç–µ—Ä —Ä–∞–≤–Ω–æ–º–µ—Ä–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è:

1. __–ü–∞—Ä–∞–º–µ—Ç—Ä a (–Ω–∏–∂–Ω—è—è –≥—Ä–∞–Ω–∏—Ü–∞):__

- –ò–∑–º–µ–Ω–µ–Ω–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ a –ø—Ä–∏ —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω–æ–π —à–∏—Ä–∏–Ω–µ –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞ $(b-a)$ –ø—Ä–∏–≤–æ–¥–∏—Ç –∫
–≥–æ—Ä–∏–∑–æ–Ω—Ç–∞–ª—å–Ω–æ–º—É —Å–¥–≤–∏–≥—É –≥—Ä–∞—Ñ–∏–∫–∞ —Ñ—É–Ω–∫—Ü–∏–∏ –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –∏ —Ñ—É–Ω–∫—Ü–∏–∏
—Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –≤–¥–æ–ª—å –æ—Å–∏ $x$, –Ω–µ –º–µ–Ω—è—è –∏—Ö —Ñ–æ—Ä–º—ã.
- –ß–µ–º –±–æ–ª—å—à–µ –∑–Ω–∞—á–µ–Ω–∏–µ $a$, —Ç–µ–º –¥–∞–ª—å—à–µ –≤–ø—Ä–∞–≤–æ —Å–º–µ—â–∞–µ—Ç—Å—è –æ–±–ª–∞—Å—Ç—å 
–Ω–µ–Ω—É–ª–µ–≤—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π —Ñ—É–Ω–∫—Ü–∏–∏ –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏.
- –ù–∞ –≥—Ä–∞—Ñ–∏–∫–µ —Ñ—É–Ω–∫—Ü–∏–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –∏–∑–º–µ–Ω–µ–Ω–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ a —Å–º–µ—â–∞–µ—Ç 
–Ω–∞—á–∞–ª–æ –Ω–∞—Ä–∞—Å—Ç–∞–Ω–∏—è —Ñ—É–Ω–∫—Ü–∏–∏ (—Ç–æ—á–∫—É, –≥–¥–µ $F(x)$ –Ω–∞—á–∏–Ω–∞–µ—Ç –≤–æ–∑—Ä–∞—Å—Ç–∞—Ç—å
–æ—Ç 0).

2. __–ü–∞—Ä–∞–º–µ—Ç—Ä b (–≤–µ—Ä—Ö–Ω—è—è –≥—Ä–∞–Ω–∏—Ü–∞) –∏–ª–∏ —à–∏—Ä–∏–Ω–∞ –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞ $(b-a)$__:

- –ü—Ä–∏ —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω–æ–º –∑–Ω–∞—á–µ–Ω–∏–∏ $a$, —É–≤–µ–ª–∏—á–µ–Ω–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ $b$ (–∏–ª–∏ 
—à–∏—Ä–∏–Ω—ã –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞ $b-a$) –ø—Ä–∏–≤–æ–¥–∏—Ç –∫ —É–º–µ–Ω—å—à–µ–Ω–∏—é –≤—ã—Å–æ—Ç—ã –ø—Ä—è–º–æ—É–≥–æ–ª—å–Ω–∏–∫–∞
—Ñ—É–Ω–∫—Ü–∏–∏ –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏, —Ç–∞–∫ –∫–∞–∫ –æ–±—â–∞—è –ø–ª–æ—â–∞–¥—å –ø–æ–¥ —Ñ—É–Ω–∫—Ü–∏–µ–π 
–ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ –¥–æ–ª–∂–Ω–∞ –æ—Å—Ç–∞–≤–∞—Ç—å—Å—è —Ä–∞–≤–Ω–æ–π $1$.
- –ß–µ–º –±–æ–ª—å—à–µ —à–∏—Ä–∏–Ω–∞ –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞ $(b-a)$, —Ç–µ–º –Ω–∏–∂–µ –∑–Ω–∞—á–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏ 
–ø–ª–æ—Ç–Ω–æ—Å—Ç–∏: $f(x) = \frac {1} {(b-a)}$.
- –£–≤–µ–ª–∏—á–µ–Ω–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ $b$ —Ç–∞–∫–∂–µ –ø—Ä–∏–≤–æ–¥–∏—Ç –∫ –±–æ–ª–µ–µ –ø–æ–ª–æ–≥–æ–º—É –Ω–∞–∫–ª–æ–Ω—É 
—Ñ—É–Ω–∫—Ü–∏–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –Ω–∞ —É—á–∞—Å—Ç–∫–µ –æ—Ç $a$ –¥–æ $b$, —Ç–∞–∫ –∫–∞–∫ 
$F(x) = \frac {(x-a)} {(b-a)}$ –ø—Ä–∏ $a \leq x \leq b$.

–¢–∞–∫–∏–º –æ–±—Ä–∞–∑–æ–º, –ø–∞—Ä–∞–º–µ—Ç—Ä $a$ –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç –ø–æ–ª–æ–∂–µ–Ω–∏–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è 
–Ω–∞ —á–∏—Å–ª–æ–≤–æ–π –æ—Å–∏, –∞ —Ä–∞–∑–Ω–æ—Å—Ç—å $b-a$ –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç —à–∏—Ä–∏–Ω—É –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞, 
–Ω–∞ –∫–æ—Ç–æ—Ä–æ–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∞ —Å–ª—É—á–∞–π–Ω–∞—è –≤–µ–ª–∏—á–∏–Ω–∞, —á—Ç–æ –≤–ª–∏—è–µ—Ç –Ω–∞ –≤—ã—Å–æ—Ç—É 
—Ñ—É–Ω–∫—Ü–∏–∏ –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ –∏ –Ω–∞–∫–ª–æ–Ω —Ñ—É–Ω–∫—Ü–∏–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è.
"""))
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
from scipy.stats import poisson
from IPython.display import display, Markdown, HTML

display(HTML())

display(HTML('<h1>–ó–∞–¥–∞—á–∞ 2: –ü—É–∞—Å—Å–æ–Ω–æ–≤—Å–∫–∞—è —Å–ª—É—á–∞–π–Ω–∞—è –≤–µ–ª–∏—á–∏–Ω–∞</h1>'))
display(HTML('<h2>1. –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∫–ª–∞—Å—Å–∞ –∏ –∑–∞–∫–æ–Ω–∞ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è</h2>'))

display(Markdown(r"""
- –ö–ª–∞—Å—Å —Å–ª—É—á–∞–π–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω—ã: –î–∏—Å–∫—Ä–µ—Ç–Ω–∞—è
- –ó–∞–∫–æ–Ω —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è: –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ü—É–∞—Å—Å–æ–Ω–∞
- –ü–∞—Ä–∞–º–µ—Ç—Ä –∑–∞–∫–æ–Ω–∞ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è: $Œª = 0.5$
"""))

display(HTML('<h2>2. –°–æ–∑–¥–∞–Ω–∏–µ —Å–ª—É—á–∞–π–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω—ã —Å –¥–∞–Ω–Ω—ã–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ–º</h2>'))

# –ü–∞—Ä–∞–º–µ—Ç—Ä —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ü—É–∞—Å—Å–æ–Ω–∞
lambda_param = 0.5
# –°–æ–∑–¥–∞–Ω–∏–µ —Å–ª—É—á–∞–π–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω—ã —Å —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ–º –ü—É–∞—Å—Å–æ–Ω–∞
X_poisson = poisson(mu=lambda_param)
# –ì–µ–Ω–µ—Ä–∞—Ü–∏—è –≤—ã–±–æ—Ä–∫–∏ –¥–ª—è –¥–µ–º–æ–Ω—Å—Ç—Ä–∞—Ü–∏–∏
sample_poisson = X_poisson.rvs(size=1000)
print(f"–ü–µ—Ä–≤—ã–µ 10 —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –≤—ã–±–æ—Ä–∫–∏ –∏–∑ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ü—É–∞—Å—Å–æ–Ω–∞ —Å Œª = {lambda_param}:")
print(sample_poisson[:10])

display(HTML('<h2>3. –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–æ–≤</h2>'))
display(HTML('<h3>3.1 –ì—Ä–∞—Ñ–∏–∫ —Ñ—É–Ω–∫—Ü–∏–∏ –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π</h3>'))

# –î–∏–∞–ø–∞–∑–æ–Ω –∑–Ω–∞—á–µ–Ω–∏–π –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–∞ (–¥–æ k=6 –æ—Ö–≤–∞—Ç—ã–≤–∞–µ—Ç –ø–æ—á—Ç–∏ –≤—Å–µ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏)
k_values = np.arange(0, 7)
pmf_values = X_poisson.pmf(k_values)

plt.figure(figsize=(10, 6))
plt.bar(k_values, pmf_values, alpha=0.7, color='skyblue', edgecolor='black')
plt.plot(k_values, pmf_values, 'ro-', ms=8)
plt.title('–ú–Ω–æ–≥–æ—É–≥–æ–ª—å–Ω–∏–∫ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ü—É–∞—Å—Å–æ–Ω–∞ —Å Œª = 0.5')
plt.xlabel('k')
plt.ylabel('P(X = k)')
plt.grid(True, alpha=0.3)
plt.xticks(k_values)
plt.show()

# –í—ã–≤–æ–¥ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π –≤ —Ç–µ–∫—Å—Ç–æ–≤–æ–º –≤–∏–¥–µ
print("–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ P(X = k) –¥–ª—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ü—É–∞—Å—Å–æ–Ω–∞ —Å Œª = 0.5:")
for k, p in zip(k_values, pmf_values):
    print(f"P(X = {k}) = {p:.6f}")

display(HTML('<h3>3.2 –ì—Ä–∞—Ñ–∏–∫ —Ñ—É–Ω–∫—Ü–∏–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è</h3>'))

# –î–∏–∞–ø–∞–∑–æ–Ω –∑–Ω–∞—á–µ–Ω–∏–π –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–∞ —Ñ—É–Ω–∫—Ü–∏–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è
k_values_cdf = np.arange(0, 7)
cdf_values = X_poisson.cdf(k_values_cdf)

plt.figure(figsize=(10, 6))
plt.step(k_values_cdf, cdf_values, 'b-', where='post', linewidth=2)
plt.scatter(k_values_cdf, cdf_values, color='red', s=50, zorder=3)
plt.title('–§—É–Ω–∫—Ü–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ü—É–∞—Å—Å–æ–Ω–∞ —Å Œª = 0.5')
plt.xlabel('x')
plt.ylabel('F(x) = P(X ‚â§ x)')
plt.grid(True, alpha=0.3)
plt.ylim(0, 1.05)
plt.xticks(k_values_cdf)
plt.show()

# –í—ã–≤–æ–¥ –∑–Ω–∞—á–µ–Ω–∏–π —Ñ—É–Ω–∫—Ü–∏–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –≤ —Ç–µ–∫—Å—Ç–æ–≤–æ–º –≤–∏–¥–µ
print("–ó–Ω–∞—á–µ–Ω–∏—è —Ñ—É–Ω–∫—Ü–∏–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è F(x) –¥–ª—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ü—É–∞—Å—Å–æ–Ω–∞ —Å Œª = 0.5:")
for k, f in zip(k_values_cdf, cdf_values):
    print(f"F({k}) = P(X ‚â§ {k}) = {f:.6f}")

display(HTML('<h2>4. –í—ã—á–∏—Å–ª–µ–Ω–∏–µ —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫</h2>'))
display(HTML('<h3>4.1 –ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ, –¥–∏—Å–ø–µ—Ä—Å–∏—è –∏ —Å—Ä–µ–¥–Ω–µ–µ –∫–≤–∞–¥—Ä–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ</h3>'))

# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–≥–æ –æ–∂–∏–¥–∞–Ω–∏—è
mean = X_poisson.mean()

# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –¥–∏—Å–ø–µ—Ä—Å–∏–∏
variance = X_poisson.var()

# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ —Å—Ä–µ–¥–Ω–µ–≥–æ –∫–≤–∞–¥—Ä–∞—Ç–∏—á–µ—Å–∫–æ–≥–æ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è
std_deviation = X_poisson.std()

print(f"–ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ: {mean}")
print(f"–î–∏—Å–ø–µ—Ä—Å–∏—è: {variance}")
print(f"–°—Ä–µ–¥–Ω–µ–µ –∫–≤–∞–¥—Ä–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ: {std_deviation}")

display(HTML('<h3>4.2 –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –º–µ–¥–∏–∞–Ω—ã –∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ P(3 < X < 7)</h3>'))

# –ö–≤–∞–Ω—Ç–∏–ª—å –ø–æ—Ä—è–¥–∫–∞ 0.15
quantile_0_15 = X_poisson.ppf(0.15)

# –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å P(0 ‚â§ X < 2)
prob_0_to_less_than_2 = X_poisson.cdf(1) - X_poisson.cdf(-1)  # P(X < 2) - P(X < 0)

print(f"–ö–≤–∞–Ω—Ç–∏–ª—å –ø–æ—Ä—è–¥–∫–∞ 0.15: {quantile_0_15}")
print(f"–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å P(0 ‚â§ X < 2): {prob_0_to_less_than_2:.6f}")

display(HTML('<h3>4.3 –ò–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è –ø–æ–ª—É—á–µ–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π</h3>'))

display(Markdown(r"""
–î–ª—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ü—É–∞—Å—Å–æ–Ω–∞ —Å –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–º $Œª = 0.5$:

- __–ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ $(Œª)$__: $Œª = 0.5$ - —ç—Ç–æ —Å—Ä–µ–¥–Ω–µ–µ 
–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ "—É—Å–ø–µ—Ö–æ–≤" –∏–ª–∏ "—Å–æ–±—ã—Ç–∏–π" –≤ —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω–æ–º –∏–Ω—Ç–µ—Ä–≤–∞–ª–µ.
–í —Å—Ä–µ–¥–Ω–µ–º —Å–ª—É—á–∞–π–Ω–∞—è –≤–µ–ª–∏—á–∏–Ω–∞ $X$ –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –∑–Ω–∞—á–µ–Ω–∏–µ $0.5$.

- __–î–∏—Å–ø–µ—Ä—Å–∏—è (—Ç–∞–∫–∂–µ —Ä–∞–≤–Ω–∞ $Œª$)__: $Œª = 0.5$ –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç —Ä–∞–∑–±—Ä–æ—Å
–∑–Ω–∞—á–µ–Ω–∏–π –≤–æ–∫—Ä—É–≥ –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–≥–æ –æ–∂–∏–¥–∞–Ω–∏—è. –ú–∞–ª–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –¥–∏—Å–ø–µ—Ä—Å–∏–∏
–æ–∑–Ω–∞—á–∞–µ—Ç, —á—Ç–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å–∫–æ–Ω—Ü–µ–Ω—Ç—Ä–∏—Ä–æ–≤–∞–Ω–æ –≤–±–ª–∏–∑–∏ —Å—Ä–µ–¥–Ω–µ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è.

- __–°—Ä–µ–¥–Ω–µ–µ –∫–≤–∞–¥—Ä–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ__: $\sqrt{Œª} = \sqrt{0.5} ‚âà 0.7071$.
–ü–æ–∫–∞–∑—ã–≤–∞–µ—Ç, –Ω–∞—Å–∫–æ–ª—å–∫–æ –≤ —Å—Ä–µ–¥–Ω–µ–º –æ—Ç–∫–ª–æ–Ω—è—é—Ç—Å—è –∑–Ω–∞—á–µ–Ω–∏—è —Å–ª—É—á–∞–π–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω—ã –æ—Ç 
–º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–≥–æ –æ–∂–∏–¥–∞–Ω–∏—è.
- __–ö–≤–∞–Ω—Ç–∏–ª—å –ø–æ—Ä—è–¥–∫–∞ $0.15$__: –≠—Ç–æ –∑–Ω–∞—á–µ–Ω–∏–µ $x$, –ø—Ä–∏ –∫–æ—Ç–æ—Ä–æ–º $P(X \leq x) = 0.15$. 
–î–ª—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ü—É–∞—Å—Å–æ–Ω–∞ —Å $Œª = 0.5$ —ç—Ç–æ $x = 0$, —á—Ç–æ –æ–∑–Ω–∞—á–∞–µ—Ç, 
—á—Ç–æ —Å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å—é $0.15$ –∏–ª–∏ –º–µ–Ω—å—à–µ —Å–ª—É—á–∞–π–Ω–∞—è –≤–µ–ª–∏—á–∏–Ω–∞ –ø—Ä–∏–º–µ—Ç 
–∑–Ω–∞—á–µ–Ω–∏–µ –Ω–µ –±–æ–ª—å—à–µ 0.
- –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å $P(0 \leq X < 2)$: –≠—Ç–æ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ —Å–ª—É—á–∞–π–Ω–∞—è 
–≤–µ–ª–∏—á–∏–Ω–∞ –ø—Ä–∏–º–µ—Ç –∑–Ω–∞—á–µ–Ω–∏–µ 0 –∏–ª–∏ 1. –ó–Ω–∞—á–µ–Ω–∏–µ –æ–∫–æ–ª–æ $0.91$ –æ–∑–Ω–∞—á–∞–µ—Ç, 
—á—Ç–æ —Å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å—é –æ–∫–æ–ª–æ 91% —Å–ª—É—á–∞–π–Ω–∞—è –≤–µ–ª–∏—á–∏–Ω–∞ –ø—Ä–∏–º–µ—Ç –∑–Ω–∞—á–µ–Ω–∏–µ –ª–∏–±–æ 
0, –ª–∏–±–æ 1, —á—Ç–æ –æ–∂–∏–¥–∞–µ–º–æ –¥–ª—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ü—É–∞—Å—Å–æ–Ω–∞ —Å –º–∞–ª—ã–º –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–º
$Œª$.
"""))

display(HTML('<h2>5. –í–ª–∏—è–Ω–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –Ω–∞ —Ö–∞—Ä–∞–∫—Ç–µ—Ä –∑–∞–∫–æ–Ω–∞ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è</h2>'))
display(HTML('<h3>5.1 –í–ª–∏—è–Ω–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –Ω–∞ —Ñ—É–Ω–∫—Ü–∏—é –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏</h3>'))


# –ü–∞—Ä–∞–º–µ—Ç—Ä—ã —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ü—É–∞—Å—Å–æ–Ω–∞ –¥–ª—è —Å—Ä–∞–≤–Ω–µ–Ω–∏—è
lambda_values = [0.5, 2, 5]
k_range = np.arange(0, 15)

plt.figure(figsize=(12, 7))

for lambda_param in lambda_values:
    poisson_dist = poisson(mu=lambda_param)
    pmf_values = poisson_dist.pmf(k_range)
    
    plt.plot(k_range, pmf_values, 'o-', label=f'Œª = {lambda_param}', linewidth=2, markersize=6)

plt.title('–ú–Ω–æ–≥–æ—É–≥–æ–ª—å–Ω–∏–∫–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ü—É–∞—Å—Å–æ–Ω–∞ —Å —Ä–∞–∑–Ω—ã–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ Œª')
plt.xlabel('k')
plt.ylabel('P(X = k)')
plt.grid(True, alpha=0.3)
plt.legend()
plt.xticks(np.arange(0, 15))
plt.show()

display(HTML('<h3>5.2 –í–ª–∏—è–Ω–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –Ω–∞ —Ñ—É–Ω–∫—Ü–∏—é —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è</h3>'))

plt.figure(figsize=(12, 7))

for lambda_param in lambda_values:
    poisson_dist = poisson(mu=lambda_param)
    cdf_values = poisson_dist.cdf(k_range)
    
    plt.step(k_range, cdf_values, '-', where='post', linewidth=2, label=f'Œª = {lambda_param}')
    plt.scatter(k_range, cdf_values, s=40, alpha=0.6)

plt.title('–§—É–Ω–∫—Ü–∏–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ü—É–∞—Å—Å–æ–Ω–∞ —Å —Ä–∞–∑–Ω—ã–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ Œª')
plt.xlabel('x')
plt.ylabel('F(x) = P(X ‚â§ x)')
plt.grid(True, alpha=0.3)
plt.legend()
plt.xticks(np.arange(0, 15, 2))
plt.ylim(0, 1.05)
plt.show()

display(HTML('<h3>5.3 –í—ã–≤–æ–¥—ã –æ –≤–ª–∏—è–Ω–∏–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –Ω–∞ —Ö–∞—Ä–∞–∫—Ç–µ—Ä –∑–∞–∫–æ–Ω–∞ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è</h3>'))

display(Markdown(r"""
–ü–∞—Ä–∞–º–µ—Ç—Ä $Œª$ –≤ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–∏ –ü—É–∞—Å—Å–æ–Ω–∞ —Å—É—â–µ—Å—Ç–≤–µ–Ω–Ω–æ –≤–ª–∏—è–µ—Ç –Ω–∞ —Ö–∞—Ä–∞–∫—Ç–µ—Ä —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è:

1. __–ü—Ä–∏ –º–∞–ª—ã—Ö –∑–Ω–∞—á–µ–Ω–∏—è—Ö $Œª (0.5)$__:

- –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å–∏–ª—å–Ω–æ –∞—Å–∏–º–º–µ—Ç—Ä–∏—á–Ω–æ, –º–∞–∫—Å–∏–º–∞–ª—å–Ω–∞—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å 
–Ω–∞–±–ª—é–¥–∞–µ—Ç—Å—è –≤ –Ω—É–ª–µ $(P(X=0) ‚âà 0.607)$
- –ó–Ω–∞—á–µ–Ω–∏—è –±—ã—Å—Ç—Ä–æ —É–±—ã–≤–∞—é—Ç, –ø–æ—á—Ç–∏ –≤—Å—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Å–æ—Å—Ä–µ–¥–æ—Ç–æ—á–µ–Ω–∞ 
–Ω–∞ –∑–Ω–∞—á–µ–Ω–∏—è—Ö $0, 1, 2$
- –§—É–Ω–∫—Ü–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –±—ã—Å—Ç—Ä–æ —Ä–∞—Å—Ç–µ—Ç –∏ –¥–æ—Å—Ç–∏–≥–∞–µ—Ç –∑–Ω–∞—á–µ–Ω–∏–π, –±–ª–∏–∑–∫–∏—Ö 
–∫ $1$, —É–∂–µ –ø—Ä–∏ $x ‚âà 3$

2. __–ü—Ä–∏ —Å—Ä–µ–¥–Ω–∏—Ö –∑–Ω–∞—á–µ–Ω–∏—è—Ö $Œª (2)$__:

- –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –≤—Å–µ –µ—â–µ –∞—Å–∏–º–º–µ—Ç—Ä–∏—á–Ω–æ, –Ω–æ –º–∞–∫—Å–∏–º—É–º –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π 
—Å–º–µ—â–∞–µ—Ç—Å—è –∫ –∑–Ω–∞—á–µ–Ω–∏—é 1-2
- –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∞ –º–µ–∂–¥—É –±–æ–ª—å—à–∏–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –∑–Ω–∞—á–µ–Ω–∏–π
–§—É–Ω–∫—Ü–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Ä–∞—Å—Ç–µ—Ç –±–æ–ª–µ–µ –ø–ª–∞–≤–Ω–æ

3. __–ü—Ä–∏ –±–æ–ª—å—à–∏—Ö –∑–Ω–∞—á–µ–Ω–∏—è—Ö $Œª (5)$__:

- –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å—Ç–∞–Ω–æ–≤–∏—Ç—Å—è –±–æ–ª–µ–µ —Å–∏–º–º–µ—Ç—Ä–∏—á–Ω—ã–º, –Ω–∞–ø–æ–º–∏–Ω–∞—è –Ω–æ—Ä–º–∞–ª—å–Ω–æ–µ
- –ú–∞–∫—Å–∏–º—É–º –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π –ø—Ä–∏—Ö–æ–¥–∏—Ç—Å—è –Ω–∞ –∑–Ω–∞—á–µ–Ω–∏—è, –±–ª–∏–∑–∫–∏–µ –∫ Œª (4-5)
- –§—É–Ω–∫—Ü–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –∏–º–µ–µ—Ç —Ö–∞—Ä–∞–∫—Ç–µ—Ä–Ω—ã–π S-–æ–±—Ä–∞–∑–Ω—ã–π –≤–∏–¥
- –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω—ã –º–µ–∂–¥—É –µ—â–µ –±–æ–ª—å—à–∏–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ–º –∑–Ω–∞—á–µ–Ω–∏–π (–¥–æ 10-12)

4. __–° —É–≤–µ–ª–∏—á–µ–Ω–∏–µ–º –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ $Œª$__:

- –£–≤–µ–ª–∏—á–∏–≤–∞–µ—Ç—Å—è –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ ($E[X] = Œª$)
- –£–≤–µ–ª–∏—á–∏–≤–∞–µ—Ç—Å—è –¥–∏—Å–ø–µ—Ä—Å–∏—è –∏ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ ($Var[X] = Œª$)
- –§–æ—Ä–º–∞ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ø—Ä–∏ –±–æ–ª—å—à–∏—Ö Œª –Ω–∞—á–∏–Ω–∞–µ—Ç –Ω–∞–ø–æ–º–∏–Ω–∞—Ç—å –Ω–æ—Ä–º–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ, —á—Ç–æ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É–µ—Ç —Ü–µ–Ω—Ç—Ä–∞–ª—å–Ω–æ–π –ø—Ä–µ–¥–µ–ª—å–Ω–æ–π —Ç–µ–æ—Ä–µ–º–µ
- –ú–Ω–æ–≥–æ—É–≥–æ–ª—å–Ω–∏–∫ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Å—Ç–∞–Ω–æ–≤–∏—Ç—Å—è –±–æ–ª–µ–µ "—Ä–∞—Å—Ç—è–Ω—É—Ç—ã–º" –ø–æ –æ—Å–∏ X
- –§—É–Ω–∫—Ü–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è F(x) –¥–æ—Å—Ç–∏–≥–∞–µ—Ç –∑–Ω–∞—á–µ–Ω–∏–π, –±–ª–∏–∑–∫–∏—Ö –∫ 1, –ø—Ä–∏ –≤—Å–µ –±–æ–ª—å—à–∏—Ö –∑–Ω–∞—á–µ–Ω–∏—è—Ö x

–î–∞–Ω–Ω—ã–π –ø–∞—Ä–∞–º–µ—Ç—Ä $Œª$ –ø–æ–ª–Ω–æ—Å—Ç—å—é –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç —Ö–∞—Ä–∞–∫—Ç–µ—Ä —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ü—É–∞—Å—Å–æ–Ω–∞, –ø–æ—Å–∫–æ–ª—å–∫—É —ç—Ç–æ –µ–¥–∏–Ω—Å—Ç–≤–µ–Ω–Ω—ã–π –ø–∞—Ä–∞–º–µ—Ç—Ä —ç—Ç–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è, –∑–∞–¥–∞—é—â–∏–π –∏ –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ, –∏ –¥–∏—Å–ø–µ—Ä—Å–∏—é."""))
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
from scipy.stats import binom
import seaborn as sns
from IPython.display import display, Markdown, HTML

display(HTML('<h1>–ó–∞–¥–∞—á–∞ 3: –ë–∏–Ω–æ–º–∏–∞–ª—å–Ω–∞—è —Å–ª—É—á–∞–π–Ω–∞—è –≤–µ–ª–∏—á–∏–Ω–∞</h1>'))
display(HTML('<h2>1. –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∫–ª–∞—Å—Å–∞ –∏ –∑–∞–∫–æ–Ω–∞ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è</h2>'))

display(Markdown(r"""
- –ö–ª–∞—Å—Å —Å–ª—É—á–∞–π–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω—ã: –î–∏—Å–∫—Ä–µ—Ç–Ω–∞—è
- –ó–∞–∫–æ–Ω —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è: –ë–∏–Ω–æ–º–∏–Ω–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ
- –ü–∞—Ä–∞–º–µ—Ç—Ä –∑–∞–∫–æ–Ω–∞ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è: 
$$\displaylines {
\begin{flalign}
&n = 8 \text{(–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —ç–ª–µ–º–µ–Ω—Ç–æ–≤)} \\
&p = 0.9 \text{(–≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ —ç–ª–µ–º–µ–Ω—Ç –Ω–µ –æ—Ç–∫–∞–∂–µ—Ç)}
\end{flalign}
}
$$
"""))

n = 8
p = 0.9

display(HTML('<h2>2. –°–æ–∑–¥–∞–Ω–∏–µ —Å–ª—É—á–∞–π–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω—ã —Å –¥–∞–Ω–Ω—ã–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ–º</h2>'))

# –°–æ–∑–¥–∞–Ω–∏–µ —Å–ª—É—á–∞–π–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω—ã —Å –±–∏–Ω–æ–º–∏–∞–ª—å–Ω—ã–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ–º
X_binom = binom(n, p)

# –ì–µ–Ω–µ—Ä–∞—Ü–∏—è –≤—ã–±–æ—Ä–∫–∏
sample_binom = X_binom.rvs(size=1000)

print(f"–ü–µ—Ä–≤—ã–µ 10 —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –≤—ã–±–æ—Ä–∫–∏ –∏–∑ –±–∏–Ω–æ–º–∏–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è B({n}, {p}):")
print(sample_binom[:10])

display(HTML('<h2>3. –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–æ–≤</h2>'))
display(HTML('<h3>3.1 –ú–Ω–æ–≥–æ—É–≥–æ–ª—å–Ω–∏–∫ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è</h3>'))

# –ó–Ω–∞—á–µ–Ω–∏—è X (–æ—Ç 0 –¥–æ n)
x_values = np.arange(0, n + 1)

# –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ P(X = k)
probabilities = X_binom.pmf(x_values)

# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –º–Ω–æ–≥–æ—É–≥–æ–ª—å–Ω–∏–∫–∞ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è
plt.figure(figsize=(10, 6))
plt.plot(x_values, probabilities, 'bo-', linewidth=2, markersize=8)
plt.title('–ú–Ω–æ–≥–æ—É–≥–æ–ª—å–Ω–∏–∫ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è B(8, 0.9)', fontsize=14)
plt.xlabel('k (—á–∏—Å–ª–æ –Ω–µ –æ—Ç–∫–∞–∑–∞–≤—à–∏—Ö —ç–ª–µ–º–µ–Ω—Ç–æ–≤)', fontsize=12)
plt.ylabel('–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å P(X = k)', fontsize=12)
plt.grid(True)
plt.xticks(x_values)
plt.tight_layout()
plt.show()

display(HTML('<h3>3.2 –§—É–Ω–∫—Ü–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è</h3>'))

# –ó–Ω–∞—á–µ–Ω–∏—è X –¥–ª—è –±–æ–ª–µ–µ –≥–ª–∞–¥–∫–æ–≥–æ –≥—Ä–∞—Ñ–∏–∫–∞
x_values_continuous = np.linspace(-0.5, n + 0.5, 1000)

# –§—É–Ω–∫—Ü–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è F(x)
cdf_values = X_binom.cdf(x_values_continuous)

# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–∞ —Ñ—É–Ω–∫—Ü–∏–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è
plt.figure(figsize=(10, 6))
plt.plot(x_values_continuous, cdf_values, 'r-', linewidth=2)
plt.title('–§—É–Ω–∫—Ü–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è B(8, 0.9)', fontsize=14)
plt.xlabel('x (—á–∏—Å–ª–æ –Ω–µ –æ—Ç–∫–∞–∑–∞–≤—à–∏—Ö —ç–ª–µ–º–µ–Ω—Ç–æ–≤)', fontsize=12)
plt.ylabel('F(x) = P(X ‚â§ x)', fontsize=12)
plt.grid(True)
plt.yticks(np.arange(0, 1.1, 0.1))
plt.xticks(np.arange(0, n + 1))
plt.tight_layout()
plt.show()

display(HTML('<h2>4. –ß–∏—Å–ª–æ–≤—ã–µ —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫–∏</h2>'))
display(HTML('<h3>4.1 –ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ, –¥–∏—Å–ø–µ—Ä—Å–∏—è –∏ –°–ö–û</h3>'))

# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–≥–æ –æ–∂–∏–¥–∞–Ω–∏—è
expected_value = X_binom.mean()
print(f"–ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ: {expected_value}")

# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –¥–∏—Å–ø–µ—Ä—Å–∏–∏
variance = X_binom.var()
print(f"–î–∏—Å–ø–µ—Ä—Å–∏—è: {variance}")

# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–≥–æ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è
std_dev = X_binom.std()
print(f"–°—Ä–µ–¥–Ω–µ–µ –∫–≤–∞–¥—Ä–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ: {std_dev}")

display(HTML('<h3>4.2 –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –∫–≤–∞–Ω—Ç–∏–ª—è –∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏</h3>'))

# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ —Ç—Ä–µ—Ç—å–µ–≥–æ –∫–≤–∞—Ä—Ç–∏–ª—è (0.75 –∫–≤–∞–Ω—Ç–∏–ª—å)
third_quartile = X_binom.ppf(0.75)
print(f"–¢—Ä–µ—Ç–∏–π –∫–≤–∞—Ä—Ç–∏–ª—å: {third_quartile}")

# –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å P(6 ‚â§ X ‚â§ 7)
prob_6_to_7 = X_binom.pmf(6) + X_binom.pmf(7)
print(f"–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å P(6 ‚â§ X ‚â§ 7): {prob_6_to_7}")

display(HTML('<h3>4.3 –ò–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤</h3>'))

display(Markdown(r"""
–ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ $E(X) = n \cdot p = 8 \cdot 0.9 = 7.2$, 
—á—Ç–æ –æ–∑–Ω–∞—á–∞–µ—Ç, —á—Ç–æ –≤ —Å—Ä–µ–¥–Ω–µ–º $7.2$ —ç–ª–µ–º–µ–Ω—Ç–∞ –∏–∑ 8 –Ω–µ –æ—Ç–∫–∞–∂—É—Ç –≤ –æ–¥–Ω–æ–º –æ–ø—ã—Ç–µ.

–î–∏—Å–ø–µ—Ä—Å–∏—è $Var(X) = n \cdot p \cdot (1-p) = 8 \cdot 0.9 \cdot 0.1 = 0.72$, —á—Ç–æ —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏–∑—É–µ—Ç
—Ä–∞–∑–±—Ä–æ—Å —á–∏—Å–ª–∞ –Ω–µ –æ—Ç–∫–∞–∑–∞–≤—à–∏—Ö —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–≥–æ –æ–∂–∏–¥–∞–Ω–∏—è.

–°—Ä–µ–¥–Ω–µ–µ –∫–≤–∞–¥—Ä–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ $œÉ = \sqrt {0.72} ‚âà 0.849$, —á—Ç–æ –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç, —á—Ç–æ
—Ç–∏–ø–∏—á–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ –æ—Ç —Å—Ä–µ–¥–Ω–µ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç –æ–∫–æ–ª–æ $0.849$ —ç–ª–µ–º–µ–Ω—Ç–æ–≤.

–¢—Ä–µ—Ç–∏–π –∫–≤–∞—Ä—Ç–∏–ª—å —Ä–∞–≤–µ–Ω 8, —á—Ç–æ –æ–∑–Ω–∞—á–∞–µ—Ç, —á—Ç–æ –≤ 75% —Å–ª—É—á–∞–µ–≤ —á–∏—Å–ª–æ –Ω–µ –æ—Ç–∫–∞–∑–∞–≤—à–∏—Ö 
—ç–ª–µ–º–µ–Ω—Ç–æ–≤ –Ω–µ –ø—Ä–µ–≤—ã—à–∞–µ—Ç 8, —Ç–æ –µ—Å—Ç—å —Å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å—é 0.75 –æ—Ç–∫–∞–∂—É—Ç –Ω–µ –±–æ–ª–µ–µ 8 —ç–ª–µ–º–µ–Ω—Ç–æ–≤.

–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å $P(6 \leq X \leq 7)$ –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç —à–∞–Ω—Å —Ç–æ–≥–æ, —á—Ç–æ —á–∏—Å–ª–æ –Ω–µ –æ—Ç–∫–∞–∑–∞–≤—à–∏—Ö —ç–ª–µ–º–µ–Ω—Ç–æ–≤
–±—É–¥–µ—Ç –æ—Ç 6 –¥–æ 7 –≤–∫–ª—é—á–∏—Ç–µ–ª—å–Ω–æ. –≠—Ç–æ –æ–∑–Ω–∞—á–∞–µ—Ç, —á—Ç–æ —Å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å—é –æ–∫–æ–ª–æ $0.66$ –∏–∑ 8 —ç–ª–µ–º–µ–Ω—Ç–æ–≤
–Ω–µ –æ—Ç–∫–∞–∂—É—Ç 6 –∏–ª–∏ 7.
"""))

display(HTML('<h2>5. –í–ª–∏—è–Ω–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –Ω–∞ —Ö–∞—Ä–∞–∫—Ç–µ—Ä —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è</h2>'))
display(HTML('<h3>5.1 –í–ª–∏—è–Ω–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ p –ø—Ä–∏ —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω–æ–º n</h3>'))

# –†–∞–∑–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ p –ø—Ä–∏ —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω–æ–º n=8
p_values = [0.3, 0.5, 0.9]
x_values = np.arange(0, n + 1)

# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –º–Ω–æ–≥–æ—É–≥–æ–ª—å–Ω–∏–∫–æ–≤ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –¥–ª—è —Ä–∞–∑–Ω—ã—Ö p
plt.figure(figsize=(12, 6))
for p_val in p_values:
    X_temp = binom(n, p_val)
    probabilities = X_temp.pmf(x_values)
    plt.plot(x_values, probabilities, 'o-', linewidth=2, markersize=6, label=f'p = {p_val}')

plt.title('–ú–Ω–æ–≥–æ—É–≥–æ–ª—å–Ω–∏–∫–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è B(8, p) –ø—Ä–∏ —Ä–∞–∑–Ω—ã—Ö p', fontsize=14)
plt.xlabel('k (—á–∏—Å–ª–æ –Ω–µ –æ—Ç–∫–∞–∑–∞–≤—à–∏—Ö —ç–ª–µ–º–µ–Ω—Ç–æ–≤)', fontsize=12)
plt.ylabel('–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å P(X = k)', fontsize=12)
plt.grid(True)
plt.xticks(x_values)
plt.legend()
plt.tight_layout()
plt.show()

# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–π —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –¥–ª—è —Ä–∞–∑–Ω—ã—Ö p
plt.figure(figsize=(12, 6))
x_values_continuous = np.linspace(-0.5, n + 0.5, 1000)
for p_val in p_values:
    X_temp = binom(n, p_val)
    cdf_values = X_temp.cdf(x_values_continuous)
    plt.plot(x_values_continuous, cdf_values, linewidth=2, label=f'p = {p_val}')

plt.title('–§—É–Ω–∫—Ü–∏–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è B(8, p) –ø—Ä–∏ —Ä–∞–∑–Ω—ã—Ö p', fontsize=14)
plt.xlabel('x (—á–∏—Å–ª–æ –Ω–µ –æ—Ç–∫–∞–∑–∞–≤—à–∏—Ö —ç–ª–µ–º–µ–Ω—Ç–æ–≤)', fontsize=12)
plt.ylabel('F(x) = P(X ‚â§ x)', fontsize=12)
plt.grid(True)
plt.yticks(np.arange(0, 1.1, 0.1))
plt.xticks(np.arange(0, n + 1))
plt.legend()
plt.tight_layout()
plt.show()

display(HTML('<h3>5.2 –í–ª–∏—è–Ω–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ n –ø—Ä–∏ —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω–æ–º p</h3>'))

# –†–∞–∑–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è —á–∏—Å–ª–∞ –∏—Å–ø—ã—Ç–∞–Ω–∏–π n –ø—Ä–∏ —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω–æ–º p=0.9
n_values = [4, 8, 16]
max_n = max(n_values)

# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –º–Ω–æ–≥–æ—É–≥–æ–ª—å–Ω–∏–∫–æ–≤ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –¥–ª—è —Ä–∞–∑–Ω—ã—Ö n
plt.figure(figsize=(14, 6))
for n_val in n_values:
    X_temp = binom(n_val, p)
    x_values = np.arange(0, n_val + 1)
    probabilities = X_temp.pmf(x_values)
    plt.plot(x_values, probabilities, 'o-', linewidth=2, markersize=6, label=f'n = {n_val}')

plt.title('–ú–Ω–æ–≥–æ—É–≥–æ–ª—å–Ω–∏–∫–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è B(n, 0.9) –ø—Ä–∏ —Ä–∞–∑–Ω—ã—Ö n', fontsize=14)
plt.xlabel('k (—á–∏—Å–ª–æ –Ω–µ –æ—Ç–∫–∞–∑–∞–≤—à–∏—Ö —ç–ª–µ–º–µ–Ω—Ç–æ–≤)', fontsize=12)
plt.ylabel('–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å P(X = k)', fontsize=12)
plt.grid(True)
plt.xticks(np.arange(0, max_n + 1, 2))
plt.legend()
plt.tight_layout()
plt.show()

# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–π —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –¥–ª—è —Ä–∞–∑–Ω—ã—Ö n
plt.figure(figsize=(14, 6))
for n_val in n_values:
    X_temp = binom(n_val, p)
    x_values_continuous = np.linspace(-0.5, n_val + 0.5, 1000)
    cdf_values = X_temp.cdf(x_values_continuous)
    plt.plot(x_values_continuous, cdf_values, linewidth=2, label=f'n = {n_val}')

plt.title('–§—É–Ω–∫—Ü–∏–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è B(n, 0.9) –ø—Ä–∏ —Ä–∞–∑–Ω—ã—Ö n', fontsize=14)
plt.xlabel('x (—á–∏—Å–ª–æ –Ω–µ –æ—Ç–∫–∞–∑–∞–≤—à–∏—Ö —ç–ª–µ–º–µ–Ω—Ç–æ–≤)', fontsize=12)
plt.ylabel('F(x) = P(X ‚â§ x)', fontsize=12)
plt.grid(True)
plt.yticks(np.arange(0, 1.1, 0.1))
plt.xticks(np.arange(0, max_n + 1, 2))
plt.legend()
plt.tight_layout()
plt.show()

display(HTML('<h3>5.3 –í—ã–≤–æ–¥—ã –æ –≤–ª–∏—è–Ω–∏–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –Ω–∞ —Ö–∞—Ä–∞–∫—Ç–µ—Ä —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è</h3>'))

display(Markdown(r"""
–ü—Ä–∏ –∞–Ω–∞–ª–∏–∑–µ –≤–ª–∏—è–Ω–∏—è –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –Ω–∞ –±–∏–Ω–æ–º–∏–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ B(n,p) –º–æ–∂–Ω–æ —Å–¥–µ–ª–∞—Ç—å —Å–ª–µ–¥—É—é—â–∏–µ –≤—ã–≤–æ–¥—ã:

1. __–í–ª–∏—è–Ω–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ $p$ (–ø—Ä–∏ —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω–æ–º $n$)__:

- –ü—Ä–∏ $p = 0.5$ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å–∏–º–º–µ—Ç—Ä–∏—á–Ω–æ –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ $n/2$
- –ü—Ä–∏ $p < 0.5$ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å–º–µ—â–∞–µ—Ç—Å—è –≤–ª–µ–≤–æ (–≤ —Å—Ç–æ—Ä–æ–Ω—É –º–µ–Ω—å—à–∏—Ö –∑–Ω–∞—á–µ–Ω–∏–π)
- –ü—Ä–∏ $p > 0.5$ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å–º–µ—â–∞–µ—Ç—Å—è –≤–ø—Ä–∞–≤–æ (–≤ —Å—Ç–æ—Ä–æ–Ω—É –±–æ–ª—å—à–∏—Ö –∑–Ω–∞—á–µ–Ω–∏–π)
- –ß–µ–º –±–ª–∏–∂–µ p –∫ 0 –∏–ª–∏ 1, —Ç–µ–º –±–æ–ª–µ–µ –∞—Å–∏–º–º–µ—Ç—Ä–∏—á–Ω—ã–º —Å—Ç–∞–Ω–æ–≤–∏—Ç—Å—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ
- –ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ —Å–º–µ—â–∞–µ—Ç—Å—è –≤–º–µ—Å—Ç–µ —Å —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ–º ($E(X) = n \cdot p$)
- –í–∏–¥ —Ñ—É–Ω–∫—Ü–∏–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Ç–∞–∫–∂–µ –º–µ–Ω—è–µ—Ç—Å—è: –ø—Ä–∏ —É–≤–µ–ª–∏—á–µ–Ω–∏–∏ $p$ –≥—Ä–∞—Ñ–∏–∫ $F(x)$ —Å—Ç–∞–Ω–æ–≤–∏—Ç—Å—è 
–±–æ–ª–µ–µ –ø–æ–ª–æ–≥–∏–º –≤ –ø—Ä–∞–≤–æ–π —á–∞—Å—Ç–∏

2. __–í–ª–∏—è–Ω–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ $n$ (–ø—Ä–∏ —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω–æ–º $p$)__:
- –° —É–≤–µ–ª–∏—á–µ–Ω–∏–µ–º $n$ —É–≤–µ–ª–∏—á–∏–≤–∞–µ—Ç—Å—è –¥–∏–∞–ø–∞–∑–æ–Ω –≤–æ–∑–º–æ–∂–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π $X$
- –ü—Ä–∏ –±–æ–ª—å—à–∏—Ö $n$ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å—Ç–∞–Ω–æ–≤–∏—Ç—Å—è –±–æ–ª–µ–µ "–≥–ª–∞–¥–∫–∏–º" –∏ –ø—Ä–∏–±–ª–∏–∂–∞–µ—Ç—Å—è –∫ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–º—É —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—é
- –ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ —É–≤–µ–ª–∏—á–∏–≤–∞–µ—Ç—Å—è –ø—Ä–æ–ø–æ—Ä—Ü–∏–æ–Ω–∞–ª—å–Ω–æ $n (E(X) = n \cdot p)$
- –î–∏—Å–ø–µ—Ä—Å–∏—è —Ç–∞–∫–∂–µ —É–≤–µ–ª–∏—á–∏–≤–∞–µ—Ç—Å—è —Å —Ä–æ—Å—Ç–æ–º $n$ ($Var(X) = n\cdot p \cdot (1-p)$)
- –§—É–Ω–∫—Ü–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Å —É–≤–µ–ª–∏—á–µ–Ω–∏–µ–º $n$ —Å—Ç–∞–Ω–æ–≤–∏—Ç—Å—è –±–æ–ª–µ–µ —Ä–∞—Å—Ç—è–Ω—É—Ç–æ–π –ø–æ –æ—Å–∏ $x$, –Ω–æ 
—Å–æ—Ö—Ä–∞–Ω—è–µ—Ç —Å–≤–æ—é S-–æ–±—Ä–∞–∑–Ω—É—é —Ñ–æ—Ä–º—É
"""))
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
from scipy.stats import expon
import seaborn as sns
from IPython.display import display, Markdown, HTML

display(HTML('<h1>–ó–∞–¥–∞—á–∞ 4: –ü–æ–∫–∞–∑–∞—Ç–µ–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –≤—Ä–µ–º–µ–Ω–∏ –≥–æ—Ä–µ–Ω–∏—è –ª–∞–º–ø–æ—á–∫–∏</h1>'))
display(HTML('<h2>1. –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∫–ª–∞—Å—Å–∞ –∏ –∑–∞–∫–æ–Ω–∞ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è</h2>'))

display(Markdown(r"""
- –ö–ª–∞—Å—Å —Å–ª—É—á–∞–π–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω—ã: –ù–µ–ø—Ä–µ—Ä—ã–≤–Ω–∞—è
- –ó–∞–∫–æ–Ω —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è: –ü–æ–∫–∞–∑–∞—Ç–µ–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ (—ç–∫—Å–ø–æ–Ω–µ–Ω—Ü–∏–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ)
- –ü–∞—Ä–∞–º–µ—Ç—Ä –∑–∞–∫–æ–Ω–∞ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è: $Œª = 1/1000 = 0.001$ (–∏–Ω—Ç–µ–Ω—Å–∏–≤–Ω–æ—Å—Ç—å –æ—Ç–∫–∞–∑–æ–≤)
"""))

display(HTML('<h2>2. –°–æ–∑–¥–∞–Ω–∏–µ —Å–ª—É—á–∞–π–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω—ã —Å –¥–∞–Ω–Ω—ã–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ–º</h2>'))

# –ü–∞—Ä–∞–º–µ—Ç—Ä –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è (Œª = 1/—Å—Ä–µ–¥–Ω–µ–µ –≤—Ä–µ–º—è)
lambda_param = 1/1000  # 1/—á–∞—Å–æ–≤
scale = 1/lambda_param  # scale = 1/Œª = 1000

# –°–æ–∑–¥–∞–Ω–∏–µ —Å–ª—É—á–∞–π–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω—ã —Å –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å–Ω—ã–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ–º
X_exp = expon(scale=scale)

# –ì–µ–Ω–µ—Ä–∞—Ü–∏—è –≤—ã–±–æ—Ä–∫–∏
sample_exp = X_exp.rvs(size=1000)

print(f"–ü–µ—Ä–≤—ã–µ 10 —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –≤—ã–±–æ—Ä–∫–∏ –∏–∑ –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Å Œª = {lambda_param}:")
print(sample_exp[:10])

display(HTML('<h2>3. –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–æ–≤</h2>'))
display(HTML('<h3>3.1 –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏ –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π</h3>'))

# –î–∏–∞–ø–∞–∑–æ–Ω –∑–Ω–∞—á–µ–Ω–∏–π –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è –≥—Ä–∞—Ñ–∏–∫–∞
x_values = np.linspace(0, 3000, 1000)
# –†–∞—Å—á–µ—Ç –∑–Ω–∞—á–µ–Ω–∏–π —Ñ—É–Ω–∫—Ü–∏–∏ –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏
pdf_values = X_exp.pdf(x_values)

# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–∞ —Ñ—É–Ω–∫—Ü–∏–∏ –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏
plt.figure(figsize=(10, 6))
plt.plot(x_values, pdf_values, 'r-', linewidth=2)
plt.title('–§—É–Ω–∫—Ü–∏—è –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è (Œª = 0.001)')
plt.xlabel('–í—Ä–µ–º—è –≥–æ—Ä–µ–Ω–∏—è –ª–∞–º–ø–æ—á–∫–∏ (—á–∞—Å—ã)')
plt.ylabel('–ü–ª–æ—Ç–Ω–æ—Å—Ç—å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏')
plt.grid(True)
plt.axvline(x=1000, color='green', linestyle='--', label='–°—Ä–µ–¥–Ω–µ–µ –≤—Ä–µ–º—è = 1000 —á–∞—Å–æ–≤')
plt.legend()
plt.show()

display(HTML('<h3>3.2 –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è</h3>'))

# –†–∞—Å—á–µ—Ç –∑–Ω–∞—á–µ–Ω–∏–π —Ñ—É–Ω–∫—Ü–∏–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è
cdf_values = X_exp.cdf(x_values)

# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–∞ —Ñ—É–Ω–∫—Ü–∏–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è
plt.figure(figsize=(10, 6))
plt.plot(x_values, cdf_values, 'b-', linewidth=2)
plt.title('–§—É–Ω–∫—Ü–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è (Œª = 0.001)')
plt.xlabel('–í—Ä–µ–º—è –≥–æ—Ä–µ–Ω–∏—è –ª–∞–º–ø–æ—á–∫–∏ (—á–∞—Å—ã)')
plt.ylabel('–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å')
plt.grid(True)
plt.axhline(y=0.65, color='green', linestyle='--', label='65%-–Ω—ã–π –∫–≤–∞–Ω—Ç–∏–ª—å')
plt.axvline(x=X_exp.ppf(0.65), color='red', linestyle='--', label=f'x = {X_exp.ppf(0.65):.2f} —á–∞—Å–æ–≤')
plt.legend()
plt.show()

display(HTML('<h2>4. –í—ã—á–∏—Å–ª–µ–Ω–∏–µ —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫</h2>'))
display(HTML('<h3>4.1 –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–≥–æ –æ–∂–∏–¥–∞–Ω–∏—è, –¥–∏—Å–ø–µ—Ä—Å–∏–∏ –∏ –°–ö–û</h3>'))

# –ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ
mean_exp = X_exp.mean()
# –î–∏—Å–ø–µ—Ä—Å–∏—è
var_exp = X_exp.var()
# –°—Ä–µ–¥–Ω–µ–µ –∫–≤–∞–¥—Ä–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ
std_exp = X_exp.std()

print(f"–ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ: {mean_exp} —á–∞—Å–æ–≤")
print(f"–î–∏—Å–ø–µ—Ä—Å–∏—è: {var_exp} —á–∞—Å–æ–≤¬≤")
print(f"–°—Ä–µ–¥–Ω–µ–µ –∫–≤–∞–¥—Ä–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ: {std_exp} —á–∞—Å–æ–≤")

display(HTML('<h3>4.2 –í—ã—á–∏—Å–ª–µ–Ω–∏–µ 65%-–Ω–æ–≥–æ –∫–≤–∞–Ω—Ç–∏–ª—è –∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏ –ø–æ–ø–∞–¥–∞–Ω–∏—è –≤ –∏–Ω—Ç–µ—Ä–≤–∞–ª</h3>'))

# 65%-–Ω—ã–π –∫–≤–∞–Ω—Ç–∏–ª—å
quantile_65 = X_exp.ppf(0.65)
print(f"65%-–Ω—ã–π –∫–≤–∞–Ω—Ç–∏–ª—å: {quantile_65:.2f} —á–∞—Å–æ–≤")

# –í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ —Å–ª—É—á–∞–π–Ω–æ –≤—ã–±—Ä–∞–Ω–Ω–∞—è –ª–∞–º–ø–æ—á–∫–∞ –±—É–¥–µ—Ç –≥–æ—Ä–µ—Ç—å –æ—Ç 950 –¥–æ 1050 —á–∞—Å–æ–≤
prob_interval = X_exp.cdf(1050) - X_exp.cdf(950)
print(f"–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –≥–æ—Ä–µ–Ω–∏—è –æ—Ç 950 –¥–æ 1050 —á–∞—Å–æ–≤: {prob_interval:.4f} ‚âà {prob_interval*100:.2f}%")

display(HTML('<h3>4.3 –ò–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è –ø–æ–ª—É—á–µ–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫</h3>'))

display(Markdown(r"""
–ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ (1000 —á–∞—Å–æ–≤) –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è–µ—Ç —Å–æ–±–æ–π —Å—Ä–µ–¥–Ω–µ–µ –≤—Ä–µ–º—è –≥–æ—Ä–µ–Ω–∏—è –ª–∞–º–ø–æ—á–∫–∏,
—á—Ç–æ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É–µ—Ç –∑–∞—è–≤–ª–µ–Ω–∏—é –ø—Ä–æ–∏–∑–≤–æ–¥–∏—Ç–µ–ª—è.

–î–∏—Å–ø–µ—Ä—Å–∏—è (1000000 $—á–∞—Å–æ–≤_2$) –∏ —Å—Ä–µ–¥–Ω–µ–µ –∫–≤–∞–¥—Ä–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ (1000 —á–∞—Å–æ–≤) –ø–æ–∫–∞–∑—ã–≤–∞—é—Ç
–∑–Ω–∞—á–∏—Ç–µ–ª—å–Ω—ã–π —Ä–∞–∑–±—Ä–æ—Å –≤—Ä–µ–º–µ–Ω–∏ –≥–æ—Ä–µ–Ω–∏—è –ª–∞–º–ø–æ—á–µ–∫, —á—Ç–æ —Ç–∏–ø–∏—á–Ω–æ –¥–ª—è –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è. 
–°—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ —Ä–∞–≤–Ω–æ –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–º—É –æ–∂–∏–¥–∞–Ω–∏—é, —á—Ç–æ —è–≤–ª—è–µ—Ç—Å—è —Ö–∞—Ä–∞–∫—Ç–µ—Ä–Ω—ã–º —Å–≤–æ–π—Å—Ç–≤–æ–º 
–ø–æ–∫–∞–∑–∞—Ç–µ–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è.

65%-–Ω—ã–π –∫–≤–∞–Ω—Ç–∏–ª—å —Ä–∞–≤–µ–Ω –ø—Ä–∏–º–µ—Ä–Ω–æ 1050 —á–∞—Å–æ–≤, —Ç–æ –µ—Å—Ç—å 65% –ª–∞–º–ø–æ—á–µ–∫ –±—É–¥—É—Ç –≥–æ—Ä–µ—Ç—å –º–µ–Ω–µ–µ 1050 —á–∞—Å–æ–≤,
–∞ 35% –ø—Ä–æ—Ä–∞–±–æ—Ç–∞—é—Ç –¥–æ–ª—å—à–µ.

–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –ª–∞–º–ø–æ—á–∫–∞ –ø—Ä–æ—Ä–∞–±–æ—Ç–∞–µ—Ç –æ—Ç 950 –¥–æ 1050 —á–∞—Å–æ–≤, —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç –ø—Ä–∏–º–µ—Ä–Ω–æ 0.0639 –∏–ª–∏ 6.39%.
–≠—Ç–æ –¥–æ–≤–æ–ª—å–Ω–æ –Ω–∏–∑–∫–∞—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å, —á—Ç–æ —É–∫–∞–∑—ã–≤–∞–µ—Ç –Ω–∞ –±–æ–ª—å—à–æ–π —Ä–∞–∑–±—Ä–æ—Å –≤—Ä–µ–º–µ–Ω–∏ –≥–æ—Ä–µ–Ω–∏—è –ª–∞–º–ø–æ—á–µ–∫ –ø—Ä–∏ 
–ø–æ–∫–∞–∑–∞—Ç–µ–ª—å–Ω–æ–º –∑–∞–∫–æ–Ω–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è.
"""))

display(HTML('<h2>5. –í–ª–∏—è–Ω–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –Ω–∞ —Ö–∞—Ä–∞–∫—Ç–µ—Ä –∑–∞–∫–æ–Ω–∞ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è</h2>'))
display(HTML('<h3>5.1 –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–æ–≤ –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Å —Ä–∞–∑–Ω—ã–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ Œª</h3>'))

plt.figure(figsize=(12, 6))

# –†–∞–∑–ª–∏—á–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è Œª
lambda_values = [0.0005, 0.001, 0.002]
colors = ['blue', 'red', 'green']
labels = ['Œª = 0.0005 (—Å—Ä–µ–¥–Ω–µ–µ = 2000 —á)', 'Œª = 0.001 (—Å—Ä–µ–¥–Ω–µ–µ = 1000 —á)', 'Œª = 0.002 (—Å—Ä–µ–¥–Ω–µ–µ = 500 —á)']

for i, lambda_val in enumerate(lambda_values):
    scale_val = 1/lambda_val
    x_exp = expon(scale=scale_val)
    pdf_vals = x_exp.pdf(x_values)
    
    plt.plot(x_values, pdf_vals, color=colors[i], linewidth=2, label=labels[i])

plt.title('–§—É–Ω–∫—Ü–∏–∏ –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Å —Ä–∞–∑–Ω—ã–º–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏ Œª')
plt.xlabel('–í—Ä–µ–º—è –≥–æ—Ä–µ–Ω–∏—è –ª–∞–º–ø–æ—á–∫–∏ (—á–∞—Å—ã)')
plt.ylabel('–ü–ª–æ—Ç–Ω–æ—Å—Ç—å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏')
plt.grid(True)
plt.legend()
plt.ylim(0, 0.004)
plt.xlim(0, 3000)
plt.show()

display(HTML('<h3>5.2 –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–æ–≤ —Ñ—É–Ω–∫—Ü–∏–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Å —Ä–∞–∑–Ω—ã–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ Œª</h3>'))

plt.figure(figsize=(12, 6))

for i, lambda_val in enumerate(lambda_values):
    scale_val = 1/lambda_val
    x_exp = expon(scale=scale_val)
    cdf_vals = x_exp.cdf(x_values)
    
    plt.plot(x_values, cdf_vals, color=colors[i], linewidth=2, label=labels[i])

plt.title('–§—É–Ω–∫—Ü–∏–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Å —Ä–∞–∑–Ω—ã–º–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏ Œª')
plt.xlabel('–í—Ä–µ–º—è –≥–æ—Ä–µ–Ω–∏—è –ª–∞–º–ø–æ—á–∫–∏ (—á–∞—Å—ã)')
plt.ylabel('–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å')
plt.grid(True)
plt.legend()
plt.ylim(0, 1.0)
plt.xlim(0, 3000)
plt.show()

display(HTML('<h3>5.3 –í—ã–≤–æ–¥—ã –æ –≤–ª–∏—è–Ω–∏–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –Ω–∞ —Ö–∞—Ä–∞–∫—Ç–µ—Ä –∑–∞–∫–æ–Ω–∞ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è</h3>'))

display(Markdown(r"""
–ü–∞—Ä–∞–º–µ—Ç—Ä $Œª$ (–∏–Ω—Ç–µ–Ω—Å–∏–≤–Ω–æ—Å—Ç—å –æ—Ç–∫–∞–∑–æ–≤) —Å—É—â–µ—Å—Ç–≤–µ–Ω–Ω–æ –≤–ª–∏—è–µ—Ç –Ω–∞ —Ö–∞—Ä–∞–∫—Ç–µ—Ä –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è:

1. __–í–ª–∏—è–Ω–∏–µ –Ω–∞ —Ñ—É–Ω–∫—Ü–∏—é –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏__:

- –ü—Ä–∏ —É–≤–µ–ª–∏—á–µ–Ω–∏–∏ $Œª$ (—É–º–µ–Ω—å—à–µ–Ω–∏–∏ —Å—Ä–µ–¥–Ω–µ–≥–æ –≤—Ä–µ–º–µ–Ω–∏ –≥–æ—Ä–µ–Ω–∏—è) –∫—Ä–∏–≤–∞—è –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ —Å—Ç–∞–Ω–æ–≤–∏—Ç—Å—è –±–æ–ª–µ–µ 
"–∫—Ä—É—Ç–æ–π" –≤–±–ª–∏–∑–∏ –Ω—É–ª—è –∏ –±—ã—Å—Ç—Ä–µ–µ —Å–ø–∞–¥–∞–µ—Ç.
- –ü—Ä–∏ —É–º–µ–Ω—å—à–µ–Ω–∏–∏ $Œª$ (—É–≤–µ–ª–∏—á–µ–Ω–∏–∏ —Å—Ä–µ–¥–Ω–µ–≥–æ –≤—Ä–µ–º–µ–Ω–∏ –≥–æ—Ä–µ–Ω–∏—è) –∫—Ä–∏–≤–∞—è —Å—Ç–∞–Ω–æ–≤–∏—Ç—Å—è –±–æ–ª–µ–µ –ø–æ–ª–æ–≥–æ–π –∏ 
—Ä–∞—Å—Ç—è–≥–∏–≤–∞–µ—Ç—Å—è –≤–¥–æ–ª—å –æ—Å–∏ –≤—Ä–µ–º–µ–Ω–∏.
- –ú–∞–∫—Å–∏–º—É–º –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ –≤—Å–µ–≥–¥–∞ –Ω–∞—Ö–æ–¥–∏—Ç—Å—è –≤ —Ç–æ—á–∫–µ $t = 0$, –Ω–æ –µ–≥–æ –≤—ã—Å–æ—Ç–∞ —Ä–∞–≤–Ω–∞ $Œª$, –ø–æ—ç—Ç–æ–º—É —É–≤–µ–ª–∏—á–µ–Ω–∏–µ $Œª$ 
–ø—Ä–∏–≤–æ–¥–∏—Ç –∫ —É–≤–µ–ª–∏—á–µ–Ω–∏—é –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è —Ñ—É–Ω–∫—Ü–∏–∏ –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏.

2. __–í–ª–∏—è–Ω–∏–µ –Ω–∞ —Ñ—É–Ω–∫—Ü–∏—é —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è__:
- –ü—Ä–∏ —É–≤–µ–ª–∏—á–µ–Ω–∏–∏ $Œª$ —Ñ—É–Ω–∫—Ü–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Ä–∞—Å—Ç–µ—Ç –±—ã—Å—Ç—Ä–µ–µ, —á—Ç–æ –æ–∑–Ω–∞—á–∞–µ—Ç, —á—Ç–æ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –æ—Ç–∫–∞–∑–∞ 
–≤ —Ç–µ—á–µ–Ω–∏–µ –∑–∞–¥–∞–Ω–Ω–æ–≥–æ –≤—Ä–µ–º–µ–Ω–∏ —É–≤–µ–ª–∏—á–∏–≤–∞–µ—Ç—Å—è.
- –ü—Ä–∏ —É–º–µ–Ω—å—à–µ–Ω–∏–∏ $Œª$ —Ñ—É–Ω–∫—Ü–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Ä–∞—Å—Ç–µ—Ç –º–µ–¥–ª–µ–Ω–Ω–µ–µ, —á—Ç–æ –æ–∑–Ω–∞—á–∞–µ—Ç –±–æ–ª—å—à—É—é –Ω–∞–¥–µ–∂–Ω–æ—Å—Ç—å –∏ 
–±–æ–ª–µ–µ –¥–ª–∏—Ç–µ–ª—å–Ω–æ–µ —Å—Ä–µ–¥–Ω–µ–µ –≤—Ä–µ–º—è —Ä–∞–±–æ—Ç—ã.

3. __–û–±—â–∏–µ –≤—ã–≤–æ–¥—ã__:
- –ü–∞—Ä–∞–º–µ—Ç—Ä $Œª$ –ø–æ–ª–Ω–æ—Å—Ç—å—é –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç —Ö–∞—Ä–∞–∫—Ç–µ—Ä –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è.
- –ü–∞—Ä–∞–º–µ—Ç—Ä $Œª$ –æ–±—Ä–∞—Ç–Ω–æ –ø—Ä–æ–ø–æ—Ä—Ü–∏–æ–Ω–∞–ª–µ–Ω –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–º—É –æ–∂–∏–¥–∞–Ω–∏—é: $E(X) = \frac {1} {Œª}$.
- –î–∏—Å–ø–µ—Ä—Å–∏—è —Ç–∞–∫–∂–µ –∑–∞–≤–∏—Å–∏—Ç –æ—Ç $Œª: D(X) = \frac {1} {Œª_2}$.
- –ß–µ–º –º–µ–Ω—å—à–µ $Œª$, —Ç–µ–º –±–æ–ª–µ–µ —Ä–∞—Å—Ç—è–Ω—É—Ç—ã–º —Å—Ç–∞–Ω–æ–≤–∏—Ç—Å—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ, –∏ —Ç–µ–º –±–æ–ª—å—à–µ —Ä–∞–∑–±—Ä–æ—Å 
–≤–æ–∑–º–æ–∂–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –≤—Ä–µ–º–µ–Ω–∏ —Ä–∞–±–æ—Ç—ã.

–≠—Ç–∏ –∑–∞–∫–æ–Ω–æ–º–µ—Ä–Ω–æ—Å—Ç–∏ –≤–∞–∂–Ω—ã –ø—Ä–∏ –ø—Ä–∞–∫—Ç–∏—á–µ—Å–∫–æ–º –ø—Ä–∏–º–µ–Ω–µ–Ω–∏–∏ –ø–æ–∫–∞–∑–∞—Ç–µ–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –¥–ª—è 
–º–æ–¥–µ–ª–∏—Ä–æ–≤–∞–Ω–∏—è –≤—Ä–µ–º–µ–Ω–∏ –±–µ–∑–æ—Ç–∫–∞–∑–Ω–æ–π —Ä–∞–±–æ—Ç—ã —É—Å—Ç—Ä–æ–π—Å—Ç–≤ –∏ —Å–∏—Å—Ç–µ–º, –ø–æ—Å–∫–æ–ª—å–∫—É –ø–æ–∑–≤–æ–ª—è—é—Ç 
–ø—Ä–æ–≥–Ω–æ–∑–∏—Ä–æ–≤–∞—Ç—å —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫–∏ –Ω–∞–¥–µ–∂–Ω–æ—Å—Ç–∏ –ø—Ä–∏ —Ä–∞–∑–ª–∏—á–Ω—ã—Ö –∏–Ω—Ç–µ–Ω—Å–∏–≤–Ω–æ—Å—Ç—è—Ö –æ—Ç–∫–∞–∑–æ–≤.
"""))
```
