```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Параметры распределения
mu = 20
sigma = 10

# Создаем диапазон значений для построения
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)

# Функция плотности вероятности
pdf = norm.pdf(x, mu, sigma)

# Функция распределения
cdf = norm.cdf(x, mu, sigma)

# Создаем графики
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

# График плотности вероятности
ax1.plot(x, pdf, 'b-', linewidth=2)
ax1.fill_between(x[(x >= mu-3*sigma) & (x <= mu+3*sigma)], 
                 pdf[(x >= mu-3*sigma) & (x <= mu+3*sigma)], 
                 alpha=0.3, color='blue')
ax1.axvline(mu, color='r', linestyle='--', label='μ = 20')
ax1.axvline(mu - sigma, color='g', linestyle=':', label='μ ± σ')
ax1.axvline(mu + sigma, color='g', linestyle=':')
ax1.axvline(mu - 2*sigma, color='orange', linestyle=':', label='μ ± 2σ')
ax1.axvline(mu + 2*sigma, color='orange', linestyle=':')
ax1.axvline(mu - 3*sigma, color='purple', linestyle=':', label='μ ± 3σ')
ax1.axvline(mu + 3*sigma, color='purple', linestyle=':')
ax1.set_title('Функция плотности вероятности')
ax1.set_xlabel('Ошибка высотометра (мм)')
ax1.set_ylabel('Плотность вероятности')
ax1.grid(True)
ax1.legend()

# График функции распределения
ax2.plot(x, cdf, 'b-', linewidth=2)
ax2.axhline(0.025, color='purple', linestyle=':', label='2.5%')
ax2.axhline(0.975, color='purple', linestyle=':', label='97.5%')
ax2.axhline(0.5, color='r', linestyle='--', label='50%')
ax2.set_title('Функция распределения')
ax2.set_xlabel('Ошибка высотометра (мм)')
ax2.set_ylabel('Вероятность P(X ≤ x)')
ax2.grid(True)
ax2.legend()

plt.tight_layout()
plt.show()
```
