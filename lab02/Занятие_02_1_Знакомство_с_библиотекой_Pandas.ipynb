## <center> 1. Типы данных Pandas

Обычно под типами данных понимают некоторое внутреннее представление, которое описывает некий набор данных, определяемый двумя критериями:
- какие значения могут принимать переменные данного типа,
- какие операции могут выполняться над переменными данного типа.

Так, в анализе данных базовыми являются два типа данных: **категориальные данные** (качественные) и **числовые данные** (количественные).
- **Категориальные данные**: нечисловая информация, такая как пол, религия, семейное положение, уровень образования и т. д.
- **Числовые данные**: измерение или подсчет, например рост, вес, возраст, зарплата, количество детей и т. д.

Основное отличие числовых данных от категориальных в том, что с числовыми данными можно выполнять арифметические операции и их результат имеет смысл. Так, *стипендия* - это числовая величина. Если умножить её на 6, то получим стипендию за шесть месяцев. *Название вуза* - категориальная величина, т.к. невозможно выполнить с ним математические операции, например, название "ТюмГУ" разделить на 17. Да и номер аудитории в учебном корпусе является категориальной величиной, хоть и выглядит как число (217, например), и его даже можно умножить, например на 5, но полученный <u>результат не имеет смысла</u>.
    
Каждый из этих двух базовых типов принято разбивать на более мелкие подтипы (дискретные, непрерывные, бинарные, порядковые, ... и прочие значения). Знать отличие этих подтипов друг от друга надо обязательно. В презентации к этому занятию **Машинное обучение и данные.pdf** рассказано о системе типов данных в машинном обучении. В языках программирования также существует понятие типов данных, при этом к существующим двум критериям отдельного типа (какие данные представлены в данном типе и какие операции над данными допустимы) рассматривается ещё один:

## <center> 2. Структура Series

### Создание Series

Есть несколько способов создать объект **Series**. В большинстве случаев при создании ему передаётся одномерный набор данных. Вот так можно создать **Series** из списка названий известных вам библиотек Python:

```python
biblio = ['NumPy','SciPy','Pandas']
bib_names = pd.Series(biblio)
print(bib_names)
```

Обратите внимание, что список имён библиотек по умолчанию проиндексирован целыми числами, начиная с 0.
Атрибут **dtype** указывает тип данных, содержащихся в данной серии. По умолчанию в библиотеке Pandas для хранения строк используется тип **object**.

#### Cоздание серии с произвольными индексами:


```python
bib_names = pd.Series(['NumPy','SciPy','Pandas','Matplotlib'], index=[1995,2001,2008,2003])
print(bib_names)
```

Мы создали список (серию), где индексами выступают годы создания данных библиотек.

#### Cоздание именованной серии случайных чисел:


```python
rnd = pd.Series(np.random.randn(4), name='randoms')
rnd
```

#### Cоздание серии из списков:


```python
l = [[1, 2]]
m=[['+','-']]
s = pd.Series(l*2+m*3)
s
```

Создана серия из двух элементов **l** и трёх элементов **m**.

#### Cоздание серии из словаря:


```python
Simps = pd.Series({'Homer':  'Dad',
                   'Marge':  'Mom',
                   'Bart':   'Son',
                   'Lisa':   'Daughter',
                   'Maggie': 'Daughter'}, name='Simpsons')
Simps
```

В созданной серии **Simpsons** хранятся роли членов семьи, а их имена стали индексами.

#### Cоздание серии последовательных чисел:


```python
nums = pd.Series(np.arange(11,33,3))
nums
```

Создана серия из подряд идущих чисел от 11 до 33 с шагом 3.

### Доступ к данным серии
К элементам серии можно обратиться как к элементам массива - имя серии и индекс в квадратных скобках:

```python
print(bib_names)
print("\n","В 2001 году создана библиотека ",bib_names[2001])
```

В качестве альтернативы можно использовать свойство **loc** объекта Series:

```python
print(bib_names.loc[2003])
```

К элементу серии можно обратиться и по его порядковому номеру (свойство **iloc**):

```python
print(bib_names.iloc[3])
```

Обратите внимание на отличие индекса (**loc**) элемента серии и его порядкового номера (**iloc**).

## <center> 3. Структура DataFrame

Объект Pandas под названием **DataFrame** — это двумерная маркированная структура данных, столбцы которой могут содержать значения разных типов. DataFrame можно рассматривать как контейнер для объектов Series. Он похож на словарь, где ключ — метка (имя) столбца, а значение — серия.

С другой стороны, DataFrame можно рассматривать как набор однотипных строк, имеющих индексы.

Объект  DataFrame мы часто будем называть **фреймом данных** или просто **фреймом**.

Подробнее о структуре DataFrame можно прочитать [здесь](https://docs-python.ru/packages/modul-pandas-analiz-dannykh-python/obekt-dataframe/).

### Создание DataFrame

#### Cоздание объекта DataFrame из двумерного списка:
Самое простое - создать DataFrame из двумерной прямоугольной таблицы - списка списков.
Каждый элемент внешнего списка представляет строку фрейма данных, элемент внутреннего списка - элемент строки фрейма данных :

```python
df = pd.DataFrame([[10, 11], [22, 33], [44, 55]])
df
```

Мы получили DataFrame, состоящий из двух серий (двух столбцов), в каждой из которых по 3 элемента.

Индексы элементов каждой серии по умолчанию 0,1 и 2. Они записаны слева от каждой строки с данными. Т.к. у серий мы не задавали названия (имена), то они обозначены порядковыми номерами (начиная с нуля).

С другой стороны, мы получили  DataFrame, состоящий из трёх **записей** (трёх строк), в каждой из которых по 2 элемента.


Дадим сериям (столбцам фрейма) имена **А** и **В**:


```python
df.columns = ['A', 'B']
df
```

Эти имена (**А** и **В**) стали заголовками столбцов образованного фрейма данных.


#### Cоздание объекта DataFrame из готовых серий:
Фрейм данных можно создать на основе словаря (dict) из готовых серий.
Сначала создадим две серии s1 и s2, содержащие по три числа и три строки соответственно. Потом создадим фрейм из этих двух серий, объединённых в один словарь. Сериям установим ключи словаря 'S1' и 'S2' соответственно. Эти ключи станут именами серий (заголовками столбцов):

```python
s1 = pd.Series([11,22,33])
s2 = pd.Series(['one','two','three'])
df = pd.DataFrame({'S1': s1,
                   'S2': s2})
df
```

В аналогичном примере ниже были использованы два объекта Series, созданные раньше: **bib_names** и **rnd**.
    Но результат получился странным. Объясните его.

```python
df_1 = pd.DataFrame({'col_1': bib_names,
                   'col_2': rnd})
df_1
```

Ответ: Тут дело в том, что элементы обеих серий (**bib_names** и **rnd**) имеют индексы. Т.к. там все индексы <u>разные</u>, то и значения записаны в <u>разные</u> строки фрейма. Если бы хотя бы один индекс совпал, то соответствующме значения <u>разных</u> серий оказались в одной строке с этим индексом.

#### Cоздание объекта DataFrame размерности 4х3 из случайных чисел
Пример ниже показывает, как можно создать такой фрейм данных одной командой. Источником данных в конструкторе объекта типа **DataFrame** является NumPy-массив построенный из 12 чисел, имеющих нормальное  распределение, и имеющий размерность 4х3 (3 серии по 4 элемента). Также в конструкторе указаны строковые значения индексов и названия столбцов (серий). Предварительно оператором <code>np.random.seed(77)</code> зафиксировано, что при каждом запуске будут генерироваться одинаковые наборы случайных чисел.

```python
np.random.seed(77)
df = pd.DataFrame(np.random.normal(size=12).reshape(4, 3),
                  index=['ind_1', 'ind_2', 'ind_3', 'ind_4'],
                  columns=['col_1', 'col_2', 'col_3'])
df
```

### CSV - файлы
Pandas поддерживает операции чтения и записи для файлов разного типа: TXT, CSV, Excel 2007+, SQL, HTML, JSON и др. Но наиболее удобным многие пользователи считают использование файл формата **.csv**, структура которого соответствует таблице DataFrame, а содержимое можно открыть и изучить в любом текстовом редакторе.

**CSV (Comma-Separated Values)** - это формат файла, используемый для хранения и обмена табличными данными. В CSV-файле данные представлены в виде текста, где отдельные записи представлены строками, а значения одной строки разделены запятыми (или другим разделителем). CSV - один из самых простых и распространенных форматов для обмена данными между различными приложениями.

**Основные характеристики CSV-файлов включают:**

- **Разделение значений**: Значения в CSV-файле разделяются определенным символом, который чаще всего является запятой, но также может быть точкой с запятой, табуляцией и т. д. Этот символ называется разделителем.
- **Строки данных**: Каждая строка в CSV-файле представляет собой запись данных. Эти строки разделяются символами новой строки (или перевода строки), и каждая строка содержит набор значений, разделенных разделителем.
- **Имена cтолбцов**: Часто в CSV-файлах первая строка содержит названия столбцов через разделитель, что позволяет понять, какие данные хранятся в каждом столбце. Это облегчает создание фрейма данных по содержанию файла.
- **Заголовок файла**: Иногда в CSV-файлах первые строки (перед названиями столбцов) содержат разную описательную информацию о содержимом файла (название датасета, источник данных, права на использование и т.п.). Т.к. подобная информация не входит в содержание фрейма данных, то эти строки надо будет пропустить при чтении данных. Ниже будет сказано, как это сделать.

#### Учёт наличия строк заголовка в CSV-файле
Иногда первые несколько строк CSV-файла могут содержать справочную информацию для пользователя (заголовок набора данных, его описание). Чтобы пропустить их при загрузке данных из файла, используют параметр  **</code> header = ... </code>**, указывающий сколько первых строк надо пропустить:

Обратите внимание, что  **len** - это функция, аргументом которой является объект данных, а **size** и **shape** - свойства объекта данных, которые вызываются через точку после его имени.

### Изменение значений отдельных столбцов фрейма
Если в обычном массиве для однотипного изменения значений всего столбца надо в цикле перебрать его элементы, то во фрейме можно преобразование всего столбца выполнить одной командой. Для этого выделяем отдельный столбец фрейма как серию, применяем ко всем его элементам нужную операцию и записываем изменённую серию на место этого столбца.

В качестве примера переведём рост и вес людей из фрейма **group** в привычные нам сантиметры и килограммы.

```python
group.head()
```

```python
group['Height']=group['Height']*2.54
group['Weight']=group['Weight']*0.4536
group.head()
```

### Изменение типа данных для отдельных столбцов фрейма
Если для дальнейшей работы с данными тип значений отдельного столбца ('column_name') необходимо изменить на другой тип('data_type'), то такое преобразование реализует метод **astype()**:

Например, после перевода в сантиметры и килограммы значения роста и веса во фрейме group стали вещественными с точностью до сотой доли сантиметра и десятой доли грамма. Такая точность для этих данных излишняя и мы опять приведём значения роста и веса к целому типу:


```python
group['Weight'] = group['Weight'].astype('int')
group['Height'] = group['Height'].astype('int')
group.head()
```

### Наличие пустых ячеек
Иногда некоторые данные во фрейме отсутствуют. При выводе это принято обозначать NaN (not a number).
Сделаем значение веса у Карла во фрейме group пустым

```python
group.loc[2,'Weight']=np.NaN;
group.head()
```

Один из способов проверить наличие NaN в DataFrame — использовать функцию **isnull()**. Она возвращает другой DataFrame, где каждое значение это булево значение, указывающее, является ли соответствующее значение в исходном DataFrame NaN.

```python
group.isnull().head()
```

Чтобы узнать, есть ли хотя бы одно значение NaN в DataFrame, можно использовать функцию **any()**, которая возвращает True, если хотя бы одно значение в DataFrame является True.

```python
group.isnull().any()
```
