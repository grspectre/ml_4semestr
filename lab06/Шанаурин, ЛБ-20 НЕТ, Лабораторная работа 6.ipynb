–∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏–µ –ø–æ–ª—É—á–µ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–∏. –î–ª—è —ç—Ç–æ–≥–æ:

3. –ù–∞–ø–∏—Å–∞—Ç—å —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è –∑–Ω–∞—á–µ–Ω–∏–π —ç–º–ø–∏—Ä–∏—á–µ—Å–∫–æ–π —Ñ—É–Ω–∫—Ü–∏–∏
—Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è.

—Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è. –ü—Ä–æ–∫–æ–º–º–µ–Ω—Ç–∏—Ä–æ–≤–∞—Ç—å –ø–æ–ª—É—á–µ–Ω–Ω—ã–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã (–≤ —Ç–µ–∫—Å—Ç–æ–≤–æ–π
—è—á–µ–π–∫–µ).
5. –í—ã—á–∏—Å–ª–∏—Ç—å –≤—ã–±–æ—Ä–æ—á–Ω–æ–µ —Å—Ä–µ–¥–Ω–µ–µ, –≤—ã–±–æ—Ä–æ—á–Ω—É—é –º–æ–¥—É –∏ –≤—ã–±–æ—Ä–æ—á–Ω—É—é –º–µ–¥–∏–∞–Ω—É.
–°–æ–ø–æ—Å—Ç–∞–≤–∏—Ç—å –ø–æ–ª—É—á–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è —Å –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–∏–º –æ–∂–∏–¥–∞–Ω–∏–µ –∏ –º–µ–¥–∏–∞–Ω–æ–π
—Å–ª—É—á–∞–π–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω—ã (–≤—ã—á–∏—Å–ª–∏—Ç—å —ç—Ç–∏ –∑–Ω–∞—á–µ–Ω–∏—è –∏–ª–∏ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –∏–∑–≤–µ—Å—Ç–Ω—ã–µ

—Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã (–≤ —Ç–µ–∫—Å—Ç–æ–≤–æ–π —è—á–µ–π–∫–µ).
6. –í—ã—á–∏—Å–ª–∏—Ç—å –≤—ã–±–æ—Ä–æ—á–Ω—É—é –¥–∏—Å–ø–µ—Ä—Å–∏—é –∏ –∏—Å–ø—Ä–∞–≤–ª–µ–Ω–Ω—É—é –¥–∏—Å–ø–µ—Ä—Å–∏—é. –°–æ–ø–æ—Å—Ç–∞–≤–∏—Ç—å
–ø–æ–ª—É—á–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è —Å –¥–∏—Å–ø–µ—Ä—Å–∏–µ–π —Å–ª—É—á–∞–π–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω—ã (–≤—ã—á–∏—Å–ª–∏—Ç—å —ç—Ç–æ
–∑–Ω–∞—á–µ–Ω–∏—è –∏–ª–∏ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –∏–∑–≤–µ—Å—Ç–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –¥–ª—è –¥–∞–Ω–Ω–æ–≥–æ –∑–∞–∫–æ–Ω–∞
—Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è). –ü—Ä–æ–∫–æ–º–º–µ–Ω—Ç–∏—Ä–æ–≤–∞—Ç—å –ø–æ–ª—É—á–µ–Ω–Ω—ã–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã (–≤ —Ç–µ–∫—Å—Ç–æ–≤–æ–π
—è—á–µ–π–∫–µ).
7. –í—ã—á–∏—Å–ª–∏—Ç—å –≤—ã–±–æ—Ä–æ—á–Ω–æ–µ —Å—Ä–µ–¥–Ω–µ–µ –∫–≤–∞–¥—Ä–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ –∏ –∏—Å–ø—Ä–∞–≤–ª–µ–Ω–Ω–æ–µ
—Å—Ä–µ–¥–Ω–µ–µ –∫–≤–∞–¥—Ä–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ. –°–æ–ø–æ—Å—Ç–∞–≤–∏—Ç—å –ø–æ–ª—É—á–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è —Å–æ
—Å—Ä–µ–¥–Ω–∏–º –∫–≤–∞–¥—Ä–∞—Ç–∏—á–µ—Å–∫–∏–º –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ–º —Å–ª—É—á–∞–π–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω—ã (–≤—ã—á–∏—Å–ª–∏—Ç—å —ç—Ç–æ
–∑–Ω–∞—á–µ–Ω–∏—è –∏–ª–∏ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –∏–∑–≤–µ—Å—Ç–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –¥–ª—è –¥–∞–Ω–Ω–æ–≥–æ –∑–∞–∫–æ–Ω–∞
—Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è). –ü—Ä–æ–∫–æ–º–º–µ–Ω—Ç–∏—Ä–æ–≤–∞—Ç—å –ø–æ–ª—É—á–µ–Ω–Ω—ã–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã (–≤ —Ç–µ–∫—Å—Ç–æ–≤–æ–π
—è—á–µ–π–∫–µ).


```python
# –ò–º–ø–æ—Ä—Ç –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã—Ö –±–∏–±–ª–∏–æ—Ç–µ–∫
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
from IPython.display import display, Markdown, HTML


def empirical_cdf(x, sample):
    return np.mean(sample <= x)


# –ù–∞—Å—Ç—Ä–æ–π–∫–∞ —Å—Ç–∏–ª—è –≥—Ä–∞—Ñ–∏–∫–æ–≤
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_context("notebook", font_scale=1.2)

# –î–ª—è –≤–æ—Å–ø—Ä–æ–∏–∑–≤–æ–¥–∏–º–æ—Å—Ç–∏ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
np.random.seed(556)

display(Markdown(r'''# –ó–∞–¥–∞–Ω–∏–µ 1
## 1. –°–æ–∑–¥–∞–Ω–∏–µ –≤—ã–±–æ—Ä–∫–∏ –∏–∑ –≥–µ–Ω–µ—Ä–∞–ª—å–Ω–æ–π —Å–æ–≤–æ–∫—É–ø–Ω–æ—Å—Ç–∏'''))

# –ü–∞—Ä–∞–º–µ—Ç—Ä—ã –±–∏–Ω–æ–º–∏–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è
n_trials = 10  # –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏—Å–ø—ã—Ç–∞–Ω–∏–π –≤ –±–∏–Ω–æ–º–∏–∞–ª—å–Ω–æ–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–∏
p_success = 0.3  # –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —É—Å–ø–µ—Ö–∞ –≤ –∫–∞–∂–¥–æ–º –∏—Å–ø—ã—Ç–∞–Ω–∏–∏
sample_size = 25  # —Ä–∞–∑–º–µ—Ä –≤—ã–±–æ—Ä–∫–∏

# –ì–µ–Ω–µ—Ä–∞—Ü–∏—è –≤—ã–±–æ—Ä–∫–∏ –∏–∑ –±–∏–Ω–æ–º–∏–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è
sample = np.random.binomial(n=n_trials, p=p_success, size=sample_size)

# –í—ã–≤–æ–¥ –≤—ã–±–æ—Ä–∫–∏
print("–í—ã–±–æ—Ä–∫–∞:", sample)
print("–†–∞–∑–º–µ—Ä –≤—ã–±–æ—Ä–∫–∏:", len(sample))

display(Markdown(r'''## 2. –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –ø–æ–ª–∏–≥–æ–Ω–∞ —á–∞—Å—Ç–æ—Ç'''))
# –ü–æ–¥—Å—á–µ—Ç —á–∞—Å—Ç–æ—Ç
value_counts = pd.Series(sample).value_counts().sort_index()
frequencies = {x: 0 for x in range(n_trials + 1)}
for value, count in value_counts.items():
    frequencies[value] = count

x_values = list(frequencies.keys())
y_values = list(frequencies.values())

# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –ø–æ–ª–∏–≥–æ–Ω–∞ —á–∞—Å—Ç–æ—Ç
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, 'b-o', linewidth=2, markersize=8)
plt.title('–ü–æ–ª–∏–≥–æ–Ω —á–∞—Å—Ç–æ—Ç –±–∏–Ω–æ–º–∏–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è', fontsize=14)
plt.xlabel('–ó–Ω–∞—á–µ–Ω–∏–µ —Å–ª—É—á–∞–π–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω—ã', fontsize=12)
plt.ylabel('–ß–∞—Å—Ç–æ—Ç–∞', fontsize=12)
plt.grid(True, alpha=0.3)
plt.xticks(range(n_trials + 1))
plt.tight_layout()
plt.show()

display(Markdown(r'''## 3. –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –ø–æ–ª–∏–≥–æ–Ω–∞ –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω—ã—Ö —á–∞—Å—Ç–æ—Ç –∏ —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–æ–≥–æ –º–Ω–æ–≥–æ—É–≥–æ–ª—å–Ω–∏–∫–∞ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è'''))
# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω—ã—Ö —á–∞—Å—Ç–æ—Ç
rel_frequencies = {k: v / sample_size for k, v in frequencies.items()}

# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏—Ö –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π –±–∏–Ω–æ–º–∏–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è
x_theoretical = np.arange(0, n_trials + 1)
theoretical_probs = stats.binom.pmf(x_theoretical, n_trials, p_success)

# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–æ–≤
plt.figure(figsize=(10, 6))
plt.plot(x_values, list(rel_frequencies.values()), 'b-o', label='–≠–º–ø–∏—Ä–∏—á–µ—Å–∫–∏–µ –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω—ã–µ —á–∞—Å—Ç–æ—Ç—ã', linewidth=2, markersize=8)
plt.plot(x_theoretical, theoretical_probs, 'r-s', label='–¢–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏–µ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏', linewidth=2, markersize=8)
plt.title('–°—Ä–∞–≤–Ω–µ–Ω–∏–µ —ç–º–ø–∏—Ä–∏—á–µ—Å–∫–æ–≥–æ –∏ —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–π', fontsize=14)
plt.xlabel('–ó–Ω–∞—á–µ–Ω–∏–µ —Å–ª—É—á–∞–π–Ω–æ–π –≤–µ–ª–∏—á–∏–Ω—ã', fontsize=12)
plt.ylabel('–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å / –û—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–∞—è —á–∞—Å—Ç–æ—Ç–∞', fontsize=12)
plt.grid(True, alpha=0.3)
plt.xticks(range(n_trials + 1))
plt.legend(fontsize=12)
plt.tight_layout()
plt.show()

display(Markdown(r"""
### –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π –∫ –ø–æ–ª—É—á–µ–Ω–Ω—ã–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º
–ù–∞ –≥—Ä–∞—Ñ–∏–∫–µ –≤–∏–¥–Ω–æ —Å—Ä–∞–≤–Ω–µ–Ω–∏–µ —ç–º–ø–∏—Ä–∏—á–µ—Å–∫–∏—Ö –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω—ã—Ö —á–∞—Å—Ç–æ—Ç —Å —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏–º–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—è–º–∏ –±–∏–Ω–æ–º–∏–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Å –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏ n=10 –∏ p=0.3. –ù–∞–±–ª—é–¥–∞—é—Ç—Å—è –Ω–µ–∫–æ—Ç–æ—Ä—ã–µ —Ä–∞—Å—Ö–æ–∂–¥–µ–Ω–∏—è –º–µ–∂–¥—É —ç–º–ø–∏—Ä–∏—á–µ—Å–∫–∏–º–∏ –∏ —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏, —á—Ç–æ —è–≤–ª—è–µ—Ç—Å—è –Ω–æ—Ä–º–∞–ª—å–Ω—ã–º —è–≤–ª–µ–Ω–∏–µ–º –¥–ª—è –≤—ã–±–æ—Ä–∫–∏ –º–∞–ª–æ–≥–æ –æ–±—ä–µ–º–∞ (n=25). –≠—Ç–∏ —Ä–∞—Å—Ö–æ–∂–¥–µ–Ω–∏—è –≤–æ–∑–Ω–∏–∫–∞—é—Ç –∏–∑-–∑–∞ —Å–ª—É—á–∞–π–Ω–æ–π –ø—Ä–∏—Ä–æ–¥—ã –≤—ã–±–æ—Ä–∫–∏. –ü—Ä–∏ —É–≤–µ–ª–∏—á–µ–Ω–∏–∏ –æ–±—ä–µ–º–∞ –≤—ã–±–æ—Ä–∫–∏ –æ–∂–∏–¥–∞–µ—Ç—Å—è, —á—Ç–æ —ç–º–ø–∏—Ä–∏—á–µ—Å–∫–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –±—É–¥–µ—Ç –≤—Å—ë –±–ª–∏–∂–µ –ø—Ä–∏–±–ª–∏–∂–∞—Ç—å—Å—è –∫ —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–æ–º—É —Å–æ–≥–ª–∞—Å–Ω–æ –∑–∞–∫–æ–Ω—É –±–æ–ª—å—à–∏—Ö —á–∏—Å–µ–ª.
## 4. –§—É–Ω–∫—Ü–∏—è –¥–ª—è –≤—ã—á–∏—Å–ª–µ–Ω–∏—è –∑–Ω–∞—á–µ–Ω–∏–π —ç–º–ø–∏—Ä–∏—á–µ—Å–∫–æ–π —Ñ—É–Ω–∫—Ü–∏–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è
"""))

# –ü—Ä–æ–≤–µ—Ä–∫–∞ —Ñ—É–Ω–∫—Ü–∏–∏
x_points = np.linspace(-1, n_trials + 1, 100)
emp_cdf_values = [empirical_cdf(x, sample) for x in x_points]

# –¢–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∞—è —Ñ—É–Ω–∫—Ü–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è
theoretical_cdf_values = stats.binom.cdf(x_points, n_trials, p_success)

# –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–π —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è
plt.figure(figsize=(10, 6))
plt.plot(x_points, emp_cdf_values, 'b-', label='–≠–º–ø–∏—Ä–∏—á–µ—Å–∫–∞—è —Ñ—É–Ω–∫—Ü–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è', linewidth=2)
plt.plot(x_points, theoretical_cdf_values, 'r--', label='–¢–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∞—è —Ñ—É–Ω–∫—Ü–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è', linewidth=2)
plt.title('–°—Ä–∞–≤–Ω–µ–Ω–∏–µ —ç–º–ø–∏—Ä–∏—á–µ—Å–∫–æ–π –∏ —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–æ–π —Ñ—É–Ω–∫—Ü–∏–π —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('F(x)', fontsize=12)
plt.grid(True, alpha=0.3)
plt.legend(fontsize=12)
plt.tight_layout()
plt.show()

display(Markdown(r"""
### –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π –∫ –ø–æ–ª—É—á–µ–Ω–Ω—ã–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º
–ì—Ä–∞—Ñ–∏–∫ –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç —Å—Ä–∞–≤–Ω–µ–Ω–∏–µ —ç–º–ø–∏—Ä–∏—á–µ—Å–∫–æ–π —Ñ—É–Ω–∫—Ü–∏–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è (–ø–æ—Å—Ç—Ä–æ–µ–Ω–Ω–æ–π –ø–æ –Ω–∞—à–µ–π –≤—ã–±–æ—Ä–∫–µ) –∏ —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–æ–π —Ñ—É–Ω–∫—Ü–∏–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –¥–ª—è –±–∏–Ω–æ–º–∏–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è. –≠–º–ø–∏—Ä–∏—á–µ—Å–∫–∞—è —Ñ—É–Ω–∫—Ü–∏—è –∏–º–µ–µ—Ç —Å—Ç—É–ø–µ–Ω—á–∞—Ç—ã–π –≤–∏–¥, –≥–¥–µ –∫–∞–∂–¥–∞—è —Å—Ç—É–ø–µ–Ω—å–∫–∞ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É–µ—Ç –Ω–∞–±–ª—é–¥–∞–µ–º–æ–º—É –∑–Ω–∞—á–µ–Ω–∏—é –≤ –≤—ã–±–æ—Ä–∫–µ. –ú–æ–∂–Ω–æ –∑–∞–º–µ—Ç–∏—Ç—å, —á—Ç–æ —ç–º–ø–∏—Ä–∏—á–µ—Å–∫–∞—è —Ñ—É–Ω–∫—Ü–∏—è –¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ —Ö–æ—Ä–æ—à–æ –∞–ø–ø—Ä–æ–∫—Å–∏–º–∏—Ä—É–µ—Ç —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫—É—é, —Ö–æ—Ç—è –∏ –∏–º–µ–µ—Ç –º–µ—Å—Ç–æ –Ω–µ–∫–æ—Ç–æ—Ä–æ–µ —Ä–∞—Å—Ö–æ–∂–¥–µ–Ω–∏–µ –∏–∑-–∑–∞ –∫–æ–Ω–µ—á–Ω–æ–≥–æ —Ä–∞–∑–º–µ—Ä–∞ –≤—ã–±–æ—Ä–∫–∏. –§—É–Ω–∫—Ü–∏—è —ç–º–ø–∏—Ä–∏—á–µ—Å–∫–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ø—Ä–∏–±–ª–∏–∂–∞–µ—Ç—Å—è –∫ —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–æ–π —Ñ—É–Ω–∫—Ü–∏–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Å —É–≤–µ–ª–∏—á–µ–Ω–∏–µ–º –æ–±—ä–µ–º–∞ –≤—ã–±–æ—Ä–∫–∏, —á—Ç–æ —è–≤–ª—è–µ—Ç—Å—è —Å–ª–µ–¥—Å—Ç–≤–∏–µ–º —Ç–µ–æ—Ä–µ–º—ã –ì–ª–∏–≤–µ–Ω–∫–æ-–ö–∞–Ω—Ç–µ–ª–ª–∏.
## 5. –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –≤—ã–±–æ—Ä–æ—á–Ω—ã—Ö —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫ –ø–æ–ª–æ–∂–µ–Ω–∏—è –∏ —Å—Ä–∞–≤–Ω–µ–Ω–∏–µ —Å —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏
"""))

# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –≤—ã–±–æ—Ä–æ—á–Ω—ã—Ö —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫
sample_mean = np.mean(sample)
sample_mode = stats.mode(sample, keepdims=True).mode[0]  # –í—ã–±–æ—Ä–æ—á–Ω–∞—è –º–æ–¥–∞
sample_median = np.median(sample)

# –¢–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏–µ —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫–∏ –±–∏–Ω–æ–º–∏–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è
theoretical_mean = n_trials * p_success
theoretical_mode = int(np.floor((n_trials + 1) * p_success))  # –î–ª—è –±–∏–Ω–æ–º–∏–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è
theoretical_median = np.floor(n_trials * p_success + 0.5)  # –ü—Ä–∏–±–ª–∏–∂–µ–Ω–∏–µ –¥–ª—è –±–∏–Ω–æ–º–∏–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è

# –í—ã–≤–æ–¥ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
print("–•–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫–∏ –ø–æ–ª–æ–∂–µ–Ω–∏—è:")
print(f"–í—ã–±–æ—Ä–æ—á–Ω–æ–µ —Å—Ä–µ–¥–Ω–µ–µ: {sample_mean:.4f}")
print(f"–¢–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–æ–µ –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ: {theoretical_mean:.4f}")
print(f"–û—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ: {abs(sample_mean - theoretical_mean):.4f}")
print("\n")
print(f"–í—ã–±–æ—Ä–æ—á–Ω–∞—è –º–æ–¥–∞: {sample_mode}")
print(f"–¢–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∞—è –º–æ–¥–∞: {theoretical_mode}")
print(f"–û—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ: {abs(sample_mode - theoretical_mode)}")
print("\n")
print(f"–í—ã–±–æ—Ä–æ—á–Ω–∞—è –º–µ–¥–∏–∞–Ω–∞: {sample_median}")
print(f"–¢–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∞—è –º–µ–¥–∏–∞–Ω–∞: {theoretical_median}")
print(f"–û—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ: {abs(sample_median - theoretical_median)}")

display(Markdown(r"""
### –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π –∫ –ø–æ–ª—É—á–µ–Ω–Ω—ã–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º
–ü—Ä–∏ —Å—Ä–∞–≤–Ω–µ–Ω–∏–∏ –≤—ã–±–æ—Ä–æ—á–Ω—ã—Ö —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫ –ø–æ–ª–æ–∂–µ–Ω–∏—è —Å –∏—Ö —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏–º–∏ –∞–Ω–∞–ª–æ–≥–∞–º–∏ –º–æ–∂–Ω–æ –Ω–∞–±–ª—é–¥–∞—Ç—å –Ω–µ–∫–æ—Ç–æ—Ä—ã–µ —Ä–∞—Å—Ö–æ–∂–¥–µ–Ω–∏—è, —á—Ç–æ —Ç–∏–ø–∏—á–Ω–æ –¥–ª—è –≤—ã–±–æ—Ä–æ–∫ –º–∞–ª–æ–≥–æ –æ–±—ä–µ–º–∞:

1. __–°—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ__: –í—ã–±–æ—Ä–æ—á–Ω–æ–µ —Å—Ä–µ–¥–Ω–µ–µ –æ—Ç–∫–ª–æ–Ω—è–µ—Ç—Å—è –æ—Ç —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–æ–≥–æ –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–≥–æ –æ–∂–∏–¥–∞–Ω–∏—è. –ú–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ –¥–ª—è –±–∏–Ω–æ–º–∏–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Å –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏ n=10 –∏ p=0.3 —Ä–∞–≤–Ω–æ 3.0, —Ç–æ–≥–¥–∞ –∫–∞–∫ –≤—ã–±–æ—Ä–æ—á–Ω–æ–µ —Å—Ä–µ–¥–Ω–µ–µ –º–æ–∂–µ—Ç –Ω–µ—Å–∫–æ–ª—å–∫–æ –æ—Ç–ª–∏—á–∞—Ç—å—Å—è –∏–∑-–∑–∞ —Å–ª—É—á–∞–π–Ω–æ—Å—Ç–∏ –≤—ã–±–æ—Ä–∫–∏.
2. __–ú–æ–¥–∞__: –í—ã–±–æ—Ä–æ—á–Ω–∞—è –º–æ–¥–∞ –º–æ–∂–µ—Ç –Ω–µ —Å–æ–≤–ø–∞–¥–∞—Ç—å —Å —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–æ–π. –î–ª—è –±–∏–Ω–æ–º–∏–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –º–æ–¥–∞ —Ä–∞–≤–Ω–∞ —Ü–µ–ª–æ–π —á–∞—Å—Ç–∏ –æ—Ç (n+1)p, —Ç–æ –µ—Å—Ç—å 3.
3. __–ú–µ–¥–∏–∞–Ω–∞__: –í—ã–±–æ—Ä–æ—á–Ω–∞—è –º–µ–¥–∏–∞–Ω–∞ —Ç–∞–∫–∂–µ –º–æ–∂–µ—Ç –æ—Ç–ª–∏—á–∞—Ç—å—Å—è –æ—Ç —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–æ–π. –î–ª—è –±–∏–Ω–æ–º–∏–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –º–µ–¥–∏–∞–Ω–∞ –ø—Ä–∏–º–µ—Ä–Ω–æ —Ä–∞–≤–Ω–∞ np (–∏–Ω–æ–≥–¥–∞ —Å –ø–æ–ø—Ä–∞–≤–∫–æ–π 0.5).

–≠—Ç–∏ —Ä–∞—Å—Ö–æ–∂–¥–µ–Ω–∏—è –≤–ø–æ–ª–Ω–µ –æ–∂–∏–¥–∞–µ–º—ã –∏ —É–º–µ–Ω—å—à–∞—é—Ç—Å—è —Å —É–≤–µ–ª–∏—á–µ–Ω–∏–µ–º –æ–±—ä–µ–º–∞ –≤—ã–±–æ—Ä–∫–∏. –û–Ω–∏ –ø–æ–¥—Ç–≤–µ—Ä–∂–¥–∞—é—Ç —Å–ª—É—á–∞–π–Ω—É—é –ø—Ä–∏—Ä–æ–¥—É –≤—ã–±–æ—Ä–∫–∏ –∏ –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–Ω–æ—Å—Ç—å –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏, –∫–æ—Ç–æ—Ä—É—é –º–æ–∂–Ω–æ –∏–∑–≤–ª–µ—á—å –∏–∑ –º–∞–ª—ã—Ö –≤—ã–±–æ—Ä–æ–∫.

## 6. –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –≤—ã–±–æ—Ä–æ—á–Ω–æ–π –¥–∏—Å–ø–µ—Ä—Å–∏–∏ –∏ –∏—Å–ø—Ä–∞–≤–ª–µ–Ω–Ω–æ–π –¥–∏—Å–ø–µ—Ä—Å–∏–∏
"""))

# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –≤—ã–±–æ—Ä–æ—á–Ω–æ–π –¥–∏—Å–ø–µ—Ä—Å–∏–∏
sample_var = np.var(sample, ddof=0)  # ddof=0 –¥–ª—è –≤—ã–±–æ—Ä–æ—á–Ω–æ–π –¥–∏—Å–ø–µ—Ä—Å–∏–∏

# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –∏—Å–ø—Ä–∞–≤–ª–µ–Ω–Ω–æ–π –¥–∏—Å–ø–µ—Ä—Å–∏–∏
corrected_var = np.var(sample, ddof=1)  # ddof=1 –¥–ª—è –∏—Å–ø—Ä–∞–≤–ª–µ–Ω–Ω–æ–π –¥–∏—Å–ø–µ—Ä—Å–∏–∏

# –¢–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∞—è –¥–∏—Å–ø–µ—Ä—Å–∏—è –±–∏–Ω–æ–º–∏–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è
theoretical_var = n_trials * p_success * (1 - p_success)

# –í—ã–≤–æ–¥ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
print("–•–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫–∏ —Ä–∞—Å—Å–µ—è–Ω–∏—è (–¥–∏—Å–ø–µ—Ä—Å–∏—è):")
print(f"–í—ã–±–æ—Ä–æ—á–Ω–∞—è –¥–∏—Å–ø–µ—Ä—Å–∏—è: {sample_var:.4f}")
print(f"–ò—Å–ø—Ä–∞–≤–ª–µ–Ω–Ω–∞—è –¥–∏—Å–ø–µ—Ä—Å–∏—è: {corrected_var:.4f}")
print(f"–¢–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∞—è –¥–∏—Å–ø–µ—Ä—Å–∏—è: {theoretical_var:.4f}")
print(f"–û—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ –≤—ã–±–æ—Ä–æ—á–Ω–æ–π –¥–∏—Å–ø–µ—Ä—Å–∏–∏ –æ—Ç —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–æ–π: {abs(sample_var - theoretical_var):.4f}")
print(f"–û—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ –∏—Å–ø—Ä–∞–≤–ª–µ–Ω–Ω–æ–π –¥–∏—Å–ø–µ—Ä—Å–∏–∏ –æ—Ç —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–æ–π: {abs(corrected_var - theoretical_var):.4f}")

display(Markdown(r"""
### –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π –∫ –ø–æ–ª—É—á–µ–Ω–Ω—ã–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º

–ü—Ä–∏ –∞–Ω–∞–ª–∏–∑–µ –¥–∏—Å–ø–µ—Ä—Å–∏–∏ –Ω–∞–±–ª—é–¥–∞—é—Ç—Å—è —Å–ª–µ–¥—É—é—â–∏–µ –∑–∞–∫–æ–Ω–æ–º–µ—Ä–Ω–æ—Å—Ç–∏:

1. –í—ã–±–æ—Ä–æ—á–Ω–∞—è –¥–∏—Å–ø–µ—Ä—Å–∏—è –≤—ã—á–∏—Å–ª—è–µ—Ç—Å—è –∫–∞–∫ —Å—Ä–µ–¥–Ω–µ–µ –∫–≤–∞–¥—Ä–∞—Ç–æ–≤ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–π –æ—Ç –≤—ã–±–æ—Ä–æ—á–Ω–æ–≥–æ —Å—Ä–µ–¥–Ω–µ–≥–æ. –û–Ω–∞ —è–≤–ª—è–µ—Ç—Å—è —Å–º–µ—â–µ–Ω–Ω–æ–π –æ—Ü–µ–Ω–∫–æ–π —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–æ–π –¥–∏—Å–ø–µ—Ä—Å–∏–∏ –∏ —Å–∏—Å—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–∏ –Ω–µ–¥–æ–æ—Ü–µ–Ω–∏–≤–∞–µ—Ç –∏—Å—Ç–∏–Ω–Ω—É—é –¥–∏—Å–ø–µ—Ä—Å–∏—é –≥–µ–Ω–µ—Ä–∞–ª—å–Ω–æ–π —Å–æ–≤–æ–∫—É–ø–Ω–æ—Å—Ç–∏.
2. –ò—Å–ø—Ä–∞–≤–ª–µ–Ω–Ω–∞—è –¥–∏—Å–ø–µ—Ä—Å–∏—è –≤–∫–ª—é—á–∞–µ—Ç –ø–æ–ø—Ä–∞–≤–æ—á–Ω—ã–π –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç n/(n-1), —á—Ç–æ –¥–µ–ª–∞–µ—Ç –µ—ë –Ω–µ—Å–º–µ—â–µ–Ω–Ω–æ–π –æ—Ü–µ–Ω–∫–æ–π —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–æ–π –¥–∏—Å–ø–µ—Ä—Å–∏–∏. –î–ª—è –≤—ã–±–æ—Ä–∫–∏ –º–∞–ª–æ–≥–æ –æ–±—ä–µ–º–∞ —Ä–∞–∑–Ω–∏—Ü–∞ –º–µ–∂–¥—É –≤—ã–±–æ—Ä–æ—á–Ω–æ–π –∏ –∏—Å–ø—Ä–∞–≤–ª–µ–Ω–Ω–æ–π –¥–∏—Å–ø–µ—Ä—Å–∏–µ–π –º–æ–∂–µ—Ç –±—ã—Ç—å –∑–Ω–∞—á–∏—Ç–µ–ª—å–Ω–æ–π.
3. –¢–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∞—è –¥–∏—Å–ø–µ—Ä—Å–∏—è –¥–ª—è –±–∏–Ω–æ–º–∏–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Ä–∞–≤–Ω–∞ np(1-p), —Ç–æ –µ—Å—Ç—å 2.1 –≤ –Ω–∞—à–µ–º —Å–ª—É—á–∞–µ.
4. –ò—Å–ø—Ä–∞–≤–ª–µ–Ω–Ω–∞—è –¥–∏—Å–ø–µ—Ä—Å–∏—è –æ–±—ã—á–Ω–æ –±–ª–∏–∂–µ –∫ —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–æ–π, —á—Ç–æ –ø–æ–¥—Ç–≤–µ—Ä–∂–¥–∞–µ—Ç –µ—ë –∫–∞—á–µ—Å—Ç–≤–æ –∫–∞–∫ –Ω–µ—Å–º–µ—â–µ–Ω–Ω–æ–π –æ—Ü–µ–Ω–∫–∏. –û—Ç–∫–ª–æ–Ω–µ–Ω–∏—è –æ—Ç —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è –æ–±—ä—è—Å–Ω—è—é—Ç—Å—è —Å–ª—É—á–∞–π–Ω–æ—Å—Ç—å—é –≤—ã–±–æ—Ä–∫–∏ –∏ —É–º–µ–Ω—å—à–∞—é—Ç—Å—è —Å —É–≤–µ–ª–∏—á–µ–Ω–∏–µ–º –æ–±—ä–µ–º–∞ –¥–∞–Ω–Ω—ã—Ö.

7. –í—ã—á–∏—Å–ª–µ–Ω–∏–µ —Å—Ä–µ–¥–Ω–µ–≥–æ –∫–≤–∞–¥—Ä–∞—Ç–∏—á–µ—Å–∫–æ–≥–æ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è
"""))

# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –≤—ã–±–æ—Ä–æ—á–Ω–æ–≥–æ —Å—Ä–µ–¥–Ω–µ–≥–æ –∫–≤–∞–¥—Ä–∞—Ç–∏—á–µ—Å–∫–æ–≥–æ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è
sample_std = np.std(sample, ddof=0)  # ddof=0 –¥–ª—è –≤—ã–±–æ—Ä–æ—á–Ω–æ–≥–æ –°–ö–û

# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –∏—Å–ø—Ä–∞–≤–ª–µ–Ω–Ω–æ–≥–æ —Å—Ä–µ–¥–Ω–µ–≥–æ –∫–≤–∞–¥—Ä–∞—Ç–∏—á–µ—Å–∫–æ–≥–æ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è
corrected_std = np.std(sample, ddof=1)  # ddof=1 –¥–ª—è –∏—Å–ø—Ä–∞–≤–ª–µ–Ω–Ω–æ–≥–æ –°–ö–û

# –¢–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–æ–µ —Å—Ä–µ–¥–Ω–µ–µ –∫–≤–∞–¥—Ä–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ
theoretical_std = np.sqrt(theoretical_var)

# –í—ã–≤–æ–¥ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
print("–•–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫–∏ —Ä–∞—Å—Å–µ—è–Ω–∏—è (—Å—Ä–µ–¥–Ω–µ–µ –∫–≤–∞–¥—Ä–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ):")
print(f"–í—ã–±–æ—Ä–æ—á–Ω–æ–µ –°–ö–û: {sample_std:.4f}")
print(f"–ò—Å–ø—Ä–∞–≤–ª–µ–Ω–Ω–æ–µ –°–ö–û: {corrected_std:.4f}")
print(f"–¢–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–æ–µ –°–ö–û: {theoretical_std:.4f}")
print(f"–û—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ –≤—ã–±–æ—Ä–æ—á–Ω–æ–≥–æ –°–ö–û –æ—Ç —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–æ–≥–æ: {abs(sample_std - theoretical_std):.4f}")
print(f"–û—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ –∏—Å–ø—Ä–∞–≤–ª–µ–Ω–Ω–æ–≥–æ –°–ö–û –æ—Ç —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–æ–≥–æ: {abs(corrected_std - theoretical_std):.4f}")

display(Markdown(r"""
### –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π –∫ –ø–æ–ª—É—á–µ–Ω–Ω—ã–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º
–°—Ä–µ–¥–Ω–µ–µ –∫–≤–∞–¥—Ä–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ (–°–ö–û) —è–≤–ª—è–µ—Ç—Å—è –∫–æ—Ä–Ω–µ–º –∏–∑ –¥–∏—Å–ø–µ—Ä—Å–∏–∏ –∏ –±–æ–ª–µ–µ –Ω–∞–≥–ª—è–¥–Ω–æ–π —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫–æ–π —Ä–∞—Å—Å–µ–∏–≤–∞–Ω–∏—è, —Ç–∞–∫ –∫–∞–∫ –≤—ã—Ä–∞–∂–∞–µ—Ç—Å—è –≤ —Ç–µ—Ö –∂–µ –µ–¥–∏–Ω–∏—Ü–∞—Ö, —á—Ç–æ –∏ —Å–∞–º–∞ —Å–ª—É—á–∞–π–Ω–∞—è –≤–µ–ª–∏—á–∏–Ω–∞:

1. __–í—ã–±–æ—Ä–æ—á–Ω–æ–µ –°–ö–û__ —è–≤–ª—è–µ—Ç—Å—è –∫–æ—Ä–Ω–µ–º –∏–∑ –≤—ã–±–æ—Ä–æ—á–Ω–æ–π –¥–∏—Å–ø–µ—Ä—Å–∏–∏ –∏, –∫–∞–∫ –∏ –≤—ã–±–æ—Ä–æ—á–Ω–∞—è –¥–∏—Å–ø–µ—Ä—Å–∏—è, –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è–µ—Ç —Å–æ–±–æ–π —Å–º–µ—â–µ–Ω–Ω—É—é –æ—Ü–µ–Ω–∫—É —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–æ–≥–æ –°–ö–û.
2. __–ò—Å–ø—Ä–∞–≤–ª–µ–Ω–Ω–æ–µ –°–ö–û__ —è–≤–ª—è–µ—Ç—Å—è –∫–æ—Ä–Ω–µ–º –∏–∑ –∏—Å–ø—Ä–∞–≤–ª–µ–Ω–Ω–æ–π –¥–∏—Å–ø–µ—Ä—Å–∏–∏ –∏ –æ–±–µ—Å–ø–µ—á–∏–≤–∞–µ—Ç –±–æ–ª–µ–µ —Ç–æ—á–Ω—É—é –æ—Ü–µ–Ω–∫—É —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–æ–≥–æ –°–ö–û –¥–ª—è –º–∞–ª—ã—Ö –≤—ã–±–æ—Ä–æ–∫.
3. __–¢–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–æ–µ –°–ö–û__ –¥–ª—è –±–∏–Ω–æ–º–∏–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Ä–∞–≤–Ω–æ $\sqrt {(np(1-p))}$, —á—Ç–æ —Å–æ—Å—Ç–∞–≤–ª—è–µ—Ç –ø—Ä–∏–º–µ—Ä–Ω–æ 1.45 –¥–ª—è –Ω–∞—à–∏—Ö –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤.

–ù–∞–±–ª—é–¥–∞–µ–º—ã–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è –≤—ã–±–æ—Ä–æ—á–Ω–æ–≥–æ –∏ –∏—Å–ø—Ä–∞–≤–ª–µ–Ω–Ω–æ–≥–æ –°–ö–û –æ—Ç —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–æ–≥–æ –æ–±—ä—è—Å–Ω—è—é—Ç—Å—è —Å–ª—É—á–∞–π–Ω–æ—Å—Ç—å—é –≤—ã–±–æ—Ä–∫–∏. –ò—Å–ø—Ä–∞–≤–ª–µ–Ω–Ω–æ–µ –°–ö–û –æ–±—ã—á–Ω–æ –±–ª–∏–∂–µ –∫ —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–æ–º—É –∑–Ω–∞—á–µ–Ω–∏—é, —á—Ç–æ –ø–æ–¥—Ç–≤–µ—Ä–∂–¥–∞–µ—Ç –µ–≥–æ —Å—Ç–∞—Ç—É—Å –∫–∞–∫ –±–æ–ª–µ–µ –Ω–∞–¥–µ–∂–Ω–æ–π –æ—Ü–µ–Ω–∫–∏ –¥–ª—è –º–∞–ª—ã—Ö –≤—ã–±–æ—Ä–æ–∫.
"""))
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
from scipy.stats import norm
from IPython.display import display, Markdown, HTML

# –ó–∞–¥–∞–µ–º –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è
mu = 5  # –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ
sigma = 2  # —Å—Ä–µ–¥–Ω–µ–∫–≤–∞–¥—Ä–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ

# –°–æ–∑–¥–∞–µ–º –≤—ã–±–æ—Ä–∫—É –∏–∑ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è
np.random.seed(312)  # –¥–ª—è –≤–æ—Å–ø—Ä–æ–∏–∑–≤–æ–¥–∏–º–æ—Å—Ç–∏ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
n = 1000  # –æ–±—ä–µ–º –≤—ã–±–æ—Ä–∫–∏
sample = np.random.normal(mu, sigma, n)

print(f"–†–∞–∑–º–µ—Ä –≤—ã–±–æ—Ä–∫–∏: {len(sample)}")
print(f"–ù–µ—Å–∫–æ–ª—å–∫–æ –ø–µ—Ä–≤—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π: {sample[:5]}")

display(Markdown(r'''# –ó–∞–¥–∞–Ω–∏–µ 2
## 1. –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã —á–∞—Å—Ç–æ—Ç'''))

plt.figure(figsize=(10, 6))
hist, bins, _ = plt.hist(sample, bins=30, color='skyblue', edgecolor='black')
plt.title('–ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ —á–∞—Å—Ç–æ—Ç')
plt.xlabel('–ó–Ω–∞—á–µ–Ω–∏–µ')
plt.ylabel('–ß–∞—Å—Ç–æ—Ç–∞')
plt.grid(alpha=0.3)
plt.show()

print(f"–ú–∞–∫—Å–∏–º–∞–ª—å–Ω–∞—è —á–∞—Å—Ç–æ—Ç–∞: {max(hist)}")

display(Markdown(r'''## 2. –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω—ã—Ö —á–∞—Å—Ç–æ—Ç –∏ —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–æ–π –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è'''))

plt.figure(figsize=(10, 6))

# –ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω—ã—Ö —á–∞—Å—Ç–æ—Ç
plt.hist(sample, bins=30, density=True, color='skyblue', edgecolor='black', alpha=0.7, label='–≠–º–ø–∏—Ä–∏—á–µ—Å–∫–∞—è –ø–ª–æ—Ç–Ω–æ—Å—Ç—å')

# –¢–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∞—è –ø–ª–æ—Ç–Ω–æ—Å—Ç—å —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è
x = np.linspace(min(sample) - 2, max(sample) + 2, 1000)
plt.plot(x, stats.norm.pdf(x, mu, sigma), 'r-', linewidth=2, label='–¢–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∞—è –ø–ª–æ—Ç–Ω–æ—Å—Ç—å')

plt.title('–ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω—ã—Ö —á–∞—Å—Ç–æ—Ç –∏ —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∞—è –ø–ª–æ—Ç–Ω–æ—Å—Ç—å')
plt.xlabel('–ó–Ω–∞—á–µ–Ω–∏–µ')
plt.ylabel('–ü–ª–æ—Ç–Ω–æ—Å—Ç—å')
plt.legend()
plt.grid(alpha=0.3)
plt.show()

display(Markdown(r"""
### –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π –∫ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º:

–ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω—ã—Ö —á–∞—Å—Ç–æ—Ç –¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ —Ö–æ—Ä–æ—à–æ –∞–ø–ø—Ä–æ–∫—Å–∏–º–∏—Ä—É–µ—Ç —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫—É—é –ø–ª–æ—Ç–Ω–æ—Å—Ç—å –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è. –ú–æ–∂–Ω–æ –∑–∞–º–µ—Ç–∏—Ç—å —Ö–∞—Ä–∞–∫—Ç–µ—Ä–Ω—É—é –∫–æ–ª–æ–∫–æ–ª–æ–æ–±—Ä–∞–∑–Ω—É—é —Ñ–æ—Ä–º—É –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Å –ø–∏–∫–æ–º –≤ —Ä–∞–π–æ–Ω–µ –∑–Ω–∞—á–µ–Ω–∏—è Œº = 5. –ù–∞–±–ª—é–¥–∞—é—Ç—Å—è –Ω–µ–∫–æ—Ç–æ—Ä—ã–µ —Ñ–ª—É–∫—Ç—É–∞—Ü–∏–∏ –≤ —ç–º–ø–∏—Ä–∏—á–µ—Å–∫–æ–π –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏, —á—Ç–æ —è–≤–ª—è–µ—Ç—Å—è –µ—Å—Ç–µ—Å—Ç–≤–µ–Ω–Ω—ã–º —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏–º —à—É–º–æ–º, –ø—Ä–∏—Å—É—â–∏–º –ª—é–±–æ–π –≤—ã–±–æ—Ä–∫–µ –∫–æ–Ω–µ—á–Ω–æ–≥–æ –æ–±—ä–µ–º–∞. –û–¥–Ω–∞–∫–æ –≤ —Ü–µ–ª–æ–º —Ñ–æ—Ä–º–∞ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É–µ—Ç —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–æ–π –∫—Ä–∏–≤–æ–π, —á—Ç–æ –≤–∏–∑—É–∞–ª—å–Ω–æ –ø–æ–¥—Ç–≤–µ—Ä–∂–¥–∞–µ—Ç, —á—Ç–æ –≤—ã–±–æ—Ä–∫–∞ –¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ –≤–∑—è—Ç–∞ –∏–∑ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Å –∑–∞–¥–∞–Ω–Ω—ã–º–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏.

## 3. –ü–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ —ç–º–ø–∏—Ä–∏—á–µ—Å–∫–æ–π –∏ —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–æ–π —Ñ—É–Ω–∫—Ü–∏–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è
"""))

plt.figure(figsize=(10, 6))

# –≠–º–ø–∏—Ä–∏—á–µ—Å–∫–∞—è —Ñ—É–Ω–∫—Ü–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è
sorted_data = np.sort(sample)
ecdf = np.arange(1, len(sample) + 1) / len(sample)
plt.step(sorted_data, ecdf, label='–≠–º–ø–∏—Ä–∏—á–µ—Å–∫–∞—è —Ñ—É–Ω–∫—Ü–∏—è', where='post')

# –¢–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∞—è —Ñ—É–Ω–∫—Ü–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è
x = np.linspace(min(sample) - 2, max(sample) + 2, 1000)
plt.plot(x, stats.norm.cdf(x, mu, sigma), 'r-', linewidth=2, label='–¢–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∞—è —Ñ—É–Ω–∫—Ü–∏—è')

plt.title('–≠–º–ø–∏—Ä–∏—á–µ—Å–∫–∞—è –∏ —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∞—è —Ñ—É–Ω–∫—Ü–∏–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è')
plt.xlabel('–ó–Ω–∞—á–µ–Ω–∏–µ')
plt.ylabel('–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å')
plt.legend()
plt.grid(alpha=0.3)
plt.show()

display(Markdown(r"""
### –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π –∫ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º:

–≠–º–ø–∏—Ä–∏—á–µ—Å–∫–∞—è —Ñ—É–Ω–∫—Ü–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è (ECDF) –æ—á–µ–Ω—å —Ö–æ—Ä–æ—à–æ —Å–æ–≥–ª–∞—Å—É–µ—Ç—Å—è —Å —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–æ–π —Ñ—É–Ω–∫—Ü–∏–µ–π —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ –∑–∞–∫–æ–Ω–∞. –û–±–µ —Ñ—É–Ω–∫—Ü–∏–∏ –∏–º–µ—é—Ç —Ö–∞—Ä–∞–∫—Ç–µ—Ä–Ω—É—é S-–æ–±—Ä–∞–∑–Ω—É—é —Ñ–æ—Ä–º—É, –ø—Ä–∏—á–µ–º —ç–º–ø–∏—Ä–∏—á–µ—Å–∫–∞—è —Ñ—É–Ω–∫—Ü–∏—è –ø–ª–∞–≤–Ω–æ –ø—Ä–∏–±–ª–∏–∂–∞–µ—Ç—Å—è –∫ —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–æ–π —Å –Ω–µ–∫–æ—Ç–æ—Ä—ã–º–∏ –Ω–µ–∑–Ω–∞—á–∏—Ç–µ–ª—å–Ω—ã–º–∏ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è–º–∏. –û—Å–Ω–æ–≤–Ω—ã–µ —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫–∏ —Ñ—É–Ω–∫—Ü–∏–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Å–æ—Ö—Ä–∞–Ω—è—é—Ç—Å—è: –æ–Ω–∞ –º–æ–Ω–æ—Ç–æ–Ω–Ω–æ –≤–æ–∑—Ä–∞—Å—Ç–∞–µ—Ç –æ—Ç 0 –¥–æ 1, –∞ –≤ —Ç–æ—á–∫–µ, —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–µ–π –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–º—É –æ–∂–∏–¥–∞–Ω–∏—é (Œº = 5), –∑–Ω–∞—á–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏ –±–ª–∏–∑–∫–æ –∫ 0.5. –ë–ª–∏–∑–∫–æ–µ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–µ —ç–º–ø–∏—Ä–∏—á–µ—Å–∫–æ–π –∏ —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–æ–π —Ñ—É–Ω–∫—Ü–∏–π —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —è–≤–ª—è–µ—Ç—Å—è –µ—â–µ –æ–¥–Ω–∏–º –ø–æ–¥—Ç–≤–µ—Ä–∂–¥–µ–Ω–∏–µ–º —Ç–æ–≥–æ, —á—Ç–æ –≤—ã–±–æ—Ä–∫–∞ –¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É–µ—Ç –Ω–æ—Ä–º–∞–ª—å–Ω–æ–º—É —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—é —Å –∑–∞–¥–∞–Ω–Ω—ã–º–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏.

## 4. –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –≤—ã–±–æ—Ä–æ—á–Ω–æ–≥–æ —Å—Ä–µ–¥–Ω–µ–≥–æ –∏ –≤—ã–±–æ—Ä–æ—á–Ω–æ–π –º–µ–¥–∏–∞–Ω—ã
"""))

# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –≤—ã–±–æ—Ä–æ—á–Ω–æ–≥–æ —Å—Ä–µ–¥–Ω–µ–≥–æ
sample_mean = np.mean(sample)
print(f"–í—ã–±–æ—Ä–æ—á–Ω–æ–µ —Å—Ä–µ–¥–Ω–µ–µ: {sample_mean:.4f}")
print(f"–¢–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–æ–µ –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ: {mu:.4f}")
print(f"–†–∞–∑–Ω–∏—Ü–∞: {abs(sample_mean - mu):.4f}")

# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –≤—ã–±–æ—Ä–æ—á–Ω–æ–π –º–µ–¥–∏–∞–Ω—ã
sample_median = np.median(sample)
print(f"\n–í—ã–±–æ—Ä–æ—á–Ω–∞—è –º–µ–¥–∏–∞–Ω–∞: {sample_median:.4f}")
print(f"–¢–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∞—è –º–µ–¥–∏–∞–Ω–∞: {mu:.4f}")  # –î–ª—è –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –º–µ–¥–∏–∞–Ω–∞ —Ä–∞–≤–Ω–∞ –º–∞—Ç–æ–∂–∏–¥–∞–Ω–∏—é
print(f"–†–∞–∑–Ω–∏—Ü–∞: {abs(sample_median - mu):.4f}")

display(Markdown(r"""
### –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π –∫ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º:

–í—ã–±–æ—Ä–æ—á–Ω–æ–µ —Å—Ä–µ–¥–Ω–µ–µ –∏ –≤—ã–±–æ—Ä–æ—á–Ω–∞—è –º–µ–¥–∏–∞–Ω–∞ –æ–∫–∞–∑–∞–ª–∏—Å—å –æ—á–µ–Ω—å –±–ª–∏–∑–∫–∏–º–∏ –∫ —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏–º –∑–Ω–∞—á–µ–Ω–∏—è–º –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–≥–æ –æ–∂–∏–¥–∞–Ω–∏—è –∏ –º–µ–¥–∏–∞–Ω—ã –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è.

–î–ª—è –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ, –º–µ–¥–∏–∞–Ω–∞ –∏ –º–æ–¥–∞ —Å–æ–≤–ø–∞–¥–∞—é—Ç –∏ —Ä–∞–≤–Ω—ã Œº = 5. –ù–µ–∑–Ω–∞—á–∏—Ç–µ–ª—å–Ω—ã–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è –≤—ã–±–æ—Ä–æ—á–Ω—ã—Ö —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫ –æ—Ç —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏—Ö –∑–Ω–∞—á–µ–Ω–∏–π –æ–±—ä—è—Å–Ω—è—é—Ç—Å—è —Å–ª—É—á–∞–π–Ω—ã–º–∏ —Ñ–ª—É–∫—Ç—É–∞—Ü–∏—è–º–∏, –ø—Ä–∏—Å—É—â–∏–º–∏ –ª—é–±–æ–π –∫–æ–Ω–µ—á–Ω–æ–π –≤—ã–±–æ—Ä–∫–µ. –ú–∞–ª–æ—Å—Ç—å —ç—Ç–∏—Ö –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–π –≥–æ–≤–æ—Ä–∏—Ç –æ –≤—ã—Å–æ–∫–æ–π —Ä–µ–ø—Ä–µ–∑–µ–Ω—Ç–∞—Ç–∏–≤–Ω–æ—Å—Ç–∏ –≤—ã–±–æ—Ä–∫–∏ –∏ –ø–æ–¥—Ç–≤–µ—Ä–∂–¥–∞–µ—Ç —Ç–æ—Ç —Ñ–∞–∫—Ç, —á—Ç–æ –≤—ã–±–æ—Ä–∫–∞ –¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ –≤–∑—è—Ç–∞ –∏–∑ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Å –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–º Œº = 5.

–ë–ª–∏–∑–æ—Å—Ç—å –≤—ã–±–æ—Ä–æ—á–Ω–æ–≥–æ —Å—Ä–µ–¥–Ω–µ–≥–æ –∏ –≤—ã–±–æ—Ä–æ—á–Ω–æ–π –º–µ–¥–∏–∞–Ω—ã –¥—Ä—É–≥ –∫ –¥—Ä—É–≥—É —Ç–∞–∫–∂–µ —è–≤–ª—è–µ—Ç—Å—è —Ö–∞—Ä–∞–∫—Ç–µ—Ä–Ω—ã–º –ø—Ä–∏–∑–Ω–∞–∫–æ–º —Å–∏–º–º–µ—Ç—Ä–∏—á–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è, –∫–∞–∫–∏–º –∏ —è–≤–ª—è–µ—Ç—Å—è –Ω–æ—Ä–º–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ.

## 5. –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –≤—ã–±–æ—Ä–æ—á–Ω–æ–π –¥–∏—Å–ø–µ—Ä—Å–∏–∏ –∏ –∏—Å–ø—Ä–∞–≤–ª–µ–Ω–Ω–æ–π –¥–∏—Å–ø–µ—Ä—Å–∏–∏
"""))

# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –≤—ã–±–æ—Ä–æ—á–Ω–æ–π –¥–∏—Å–ø–µ—Ä—Å–∏–∏
sample_var = np.var(sample)
print(f"–í—ã–±–æ—Ä–æ—á–Ω–∞—è –¥–∏—Å–ø–µ—Ä—Å–∏—è: {sample_var:.4f}")

# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –∏—Å–ø—Ä–∞–≤–ª–µ–Ω–Ω–æ–π –¥–∏—Å–ø–µ—Ä—Å–∏–∏
corrected_var = np.var(sample, ddof=1)
print(f"–ò—Å–ø—Ä–∞–≤–ª–µ–Ω–Ω–∞—è –¥–∏—Å–ø–µ—Ä—Å–∏—è: {corrected_var:.4f}")

# –¢–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∞—è –¥–∏—Å–ø–µ—Ä—Å–∏—è
theoretical_var = sigma**2
print(f"–¢–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∞—è –¥–∏—Å–ø–µ—Ä—Å–∏—è: {theoretical_var:.4f}")

print(f"–†–∞–∑–Ω–∏—Ü–∞ –º–µ–∂–¥—É –≤—ã–±–æ—Ä–æ—á–Ω–æ–π –∏ —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–æ–π: {abs(sample_var - theoretical_var):.4f}")
print(f"–†–∞–∑–Ω–∏—Ü–∞ –º–µ–∂–¥—É –∏—Å–ø—Ä–∞–≤–ª–µ–Ω–Ω–æ–π –∏ —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–æ–π: {abs(corrected_var - theoretical_var):.4f}")

display(Markdown(r"""
### –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π –∫ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º:

–í—ã–±–æ—Ä–æ—á–Ω–∞—è –¥–∏—Å–ø–µ—Ä—Å–∏—è, –≤—ã—á–∏—Å–ª–µ–Ω–Ω–∞—è –∫–∞–∫ —Å—Ä–µ–¥–Ω–∏–π –∫–≤–∞–¥—Ä–∞—Ç –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–π –æ—Ç –≤—ã–±–æ—Ä–æ—á–Ω–æ–≥–æ —Å—Ä–µ–¥–Ω–µ–≥–æ, –¥–∞–µ—Ç –æ—Ü–µ–Ω–∫—É –¥–∏—Å–ø–µ—Ä—Å–∏–∏ –≥–µ–Ω–µ—Ä–∞–ª—å–Ω–æ–π —Å–æ–≤–æ–∫—É–ø–Ω–æ—Å—Ç–∏, –Ω–æ —ç—Ç–∞ –æ—Ü–µ–Ω–∫–∞ —è–≤–ª—è–µ—Ç—Å—è —Å–º–µ—â–µ–Ω–Ω–æ–π. –ò—Å–ø—Ä–∞–≤–ª–µ–Ω–Ω–∞—è –¥–∏—Å–ø–µ—Ä—Å–∏—è, –≥–¥–µ —Å—É–º–º–∞ –∫–≤–∞–¥—Ä–∞—Ç–æ–≤ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–π –¥–µ–ª–∏—Ç—Å—è –Ω–∞ (n-1), –∞ –Ω–µ –Ω–∞ n, —è–≤–ª—è–µ—Ç—Å—è –Ω–µ—Å–º–µ—â–µ–Ω–Ω–æ–π –æ—Ü–µ–Ω–∫–æ–π –¥–∏—Å–ø–µ—Ä—Å–∏–∏ –≥–µ–Ω–µ—Ä–∞–ª—å–Ω–æ–π —Å–æ–≤–æ–∫—É–ø–Ω–æ—Å—Ç–∏.

–¢–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∞—è –¥–∏—Å–ø–µ—Ä—Å–∏—è –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Ä–∞–≤–Ω–∞ œÉ¬≤ = 4 (—Ç–∞–∫ –∫–∞–∫ œÉ = 2). –ö–∞–∫ –≤–∏–¥–Ω–æ –∏–∑ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤, –∏—Å–ø—Ä–∞–≤–ª–µ–Ω–Ω–∞—è –¥–∏—Å–ø–µ—Ä—Å–∏—è –±–ª–∏–∂–µ –∫ —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–æ–º—É –∑–Ω–∞—á–µ–Ω–∏—é, —á–µ–º –≤—ã–±–æ—Ä–æ—á–Ω–∞—è –¥–∏—Å–ø–µ—Ä—Å–∏—è. –≠—Ç–æ –æ–∂–∏–¥–∞–µ–º–æ, –ø–æ—Å–∫–æ–ª—å–∫—É –∏—Å–ø—Ä–∞–≤–ª–µ–Ω–Ω–∞—è –¥–∏—Å–ø–µ—Ä—Å–∏—è —è–≤–ª—è–µ—Ç—Å—è –Ω–µ—Å–º–µ—â–µ–Ω–Ω–æ–π –æ—Ü–µ–Ω–∫–æ–π –¥–∏—Å–ø–µ—Ä—Å–∏–∏ –≥–µ–Ω–µ—Ä–∞–ª—å–Ω–æ–π —Å–æ–≤–æ–∫—É–ø–Ω–æ—Å—Ç–∏.

–†–∞–∑–Ω–∏—Ü–∞ –º–µ–∂–¥—É –∏—Å–ø—Ä–∞–≤–ª–µ–Ω–Ω–æ–π –¥–∏—Å–ø–µ—Ä—Å–∏–µ–π –∏ —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–æ–π –¥–∏—Å–ø–µ—Ä—Å–∏–µ–π —è–≤–ª—è–µ—Ç—Å—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–º —Å–ª—É—á–∞–π–Ω—ã—Ö —Ñ–ª—É–∫—Ç—É–∞—Ü–∏–π –≤ –≤—ã–±–æ—Ä–∫–µ. –ü—Ä–∏ —É–≤–µ–ª–∏—á–µ–Ω–∏–∏ –æ–±—ä–µ–º–∞ –≤—ã–±–æ—Ä–∫–∏ —ç—Ç–∞ —Ä–∞–∑–Ω–∏—Ü–∞ –±—É–¥–µ—Ç —Å—Ç—Ä–µ–º–∏—Ç—å—Å—è –∫ –Ω—É–ª—é.

## 6. –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –≤—ã–±–æ—Ä–æ—á–Ω–æ–≥–æ –∏ –∏—Å–ø—Ä–∞–≤–ª–µ–Ω–Ω–æ–≥–æ —Å—Ä–µ–¥–Ω–µ–≥–æ –∫–≤–∞–¥—Ä–∞—Ç–∏—á–µ—Å–∫–æ–≥–æ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è
"""))

# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –≤—ã–±–æ—Ä–æ—á–Ω–æ–≥–æ —Å—Ä–µ–¥–Ω–µ–≥–æ –∫–≤–∞–¥—Ä–∞—Ç–∏—á–µ—Å–∫–æ–≥–æ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è
sample_std = np.std(sample)
print(f"–í—ã–±–æ—Ä–æ—á–Ω–æ–µ —Å—Ä–µ–¥–Ω–µ–µ –∫–≤–∞–¥—Ä–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ: {sample_std:.4f}")

# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –∏—Å–ø—Ä–∞–≤–ª–µ–Ω–Ω–æ–≥–æ —Å—Ä–µ–¥–Ω–µ–≥–æ –∫–≤–∞–¥—Ä–∞—Ç–∏—á–µ—Å–∫–æ–≥–æ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è
corrected_std = np.std(sample, ddof=1)
print(f"–ò—Å–ø—Ä–∞–≤–ª–µ–Ω–Ω–æ–µ —Å—Ä–µ–¥–Ω–µ–µ –∫–≤–∞–¥—Ä–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ: {corrected_std:.4f}")

# –¢–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–æ–µ —Å—Ä–µ–¥–Ω–µ–µ –∫–≤–∞–¥—Ä–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ
theoretical_std = sigma
print(f"–¢–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–æ–µ —Å—Ä–µ–¥–Ω–µ–µ –∫–≤–∞–¥—Ä–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ: {theoretical_std:.4f}")

print(f"–†–∞–∑–Ω–∏—Ü–∞ –º–µ–∂–¥—É –≤—ã–±–æ—Ä–æ—á–Ω—ã–º –∏ —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏–º: {abs(sample_std - theoretical_std):.4f}")
print(f"–†–∞–∑–Ω–∏—Ü–∞ –º–µ–∂–¥—É –∏—Å–ø—Ä–∞–≤–ª–µ–Ω–Ω—ã–º –∏ —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏–º: {abs(corrected_std - theoretical_std):.4f}")

display(Markdown(r"""
### –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π –∫ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º:

–°—Ä–µ–¥–Ω–µ–µ –∫–≤–∞–¥—Ä–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ (–°–ö–û) —è–≤–ª—è–µ—Ç—Å—è –∫–≤–∞–¥—Ä–∞—Ç–Ω—ã–º –∫–æ—Ä–Ω–µ–º –∏–∑ –¥–∏—Å–ø–µ—Ä—Å–∏–∏ –∏ –∏–∑–º–µ—Ä—è–µ—Ç—Å—è –≤ —Ç–µ—Ö –∂–µ –µ–¥–∏–Ω–∏—Ü–∞—Ö, —á—Ç–æ –∏ —Å–∞–º–∞ —Å–ª—É—á–∞–π–Ω–∞—è –≤–µ–ª–∏—á–∏–Ω–∞. –í—ã–±–æ—Ä–æ—á–Ω–æ–µ –°–ö–û, –∫–∞–∫ –∏ –≤—ã–±–æ—Ä–æ—á–Ω–∞—è –¥–∏—Å–ø–µ—Ä—Å–∏—è, —è–≤–ª—è–µ—Ç—Å—è —Å–º–µ—â–µ–Ω–Ω–æ–π –æ—Ü–µ–Ω–∫–æ–π –°–ö–û –≥–µ–Ω–µ—Ä–∞–ª—å–Ω–æ–π —Å–æ–≤–æ–∫—É–ø–Ω–æ—Å—Ç–∏. –ò—Å–ø—Ä–∞–≤–ª–µ–Ω–Ω–æ–µ –°–ö–û, –≤—ã—á–∏—Å–ª—è–µ–º–æ–µ –∫–∞–∫ –∫–≤–∞–¥—Ä–∞—Ç–Ω—ã–π –∫–æ—Ä–µ–Ω—å –∏–∑ –∏—Å–ø—Ä–∞–≤–ª–µ–Ω–Ω–æ–π –¥–∏—Å–ø–µ—Ä—Å–∏–∏, —è–≤–ª—è–µ—Ç—Å—è –Ω–µ—Å–º–µ—â–µ–Ω–Ω–æ–π –æ—Ü–µ–Ω–∫–æ–π –°–ö–û –≥–µ–Ω–µ—Ä–∞–ª—å–Ω–æ–π —Å–æ–≤–æ–∫—É–ø–Ω–æ—Å—Ç–∏.

–¢–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–æ–µ –°–ö–û –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Ä–∞–≤–Ω–æ œÉ = 2. –ö–∞–∫ –≤–∏–¥–Ω–æ –∏–∑ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤, –∏—Å–ø—Ä–∞–≤–ª–µ–Ω–Ω–æ–µ –°–ö–û –±–ª–∏–∂–µ –∫ —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–æ–º—É –∑–Ω–∞—á–µ–Ω–∏—é, —á–µ–º –≤—ã–±–æ—Ä–æ—á–Ω–æ–µ –°–ö–û. –≠—Ç–æ –∞–Ω–∞–ª–æ–≥–∏—á–Ω–æ —Å–∏—Ç—É–∞—Ü–∏–∏ —Å –¥–∏—Å–ø–µ—Ä—Å–∏–µ–π –∏ –æ–±—ä—è—Å–Ω—è–µ—Ç—Å—è —Ç–µ–º –∂–µ —Å–≤–æ–π—Å—Ç–≤–æ–º —Å–º–µ—â–µ–Ω–Ω–æ—Å—Ç–∏/–Ω–µ—Å–º–µ—â–µ–Ω–Ω–æ—Å—Ç–∏ –æ—Ü–µ–Ω–æ–∫.

–†–∞–∑–Ω–∏—Ü–∞ –º–µ–∂–¥—É –∏—Å–ø—Ä–∞–≤–ª–µ–Ω–Ω—ã–º –°–ö–û –∏ —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏–º –°–ö–û, –∫–∞–∫ –∏ –≤ —Å–ª—É—á–∞–µ —Å –¥–∏—Å–ø–µ—Ä—Å–∏–µ–π, —è–≤–ª—è–µ—Ç—Å—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–º —Å–ª—É—á–∞–π–Ω—ã—Ö —Ñ–ª—É–∫—Ç—É–∞—Ü–∏–π –≤ –≤—ã–±–æ—Ä–∫–µ. –ü—Ä–∏ —É–≤–µ–ª–∏—á–µ–Ω–∏–∏ –æ–±—ä–µ–º–∞ –≤—ã–±–æ—Ä–∫–∏ —ç—Ç–∞ —Ä–∞–∑–Ω–∏—Ü–∞ –±—É–¥–µ—Ç —Å—Ç—Ä–µ–º–∏—Ç—å—Å—è –∫ –Ω—É–ª—é.

### –ó–∞–∫–ª—é—á–µ–Ω–∏–µ
–ü—Ä–æ–≤–µ–¥–µ–Ω–Ω–æ–µ –∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏–µ –≤—ã–±–æ—Ä–∫–∏ –∏–∑ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è —Å –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏ Œº = 5 –∏ œÉ = 2 –ø–æ–∫–∞–∑–∞–ª–æ —Ö–æ—Ä–æ—à–µ–µ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–µ —ç–º–ø–∏—Ä–∏—á–µ—Å–∫–∏—Ö –∏ —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏—Ö —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫. –ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω—ã—Ö —á–∞—Å—Ç–æ—Ç –∏ —ç–º–ø–∏—Ä–∏—á–µ—Å–∫–∞—è —Ñ—É–Ω–∫—Ü–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –≤–∏–∑—É–∞–ª—å–Ω–æ –±–ª–∏–∑–∫–∏ –∫ —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏–º –∞–Ω–∞–ª–æ–≥–∞–º. –í—ã–±–æ—Ä–æ—á–Ω—ã–µ —á–∏—Å–ª–æ–≤—ã–µ —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫–∏ (—Å—Ä–µ–¥–Ω–µ–µ, –º–µ–¥–∏–∞–Ω–∞, –¥–∏—Å–ø–µ—Ä—Å–∏—è, –°–ö–û) –æ—á–µ–Ω—å –±–ª–∏–∑–∫–∏ –∫ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏–º —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏–º –∑–Ω–∞—á–µ–Ω–∏—è–º, —á—Ç–æ –ø–æ–¥—Ç–≤–µ—Ä–∂–¥–∞–µ—Ç –≤—ã—Å–æ–∫—É—é —Ä–µ–ø—Ä–µ–∑–µ–Ω—Ç–∞—Ç–∏–≤–Ω–æ—Å—Ç—å –≤—ã–±–æ—Ä–∫–∏. –ò—Å–ø—Ä–∞–≤–ª–µ–Ω–Ω—ã–µ –æ—Ü–µ–Ω–∫–∏ –¥–∏—Å–ø–µ—Ä—Å–∏–∏ –∏ –°–ö–û –æ–∫–∞–∑–∞–ª–∏—Å—å –±–ª–∏–∂–µ –∫ —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–∏–º –∑–Ω–∞—á–µ–Ω–∏—è–º, —á–µ–º –æ–±—ã—á–Ω—ã–µ –≤—ã–±–æ—Ä–æ—á–Ω—ã–µ –æ—Ü–µ–Ω–∫–∏, —á—Ç–æ –ø–æ–¥—Ç–≤–µ—Ä–∂–¥–∞–µ—Ç —Ü–µ–ª–µ—Å–æ–æ–±—Ä–∞–∑–Ω–æ—Å—Ç—å –∏—Ö –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è –ø—Ä–∏ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–º –∞–Ω–∞–ª–∏–∑–µ.
"""))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display, Markdown, HTML

# –°–æ–∑–¥–∞–µ–º DataFrame –∏–∑ –ø—Ä–µ–¥–æ—Å—Ç–∞–≤–ª–µ–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö
data = pd.read_csv('–í–∞—Ä–∏–∞–Ω—Ç 26.3.csv')

# –ü—Ä–æ—Å–º–æ—Ç—Ä –ø–µ—Ä–≤—ã—Ö –∑–∞–ø–∏—Å–µ–π
print("–ü–µ—Ä–≤—ã–µ 5 –∑–∞–ø–∏—Å–µ–π:")
display(data.head())

# –û–±—â–∞—è –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ –Ω–∞–±–æ—Ä–µ –¥–∞–Ω–Ω—ã—Ö
display(Markdown("### –û–±—â–∞—è –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—è –æ –Ω–∞–±–æ—Ä–µ –¥–∞–Ω–Ω—ã—Ö:"))
display(data.info())

# –°—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–µ –æ–ø–∏—Å–∞–Ω–∏–µ —á–∏—Å–ª–æ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
display(Markdown("### –°—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–µ –æ–ø–∏—Å–∞–Ω–∏–µ —á–∏—Å–ª–æ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤:"))
display(data.describe())

display(Markdown(r"""
### 2. –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Ç–∏–ø–æ–≤ –¥–∞–Ω–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
#### Age (–≤–æ–∑—Ä–∞—Å—Ç)

–¢–∏–ø –¥–∞–Ω–Ω—ã—Ö: –ö–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω—ã–π, –¥–∏—Å–∫—Ä–µ—Ç–Ω—ã–π

–û–±–æ—Å–Ω–æ–≤–∞–Ω–∏–µ: –í–æ–∑—Ä–∞—Å—Ç –∏–∑–º–µ—Ä—è–µ—Ç—Å—è –≤ —Ü–µ–ª—ã—Ö –≥–æ–¥–∞—Ö –∏ –ø—Ä–∏–Ω–∏–º–∞–µ—Ç –∫–æ–Ω–µ—á–Ω–æ–µ –º–Ω–æ–∂–µ—Å—Ç–≤–æ –∑–Ω–∞—á–µ–Ω–∏–π. –≠—Ç–æ —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫–∞, –∫–æ—Ç–æ—Ä—É—é –º–æ–∂–Ω–æ –∏–∑–º–µ—Ä–∏—Ç—å —á–∏—Å–ª–æ–º –∏ –Ω–∞–¥ –∫–æ—Ç–æ—Ä–æ–π –¥–æ–ø—É—Å—Ç–∏–º—ã –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–∏–µ –æ–ø–µ—Ä–∞—Ü–∏–∏.

#### Weight (–≤–µ—Å)
–¢–∏–ø –¥–∞–Ω–Ω—ã—Ö: –ö–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω—ã–π, –Ω–µ–ø—Ä–µ—Ä—ã–≤–Ω—ã–π

–û–±–æ—Å–Ω–æ–≤–∞–Ω–∏–µ: –í–µ—Å –∏–∑–º–µ—Ä—è–µ—Ç—Å—è –≤ —á–∏—Å–ª–æ–≤–æ–π —à–∫–∞–ª–µ —Å –¥–µ—Å—è—Ç–∏—á–Ω—ã–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏, –≥–¥–µ –≤–æ–∑–º–æ–∂–Ω—ã –ª—é–±—ã–µ –ø—Ä–æ–º–µ–∂—É—Ç–æ—á–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è. –ù–∞–¥ –≤–µ—Å–æ–º –¥–æ–ø—É—Å—Ç–∏–º—ã –≤—Å–µ –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–∏–µ –æ–ø–µ—Ä–∞—Ü–∏–∏.

#### CAEC (–ø–µ—Ä–µ–∫—É—Å—ã –º–µ–∂–¥—É –æ—Å–Ω–æ–≤–Ω—ã–º–∏ –ø—Ä–∏–µ–º–∞–º–∏ –ø–∏—â–∏)
–¢–∏–ø –¥–∞–Ω–Ω—ã—Ö: –ö–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–π, –ø–æ—Ä—è–¥–∫–æ–≤—ã–π

–û–±–æ—Å–Ω–æ–≤–∞–Ω–∏–µ: –î–∞–Ω–Ω—ã–π –ø—Ä–∏–∑–Ω–∞–∫ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ "Always", "Frequently", "Sometimes", "Never", –∫–æ—Ç–æ—Ä—ã–µ –æ—Ç—Ä–∞–∂–∞—é—Ç —á–∞—Å—Ç–æ—Ç—É –¥–µ–π—Å—Ç–≤–∏—è –∏ –∏–º–µ—é—Ç –µ—Å—Ç–µ—Å—Ç–≤–µ–Ω–Ω—ã–π –ø–æ—Ä—è–¥–æ–∫ (–æ—Ç "–Ω–∏–∫–æ–≥–¥–∞" –¥–æ "–≤—Å–µ–≥–¥–∞").

#### FAF (–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –¥–Ω–µ–π –≤ –Ω–µ–¥–µ–ª—é —Å —Ñ–∏–∑–∏—á–µ—Å–∫–æ–π –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç—å—é)
–¢–∏–ø –¥–∞–Ω–Ω—ã—Ö: –ö–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω—ã–π, –¥–∏—Å–∫—Ä–µ—Ç–Ω—ã–π

–û–±–æ—Å–Ω–æ–≤–∞–Ω–∏–µ: –≠—Ç–æ—Ç –ø—Ä–∏–∑–Ω–∞–∫ –ø—Ä–∏–Ω–∏–º–∞–µ—Ç —Ü–µ–ª–æ—á–∏—Å–ª–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –æ—Ç 0 –¥–æ 7, –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è—è –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –¥–Ω–µ–π –≤ –Ω–µ–¥–µ–ª—é. –ó–Ω–∞—á–µ–Ω–∏—è —Å—á–µ—Ç–Ω—ã–µ –∏ –∏–º–µ—é—Ç —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ –∏–Ω—Ç–µ—Ä–≤–∞–ª—ã.

#### MTRANS (–≤–∏–¥ —Ç—Ä–∞–Ω—Å–ø–æ—Ä—Ç–∞)
–¢–∏–ø –¥–∞–Ω–Ω—ã—Ö: –ö–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–π, –Ω–æ–º–∏–Ω–∞–ª—å–Ω—ã–π

–û–±–æ—Å–Ω–æ–≤–∞–Ω–∏–µ: –ü—Ä–µ–¥—Å—Ç–∞–≤–ª—è–µ—Ç —Å–æ–±–æ–π –≤–∏–¥—ã —Ç—Ä–∞–Ω—Å–ø–æ—Ä—Ç–∞ (Public_Transportation, Automobile, Walking, Motorbike), –∫–æ—Ç–æ—Ä—ã–µ –Ω–µ –∏–º–µ—é—Ç –µ—Å—Ç–µ—Å—Ç–≤–µ–Ω–Ω–æ–≥–æ –ø–æ—Ä—è–¥–∫–∞ –∏–ª–∏ –∏–µ—Ä–∞—Ä—Ö–∏–∏.

### 3. –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
"""))

# –ù–∞—Å—Ç—Ä–æ–π–∫–∞ —Å—Ç–∏–ª—è –≥—Ä–∞—Ñ–∏–∫–æ–≤
#print(plt.style.available)
plt.style.use('seaborn-v0_8-whitegrid')
fig, axes = plt.subplots(2, 3, figsize=(20, 12))

# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –≤–æ–∑—Ä–∞—Å—Ç–∞ (Age)
sns.histplot(data=data, x='Age', kde=True, ax=axes[0, 0])
axes[0, 0].set_title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –≤–æ–∑—Ä–∞—Å—Ç–∞', fontsize=14)
axes[0, 0].set_xlabel('–í–æ–∑—Ä–∞—Å—Ç (–ª–µ—Ç)', fontsize=12)
axes[0, 0].set_ylabel('–ß–∞—Å—Ç–æ—Ç–∞', fontsize=12)

# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –≤–µ—Å–∞ (Weight)
sns.histplot(data=data, x='Weight', kde=True, ax=axes[0, 1])
axes[0, 1].set_title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –≤–µ—Å–∞', fontsize=14)
axes[0, 1].set_xlabel('–í–µ—Å (–∫–≥)', fontsize=12)
axes[0, 1].set_ylabel('–ß–∞—Å—Ç–æ—Ç–∞', fontsize=12)

# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –ø–µ—Ä–µ–∫—É—Å–æ–≤ (CAEC)
caec_order = ['Never', 'Sometimes', 'Frequently', 'Always']
sns.countplot(data=data, x='CAEC', order=caec_order, ax=axes[0, 2])
axes[0, 2].set_title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —á–∞—Å—Ç–æ—Ç—ã –ø–µ—Ä–µ–∫—É—Å–æ–≤', fontsize=14)
axes[0, 2].set_xlabel('–ß–∞—Å—Ç–æ—Ç–∞ –ø–µ—Ä–µ–∫—É—Å–æ–≤', fontsize=12)
axes[0, 2].set_ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ', fontsize=12)
axes[0, 2].tick_params(axis='x', rotation=45)

# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —Ñ–∏–∑–∏—á–µ—Å–∫–æ–π –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç–∏ (FAF)
sns.countplot(data=data, x='FAF', ax=axes[1, 0])
axes[1, 0].set_title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –¥–Ω–µ–π —Ñ–∏–∑–∏—á–µ—Å–∫–æ–π –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç–∏', fontsize=14)
axes[1, 0].set_xlabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –¥–Ω–µ–π –≤ –Ω–µ–¥–µ–ª—é', fontsize=12)
axes[1, 0].set_ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ', fontsize=12)

# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —Ç—Ä–∞–Ω—Å–ø–æ—Ä—Ç–∞ (MTRANS)
sns.countplot(data=data, x='MTRANS', ax=axes[1, 1])
axes[1, 1].set_title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –≤–∏–¥–æ–≤ —Ç—Ä–∞–Ω—Å–ø–æ—Ä—Ç–∞', fontsize=14)
axes[1, 1].set_xlabel('–í–∏–¥ —Ç—Ä–∞–Ω—Å–ø–æ—Ä—Ç–∞', fontsize=12)
axes[1, 1].set_ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ', fontsize=12)
axes[1, 1].tick_params(axis='x', rotation=45)

# –£–¥–∞–ª–µ–Ω–∏–µ –ª–∏—à–Ω–µ–≥–æ –≥—Ä–∞—Ñ–∏–∫–∞
axes[1, 2].axis('off')

plt.tight_layout()
plt.show()

display(Markdown(r"""### 4. –í—ã—á–∏—Å–ª–µ–Ω–∏–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏—Ö –æ—Ü–µ–Ω–æ–∫ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞"""))

# –°–æ–∑–¥–∞–¥–∏–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –ø–æ–ª—É—á–µ–Ω–∏—è —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫ –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç —Ç–∏–ø–∞ –ø—Ä–∏–∑–Ω–∞–∫–∞
def get_statistics(data, column_name, column_type):
    stats = {}
    
    # –ú–æ–¥–∞ –¥–ª—è –≤—Å–µ—Ö —Ç–∏–ø–æ–≤ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
    stats['–ú–æ–¥–∞'] = data[column_name].mode()[0]
    
    if column_type in ['–∫–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω—ã–π_–¥–∏—Å–∫—Ä–µ—Ç–Ω—ã–π', '–∫–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω—ã–π_–Ω–µ–ø—Ä–µ—Ä—ã–≤–Ω—ã–π']:
        # –ö–≤–∞—Ä—Ç–∏–ª–∏, –º–µ–¥–∏–∞–Ω–∞, —Å—Ä–µ–¥–Ω–µ–µ, –¥–∏—Å–ø–µ—Ä—Å–∏—è –∏ –°–ö–û –¥–ª—è –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
        stats['–ü–µ—Ä–≤—ã–π –∫–≤–∞—Ä—Ç–∏–ª—å'] = data[column_name].quantile(0.25)
        stats['–ú–µ–¥–∏–∞–Ω–∞'] = data[column_name].median()
        stats['–¢—Ä–µ—Ç–∏–π –∫–≤–∞—Ä—Ç–∏–ª—å'] = data[column_name].quantile(0.75)
        stats['–°—Ä–µ–¥–Ω–µ–µ'] = data[column_name].mean()
        stats['–î–∏—Å–ø–µ—Ä—Å–∏—è'] = data[column_name].var()
        stats['–ò—Å–ø—Ä–∞–≤–ª–µ–Ω–Ω–∞—è –¥–∏—Å–ø–µ—Ä—Å–∏—è'] = data[column_name].var(ddof=1)
        stats['–°–ö–û'] = data[column_name].std()
        stats['–ò—Å–ø—Ä–∞–≤–ª–µ–Ω–Ω–æ–µ –°–ö–û'] = data[column_name].std(ddof=1)
    
    elif column_type == '–∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–π_–ø–æ—Ä—è–¥–∫–æ–≤—ã–π':
        # –î–ª—è –ø–æ—Ä—è–¥–∫–æ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –º–æ–∂–Ω–æ –≤—ã—á–∏—Å–ª–∏—Ç—å –º–µ–¥–∏–∞–Ω—É –∏ –∫–≤–∞—Ä—Ç–∏–ª–∏, 
        # –µ—Å–ª–∏ –æ–ø—Ä–µ–¥–µ–ª–∏—Ç—å –ø–æ—Ä—è–¥–æ–∫ –∫–∞—Ç–µ–≥–æ—Ä–∏–π
        if column_name == 'CAEC':
            caec_order = {'Never': 0, 'Sometimes': 1, 'Frequently': 2, 'Always': 3}
            data_numeric = data[column_name].map(caec_order)
            stats['–ü–µ—Ä–≤—ã–π –∫–≤–∞—Ä—Ç–∏–ª—å'] = data_numeric.quantile(0.25)
            stats['–ú–µ–¥–∏–∞–Ω–∞'] = data_numeric.median()
            stats['–¢—Ä–µ—Ç–∏–π –∫–≤–∞—Ä—Ç–∏–ª—å'] = data_numeric.quantile(0.75)
    
    return stats

# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞
age_stats = get_statistics(data, 'Age', '–∫–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω—ã–π_–¥–∏—Å–∫—Ä–µ—Ç–Ω—ã–π')
weight_stats = get_statistics(data, 'Weight', '–∫–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω—ã–π_–Ω–µ–ø—Ä–µ—Ä—ã–≤–Ω—ã–π')
caec_stats = get_statistics(data, 'CAEC', '–∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–π_–ø–æ—Ä—è–¥–∫–æ–≤—ã–π')
faf_stats = get_statistics(data, 'FAF', '–∫–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω—ã–π_–¥–∏—Å–∫—Ä–µ—Ç–Ω—ã–π')
mtrans_stats = get_statistics(data, 'MTRANS', '–∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–π_–Ω–æ–º–∏–Ω–∞–ª—å–Ω—ã–π')

# –í—ã–≤–æ–¥ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
print("–°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏ –¥–ª—è –≤–æ–∑—Ä–∞—Å—Ç–∞ (Age):")
for key, value in age_stats.items():
    print(f"{key}: {value}")

print("\n–°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏ –¥–ª—è –≤–µ—Å–∞ (Weight):")
for key, value in weight_stats.items():
    print(f"{key}: {value}")

print("\n–°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏ –¥–ª—è –ø–µ—Ä–µ–∫—É—Å–æ–≤ (CAEC):")
for key, value in caec_stats.items():
    print(f"{key}: {value}")

print("\n–°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏ –¥–ª—è —Ñ–∏–∑–∏—á–µ—Å–∫–æ–π –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç–∏ (FAF):")
for key, value in faf_stats.items():
    print(f"{key}: {value}")

print("\n–°—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏ –¥–ª—è —Ç—Ä–∞–Ω—Å–ø–æ—Ä—Ç–∞ (MTRANS):")
for key, value in mtrans_stats.items():
    print(f"{key}: {value}")

display(Markdown(r"""
### 5. –ò–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è –≤—ã–±–æ—Ä–∞ –æ—Ü–µ–Ω–æ–∫ –∏ –ø–æ–ª—É—á–µ–Ω–Ω—ã—Ö —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤

#### Age (–≤–æ–∑—Ä–∞—Å—Ç)

–í—ã–±–æ—Ä –æ—Ü–µ–Ω–æ–∫: –î–ª—è –≤–æ–∑—Ä–∞—Å—Ç–∞, –∫–∞–∫ –¥–ª—è –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω–æ–≥–æ –¥–∏—Å–∫—Ä–µ—Ç–Ω–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞, –ø—Ä–∏–º–µ–Ω–∏–º—ã –≤—Å–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏–µ –æ—Ü–µ–Ω–∫–∏: –º–æ–¥–∞, –∫–≤–∞—Ä—Ç–∏–ª–∏, –º–µ–¥–∏–∞–Ω–∞, —Å—Ä–µ–¥–Ω–µ–µ, –¥–∏—Å–ø–µ—Ä—Å–∏—è –∏ –°–ö–û.

__–ò–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤__:

–ú–æ–¥–∞ –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç –Ω–∞–∏–±–æ–ª–µ–µ —á–∞—Å—Ç–æ –≤—Å—Ç—Ä–µ—á–∞—é—â–∏–π—Å—è –≤–æ–∑—Ä–∞—Å—Ç –≤ –≤—ã–±–æ—Ä–∫–µ.

–ö–≤–∞—Ä—Ç–∏–ª–∏ –∏ –º–µ–¥–∏–∞–Ω–∞ –æ–ø–∏—Å—ã–≤–∞—é—Ç —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –≤–æ–∑—Ä–∞—Å—Ç–æ–≤.

–°—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –æ—Ç—Ä–∞–∂–∞–µ—Ç —Å—Ä–µ–¥–Ω–∏–π –≤–æ–∑—Ä–∞—Å—Ç —É—á–∞—Å—Ç–Ω–∏–∫–æ–≤.

–î–∏—Å–ø–µ—Ä—Å–∏—è –∏ –°–ö–û –æ—Ç—Ä–∞–∂–∞—é—Ç —Ä–∞–∑–±—Ä–æ—Å –≤–æ–∑—Ä–∞—Å—Ç–æ–≤ –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ —Å—Ä–µ–¥–Ω–µ–≥–æ.

–ò—Å—Ö–æ–¥—è –∏–∑ –¥–∞–Ω–Ω—ã—Ö, –≤–∏–¥–Ω–æ, —á—Ç–æ –≤—ã–±–æ—Ä–∫–∞ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∞ –ø—Ä–µ–∏–º—É—â–µ—Å—Ç–≤–µ–Ω–Ω–æ –º–æ–ª–æ–¥—ã–º–∏ –ª—é–¥—å–º–∏, —Å –±–æ–ª—å—à–∏–Ω—Å—Ç–≤–æ–º —É—á–∞—Å—Ç–Ω–∏–∫–æ–≤ –≤ –≤–æ–∑—Ä–∞—Å—Ç–µ –¥–æ 30 –ª–µ—Ç.

#### Weight (–≤–µ—Å)
–í—ã–±–æ—Ä –æ—Ü–µ–Ω–æ–∫: –î–ª—è –≤–µ—Å–∞, –∫–∞–∫ –¥–ª—è –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω–æ–≥–æ –Ω–µ–ø—Ä–µ—Ä—ã–≤–Ω–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞, —Ç–∞–∫–∂–µ –ø—Ä–∏–º–µ–Ω–∏–º—ã –≤—Å–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏–µ –æ—Ü–µ–Ω–∫–∏.

__–ò–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤__:

–ó–Ω–∞—á–∏—Ç–µ–ª—å–Ω—ã–π —Ä–∞–∑–±—Ä–æ—Å –≤ –≤–µ—Å–µ —É—á–∞—Å—Ç–Ω–∏–∫–æ–≤ —É–∫–∞–∑—ã–≤–∞–µ—Ç –Ω–∞ —Ä–∞–∑–Ω–æ—Ä–æ–¥–Ω–æ—Å—Ç—å –≤—ã–±–æ—Ä–∫–∏.

–ù–∞–ª–∏—á–∏–µ —ç–∫—Å—Ç—Ä–µ–º–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –º–æ–∂–µ—Ç —É–∫–∞–∑—ã–≤–∞—Ç—å –Ω–∞ –Ω–∞–ª–∏—á–∏–µ –ª—é–¥–µ–π —Å –∏–∑–±—ã—Ç–æ—á–Ω—ã–º –≤–µ—Å–æ–º –∏–ª–∏ –æ–∂–∏—Ä–µ–Ω–∏–µ–º.

–°—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –∏ –º–µ–¥–∏–∞–Ω–∞ –ø–æ–∑–≤–æ–ª—è—é—Ç —Å—É–¥–∏—Ç—å –æ —Ü–µ–Ω—Ç—Ä–∞–ª—å–Ω–æ–π —Ç–µ–Ω–¥–µ–Ω—Ü–∏–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –≤–µ—Å–∞.

CAEC (–ø–µ—Ä–µ–∫—É—Å—ã –º–µ–∂–¥—É –æ—Å–Ω–æ–≤–Ω—ã–º–∏ –ø—Ä–∏–µ–º–∞–º–∏ –ø–∏—â–∏)

–í—ã–±–æ—Ä –æ—Ü–µ–Ω–æ–∫: –î–ª—è –ø–æ—Ä—è–¥–∫–æ–≤–æ–≥–æ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞ –ø—Ä–∏–º–µ–Ω–∏–º—ã –º–æ–¥–∞, –º–µ–¥–∏–∞–Ω–∞ –∏ –∫–≤–∞—Ä—Ç–∏–ª–∏ (–µ—Å–ª–∏ –∑–∞–¥–∞–Ω –ø–æ—Ä—è–¥–æ–∫ –∫–∞—Ç–µ–≥–æ—Ä–∏–π).

–ò–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤:

–ú–æ–¥–∞ –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç –Ω–∞–∏–±–æ–ª–µ–µ —Ä–∞—Å–ø—Ä–æ—Å—Ç—Ä–∞–Ω–µ–Ω–Ω—É—é —á–∞—Å—Ç–æ—Ç—É –ø–µ—Ä–µ–∫—É—Å–æ–≤ - "Sometimes".

–ú–µ–¥–∏–∞–Ω–∞ –∏ –∫–≤–∞—Ä—Ç–∏–ª–∏ –ø–æ–∑–≤–æ–ª—è—é—Ç –ø–æ–Ω—è—Ç—å —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø–æ —à–∫–∞–ª–µ "Never" - "Always".

–ë–æ–ª—å—à–∏–Ω—Å—Ç–≤–æ —É—á–∞—Å—Ç–Ω–∏–∫–æ–≤ –∏–Ω–æ–≥–¥–∞ –¥–µ–ª–∞—é—Ç –ø–µ—Ä–µ–∫—É—Å—ã –º–µ–∂–¥—É –æ—Å–Ω–æ–≤–Ω—ã–º–∏ –ø—Ä–∏–µ–º–∞–º–∏ –ø–∏—â–∏.

#### FAF (–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –¥–Ω–µ–π –≤ –Ω–µ–¥–µ–ª—é —Å —Ñ–∏–∑–∏—á–µ—Å–∫–æ–π –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç—å—é)

–í—ã–±–æ—Ä –æ—Ü–µ–Ω–æ–∫: –î–ª—è –∫–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω–æ–≥–æ –¥–∏—Å–∫—Ä–µ—Ç–Ω–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞ –ø—Ä–∏–º–µ–Ω–∏–º—ã –≤—Å–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏–µ –æ—Ü–µ–Ω–∫–∏.

__–ò–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤__:

–ù–∏–∑–∫–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è –º–æ–¥—ã –∏ –º–µ–¥–∏–∞–Ω—ã —É–∫–∞–∑—ã–≤–∞—é—Ç –Ω–∞ —Ç–æ, —á—Ç–æ –±–æ–ª—å—à–∏–Ω—Å—Ç–≤–æ —É—á–∞—Å—Ç–Ω–∏–∫–æ–≤ –∏–º–µ—é—Ç –Ω–∏–∑–∫—É—é —Ñ–∏–∑–∏—á–µ—Å–∫—É—é –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç—å.

–ú–∞–ª–µ–Ω—å–∫–æ–µ –°–ö–û –≥–æ–≤–æ—Ä–∏—Ç –æ –Ω–µ–±–æ–ª—å—à–æ–º —Ä–∞–∑–±—Ä–æ—Å–µ –∑–Ω–∞—á–µ–Ω–∏–π.

–£—á–∞—Å—Ç–Ω–∏–∫–∏ –∏—Å—Å–ª–µ–¥–æ–≤–∞–Ω–∏—è –≤ —Å—Ä–µ–¥–Ω–µ–º —É–¥–µ–ª—è—é—Ç —Ñ–∏–∑–∏—á–µ—Å–∫–æ–π –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç–∏ –º–µ–Ω–µ–µ –æ–¥–Ω–æ–≥–æ –¥–Ω—è –≤ –Ω–µ–¥–µ–ª—é, —á—Ç–æ –º–æ–∂–µ—Ç —Å–≤–∏–¥–µ—Ç–µ–ª—å—Å—Ç–≤–æ–≤–∞—Ç—å –æ –º–∞–ª–æ–ø–æ–¥–≤–∏–∂–Ω–æ–º –æ–±—Ä–∞–∑–µ –∂–∏–∑–Ω–∏.

#### MTRANS (–≤–∏–¥ —Ç—Ä–∞–Ω—Å–ø–æ—Ä—Ç–∞)

–í—ã–±–æ—Ä –æ—Ü–µ–Ω–æ–∫: –î–ª—è –Ω–æ–º–∏–Ω–∞–ª—å–Ω–æ–≥–æ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω–æ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–∞ –ø—Ä–∏–º–µ–Ω–∏–º–∞ —Ç–æ–ª—å–∫–æ –º–æ–¥–∞.

–ò–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∞—Ü–∏—è —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤:

–ú–æ–¥–∞ –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç, —á—Ç–æ –Ω–∞–∏–±–æ–ª–µ–µ —Ä–∞—Å–ø—Ä–æ—Å—Ç—Ä–∞–Ω–µ–Ω–Ω—ã–º –≤–∏–¥–æ–º —Ç—Ä–∞–Ω—Å–ø–æ—Ä—Ç–∞ —Å—Ä–µ–¥–∏ —É—á–∞—Å—Ç–Ω–∏–∫–æ–≤ —è–≤–ª—è–µ—Ç—Å—è –æ–±—â–µ—Å—Ç–≤–µ–Ω–Ω—ã–π —Ç—Ä–∞–Ω—Å–ø–æ—Ä—Ç.

–ë–æ–ª—å—à–∏–Ω—Å—Ç–≤–æ —É—á–∞—Å—Ç–Ω–∏–∫–æ–≤ –ø—Ä–µ–¥–ø–æ—á–∏—Ç–∞—é—Ç –ø–µ—Ä–µ–¥–≤–∏–≥–∞—Ç—å—Å—è –Ω–∞ –æ–±—â–µ—Å—Ç–≤–µ–Ω–Ω–æ–º —Ç—Ä–∞–Ω—Å–ø–æ—Ä—Ç–µ, —á—Ç–æ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É–µ—Ç –ª–∞—Ç–∏–Ω–æ–∞–º–µ—Ä–∏–∫–∞–Ω—Å–∫–∏–º —Ä–µ–∞–ª–∏—è–º –∏ –≥–æ—Ä–æ–¥—Å–∫–æ–º—É –æ–±—Ä–∞–∑—É –∂–∏–∑–Ω–∏.

#### –í—ã–≤–æ–¥—ã

–ü—Ä–æ–≤–µ–¥–µ–Ω–Ω—ã–π –∞–Ω–∞–ª–∏–∑ –¥–∞–Ω–Ω—ã—Ö –æ —Ñ–∏–∑–∏—á–µ—Å–∫–æ–º —Å–æ—Å—Ç–æ—è–Ω–∏–∏ –∏ –æ–±—Ä–∞–∑–µ –∂–∏–∑–Ω–∏ –ª—é–¥–µ–π –∏–∑ –ª–∞—Ç–∏–Ω–æ–∞–º–µ—Ä–∏–∫–∞–Ω—Å–∫–∏—Ö —Å—Ç—Ä–∞–Ω –ø–æ–∫–∞–∑–∞–ª, —á—Ç–æ –≤—ã–±–æ—Ä–∫–∞ –ø—Ä–µ–∏–º—É—â–µ—Å—Ç–≤–µ–Ω–Ω–æ —Å–æ—Å—Ç–æ–∏—Ç –∏–∑ –º–æ–ª–æ–¥—ã—Ö –ª—é–¥–µ–π, –∫–æ—Ç–æ—Ä—ã–µ –∏–Ω–æ–≥–¥–∞ –ø–µ—Ä–µ–∫—É—Å—ã–≤–∞—é—Ç –º–µ–∂–¥—É –æ—Å–Ω–æ–≤–Ω—ã–º–∏ –ø—Ä–∏–µ–º–∞–º–∏ –ø–∏—â–∏, –∏–º–µ—é—Ç –Ω–∏–∑–∫—É—é —Ñ–∏–∑–∏—á–µ—Å–∫—É—é –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç—å –∏ –ø–æ–ª—å–∑—É—é—Ç—Å—è –æ–±—â–µ—Å—Ç–≤–µ–Ω–Ω—ã–º —Ç—Ä–∞–Ω—Å–ø–æ—Ä—Ç–æ–º. –í–µ—Å —É—á–∞—Å—Ç–Ω–∏–∫–æ–≤ –∏–º–µ–µ—Ç –∑–Ω–∞—á–∏—Ç–µ–ª—å–Ω—ã–π —Ä–∞–∑–±—Ä–æ—Å, —á—Ç–æ –º–æ–∂–µ—Ç –±—ã—Ç—å —Å–≤—è–∑–∞–Ω–æ —Å —Ä–∞–∑–ª–∏—á–Ω—ã–º–∏ —Ñ–∞–∫—Ç–æ—Ä–∞–º–∏, –≤–∫–ª—é—á–∞—è –ø–∏—â–µ–≤—ã–µ –ø—Ä–∏–≤—ã—á–∫–∏ –∏ —Ñ–∏–∑–∏—á–µ—Å–∫—É—é –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç—å.
"""))
```
